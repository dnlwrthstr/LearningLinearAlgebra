# Vector operations

You already have an understanding of the notion of vectors and why they are useful. However, vectors become truly powerful when we can **perform operations on them**.

Consider real numbers that represent distances between cities. Knowing the distances alone is not enough if we want to compute something practical—such as the **shortest route** between locations. To achieve this, we rely on arithmetic operations.

The same idea applies to vectors. Vectors are most useful when we can **combine, transform, and manipulate** them to obtain new vectors. These operations allow us to model real-world phenomena and solve practical problems.

This capability is fundamental not only in core areas of mathematics such as **linear algebra** and **calculus**, but also in modern fields of computer science, including:

- **Machine learning**
- **Video game development**
- **Computer graphics and graphic design**

In all these areas, vector operations are essential tools for representing motion, transformations, optimization, and structure.

## Vector operations

Like real numbers, vectors have basic operations such as **addition** and **multiplication**, with subtraction and division acting as their respective inverses (where applicable). Let’s examine these vector operations in detail.


## Addition

To add two vectors, they must have the **same dimension**, meaning they contain the same number of elements. If vectors are of different sizes, addition is not defined.

Vector addition is performed **component-wise**.

Let
$$
\vec{a} = (2,\,3,\,-5), \qquad
\vec{b} = (-1,\,0,\,2).
$$

Their sum is the vector $\vec{c}$:
$$
\vec{c} = \vec{a} + \vec{b}
= (2,3,-5) + (-1,0,2)
$$

$$
= (2+(-1),\,3+0,\,-5+2)
= (1,\,3,\,-3).
$$

As you can see, vector addition is straightforward.
Vector **subtraction** works the same way, since subtraction is simply addition with a negative sign.

## Multiplication by a scalar

Scalar multiplication means multiplying **each component** of a vector by a real number.

Let
$$
\vec{a} = (7,\,2,\,0,\,4,\,1), \qquad \lambda = 2.
$$

Then
$$
\vec{c} = \lambda \cdot \vec{a}
= 2 \cdot (7,2,0,4,1)
= (14,\,4,\,0,\,8,\,2).
$$

### Special cases

**Multiplication by zero**
$$
0 \cdot \vec{a}
= (0\cdot7,\,0\cdot2,\,0\cdot0,\,0\cdot4,\,0\cdot1)
= (0,\,0,\,0,\,0,\,0).
$$

Any vector multiplied by $0$ becomes the **zero vector**.

**Zero vector multiplied by a scalar**
$$
5 \cdot (0,0,0)
= (5\cdot0,\,5\cdot0,\,5\cdot0)
= (0,0,0).
$$

The zero vector remains unchanged under scalar multiplication.

> Note: Vector-to-vector multiplication does exist (such as dot and cross products), but it will be discussed later.

## Combining vector operations

Vector operations obey the same algebraic rules as real numbers.

Let
$$
\vec{a} = (4,\,-5,\,-1), \quad
\vec{b} = (3,\,3,\,33), \quad
\vec{c} = (-7,\,8,\,-13).
$$

Compute
$$
3\vec{a} + 2\vec{b} + 4\vec{c}.
$$

Step by step:
$$
3\vec{a} = (12,\,-15,\,-3),
$$
$$
2\vec{b} = (6,\,6,\,66),
$$
$$
4\vec{c} = (-28,\,32,\,-52).
$$

Now add them:
$$
(12,-15,-3) + (6,6,66) + (-28,32,-52)
$$

$$
= (12+6-28,\,-15+6+32,\,-3+66-52)
$$

$$
= (-10,\,23,\,11).
$$

So the final result is
$$
3\vec{a} + 2\vec{b} + 4\vec{c} = (-10,\,23,\,11).
$$

## Geometric interpretation of operations

Let’s look at the **geometric meaning** behind basic vector operations.

## Vector addition

As discussed earlier, the sum of two vectors $\vec{a}$ and $\vec{b}$ is itself a vector.
Geometrically, vector addition can be visualized using two equivalent rules:

### Triangle (head-to-tail) law

The **triangle law** is used when the **head of the first vector** $\vec{a}$ is placed at the **tail of the second vector** $\vec{b}$.

- Place the tail of $\vec{b}$ at the head of $\vec{a}$
- The resulting vector $\vec{a} + \vec{b}$ goes from the tail of $\vec{a}$ to the head of $\vec{b}$

This method is also called the **head-to-tail rule** and is often the easiest to remember.

### Parallelogram law

The **parallelogram law** is used when both vectors share the **same tail**.

- Draw $\vec{a}$ and $\vec{b}$ starting from the same point
- Complete a parallelogram by drawing lines parallel to each vector
- The diagonal of the parallelogram starting at the common tail represents
  the vector $\vec{a} + \vec{b}$

Both constructions give the **same resulting vector**, confirming that vector addition is independent of how vectors are placed, as long as their direction and magnitude are preserved.

## Multiplication by a scalar

When a vector $\vec{a}$ is multiplied by a scalar $\lambda$, the result is a new vector $\lambda \vec{a}$.

Geometrically:

- If $\lambda > 0$, the vector keeps the **same direction** as $\vec{a}$
- If $\lambda < 0$, the vector points in the **opposite direction**
- The **length** of the vector is scaled by a factor of $|\lambda|$

### Example

Let
$$
\vec{a} = (2,\,1).
$$

Multiplying by $2$:
$$
2\vec{a} = (4,\,2),
$$
which doubles the length of the vector while keeping its direction.

Multiplying by $-\tfrac{1}{2}$:
$$
-\tfrac{1}{2}\vec{a} = (-1,\,-\tfrac{1}{2}),
$$
which reverses the direction of $\vec{a}$ and scales its length by a factor of $\tfrac{1}{2}$.

### Geometric takeaway

- **Vector addition** corresponds to chaining movements in space
- **Scalar multiplication** stretches, shrinks, or reverses vectors
- These geometric interpretations are foundational for physics, computer graphics, machine learning, and game development

```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
a = np.array([2, 1])
b = np.array([1, 3])

# Plot setup
plt.figure()
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

# Draw vector a
plt.arrow(0, 0, a[0], a[1], length_includes_head=True, head_width=0.1)
plt.text(a[0], a[1], r"$\vec a$")

# Draw vector b from head of a
plt.arrow(a[0], a[1], b[0], b[1], length_includes_head=True, head_width=0.1)
plt.text(a[0] + b[0], a[1] + b[1], r"$\vec b$")

# Draw result vector a + b
plt.arrow(0, 0, a[0] + b[0], a[1] + b[1],
          length_includes_head=True, linestyle="dashed", head_width=0.1)
plt.text(a[0] + b[0], a[1] + b[1] + 0.1, r"$\vec a + \vec b$")

plt.title("Vector Addition (Triangle / Head-to-Tail Law)")
plt.axis("equal")
plt.show()
```

```python
plt.figure()
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

# Draw vectors from the same origin
plt.arrow(0, 0, a[0], a[1], length_includes_head=True, head_width=0.1)
plt.arrow(0, 0, b[0], b[1], length_includes_head=True, head_width=0.1)

# Parallelogram sides (dashed)
plt.arrow(a[0], a[1], b[0], b[1], linestyle="dashed", head_width=0)
plt.arrow(b[0], b[1], a[0], a[1], linestyle="dashed", head_width=0)

# Result vector
plt.arrow(0, 0, a[0] + b[0], a[1] + b[1],
          length_includes_head=True, color="black", head_width=0.1)

plt.text(a[0], a[1], r"$\vec a$")
plt.text(b[0], b[1], r"$\vec b$")
plt.text(a[0] + b[0], a[1] + b[1], r"$\vec a + \vec b$")

plt.title("Vector Addition (Parallelogram Law)")
plt.axis("equal")
plt.show()
```

```python
# Original vector
a = np.array([2, 1])

# Scalars
scalars = [2, -0.5]

plt.figure()
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

# Original vector
plt.arrow(0, 0, a[0], a[1], length_includes_head=True, head_width=0.1)
plt.text(a[0], a[1], r"$\vec a$")

# Scaled vectors
for k in scalars:
    scaled = k * a
    plt.arrow(0, 0, scaled[0], scaled[1],
              length_includes_head=True, linestyle="dashed", head_width=0.1)
    plt.text(scaled[0], scaled[1], rf"${k}\vec a$")

plt.title("Scalar Multiplication of a Vector")
plt.axis("equal")
plt.show()
```

```python
a = np.array([4, -5])
b = np.array([3, 2])

result = 3*a + 2*b

plt.figure()
plt.axhline(0)
plt.axvline(0)
plt.grid(True)

plt.arrow(0, 0, a[0], a[1], head_width=0.1)
plt.arrow(0, 0, b[0], b[1], head_width=0.1)
plt.arrow(0, 0, result[0], result[1], head_width=0.1, linestyle="dashed")

plt.text(a[0], a[1], r"$\vec a$")
plt.text(b[0], b[1], r"$\vec b$")
plt.text(result[0], result[1], r"$3\vec a + 2\vec b$")

plt.title("Linear Combination of Vectors")
plt.axis("equal")
plt.show()
```

# Properties of Vector Operations

As far as properties resulting from the discussed operations are concerned, let's point out the three main ones.

## Commutativity of Addition

This one is simple: it doesn't matter in what order we add vectors. That means

$$
\vec a + \vec b = \vec b + \vec a
$$

It is easy to check: we add the corresponding coordinates as numbers and do not care in what order we perform that operation.


## Associativity of Addition

This property is related to the order of parentheses. The main idea is that during the addition of more than two vectors, it is not important how the parentheses are arranged:

$$
\vec a + (\vec b + \vec c) = (\vec a + \vec b) + \vec c
$$

It is also easy to check if you simply calculate the expression.

You can also easily prove it using the triangle law. First, focus on the left side of the statement. When you add $\vec c$ to $\vec b$ using the triangle law, you obtain the yellow vector $\vec b + \vec c$. After that, apply the triangle law once more to add this vector to $\vec a$, which gives the black vector

$$
\vec a + (\vec b + \vec c).
$$

Nearly the same process is done for the right side of the statement. You add $\vec b$ to $\vec a$ and obtain the yellow vector $\vec a + \vec b$. Afterward, you add $\vec c$ to this vector. Once again, you obtain the same black vector

$$
\vec a + \vec b + \vec c.
$$

**Proof of associativity of addition**

## Distributivity of Scalar Multiplication over Addition

This property shows the rule for opening parentheses during scalar multiplication. The rule is:

$$
\lambda \cdot (\vec a + \vec b) = \lambda \vec a + \lambda \vec b
$$

To verify it, you only need to calculate the values on both sides of the equation and see that they are exactly the same. There is also a geometric proof of this property.

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# Vector definitions
# -----------------------
a = np.array([3.5, 1.0])
b = np.array([1.2, 3.0])
lam = 2.0

a_b = a + b
lam_a = lam * a
lam_b = lam * b
lam_ab = lam * a_b

# -----------------------
# Plot setup
# -----------------------
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_aspect("equal")
ax.axis("off")

# -----------------------
# Helper function
# -----------------------
def draw_vec(start, vec, color, lw=2.5, style="-", z=3):
    ax.arrow(
        start[0], start[1],
        vec[0], vec[1],
        head_width=0.18,
        head_length=0.28,
        length_includes_head=True,
        linewidth=lw,
        linestyle=style,
        color=color,
        zorder=z
    )

def label_vec(start, vec, text, offset=(0,0)):
    mid = start + 0.55 * vec + np.array(offset)
    ax.text(mid[0], mid[1], text, fontsize=14, color="white")

# -----------------------
# Colors
# -----------------------
blue = "#4C8DFF"
orange = "#D28B5C"
red = "#FF4C4C"
gray = "#BBBBBB"

# -----------------------
# Base vectors
# -----------------------
draw_vec((0,0), a, blue)
draw_vec((0,0), b, orange)

label_vec((0,0), a, r"$\vec a$", offset=(0.1,-0.15))
label_vec((0,0), b, r"$\vec b$", offset=(-0.25,0.1))

# -----------------------
# a + b
# -----------------------
draw_vec((0,0), a_b, red, lw=3.2)
label_vec((0,0), a_b, r"$\vec a+\vec b$", offset=(0.1,0.1))

# -----------------------
# λa and λb
# -----------------------
draw_vec((0,0), lam_a, blue, style="--", lw=2.3)
draw_vec((0,0), lam_b, orange, style="--", lw=2.3)

label_vec((0,0), lam_a, r"$\lambda\vec a$", offset=(0.15,-0.2))
label_vec((0,0), lam_b, r"$\lambda\vec b$", offset=(-0.3,0.15))

# -----------------------
# λ(a + b)
# -----------------------
draw_vec((0,0), lam_ab, red, style=":", lw=3)
label_vec((0,0), lam_ab, r"$\lambda(\vec a+\vec b)$", offset=(0.1,0.15))

# -----------------------
# Parallelogram construction
# -----------------------
draw_vec(lam_a, lam_b, orange, style=":", lw=2)
draw_vec(lam_b, lam_a, blue, style=":", lw=2)

# -----------------------
# Limits
# -----------------------
pts = np.array([
    [0,0], a, b, a_b, lam_a, lam_b, lam_ab
])
pad = 1.2
ax.set_xlim(pts[:,0].min()-pad, pts[:,0].max()+pad)
ax.set_ylim(pts[:,1].min()-pad, pts[:,1].max()+pad)

plt.show()
```
