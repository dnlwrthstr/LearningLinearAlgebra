# Euclidean Spaces

The dot product is a magnificent tool when it comes to $\mathbb{R}^n$. We saw how it seamlessly connects geometric notions with familiar algebraic manipulations. This naturally leads to the question: could we define a similar operation for an arbitrary vector space $V$?

When we first defined the dot product, we relied on the standard basis of $\mathbb{R}^n$. This raises an even more fundamental question: can such an operation be defined without referring to any specific basis at all?

The answer to both questions is yes. In fact, equipping a vector space with this new operation leads to a remarkable mathematical structure. This construction is at least as powerful for expressing geometric concepts in algebraic terms as $\mathbb{R}^n$ endowed with the dot product.

## Increasing the Abstraction Level

The dot product was so great because of its valuable features and properties. There are three of them which are the most important for us.

Let's say that $V$ is an arbitrary real $n$-dimensional vector space, and $\vec v$ and $\vec w$ are any two vectors from $V$. We call an operation that takes $\vec v$ and $\vec w$ and outputs a real number an **inner product** if it is **bilinear**, **symmetric**, and **positive definite**.

To denote the inner product, we use the angle-bracket notation $\langle \vec v, \vec w \rangle$. Here is what bilinear, symmetric, and positive definite mean.

### Bilinearity

For any vectors $\vec a_1, \vec a_2, \vec b_1, \vec b_2 \in V$ and scalars $\alpha_1, \alpha_2, \beta_1, \beta_2 \in \mathbb{R}$,

$$
\langle \alpha_1 \vec a_1 + \alpha_2 \vec a_2,\; \beta_1 \vec b_1 + \beta_2 \vec b_2 \rangle
=
\alpha_1 \beta_1 \langle \vec a_1, \vec b_1 \rangle
+
\alpha_1 \beta_2 \langle \vec a_1, \vec b_2 \rangle
+
\alpha_2 \beta_1 \langle \vec a_2, \vec b_1 \rangle
+
\alpha_2 \beta_2 \langle \vec a_2, \vec b_2 \rangle .
$$

### Symmetry

For all $\vec a, \vec b \in V$,

$$
\langle \vec a, \vec b \rangle = \langle \vec b, \vec a \rangle .
$$

### Positive Definiteness

For all $\vec a \in V$,

$$
\langle \vec a, \vec a \rangle \ge 0 .
$$

Note that these three properties hold for the dot product. Therefore, the dot product is an example of an inner product.

## Inner Products of Basis Vectors Are the Key

In the case of the dot product, the inner products of the basis vectors with themselves were equal to $1$, and with the other basis vectors equal to $0$.
For a general inner product, we have much more freedom in choosing these values. However, the inner products of basis vectors cannot be completely arbitrary.

Consider a two-dimensional real vector space $V$ with an arbitrary basis
$$
\{\vec e_1, \vec e_2\}.
$$

Suppose the inner products of the basis vectors are given by
$$
\langle \vec e_1, \vec e_1 \rangle = a, \qquad
\langle \vec e_1, \vec e_2 \rangle = a, \qquad
\langle \vec e_2, \vec e_2 \rangle = b,
$$
where $a$ and $b$ are real numbers.

Now consider two arbitrary vectors
$$
\vec v = v_1 \vec e_1 + v_2 \vec e_2,
\qquad
\vec w = w_1 \vec e_1 + w_2 \vec e_2 .
$$

By bilinearity,
$$
\begin{aligned}
\langle \vec v, \vec w \rangle
&= \langle v_1 \vec e_1 + v_2 \vec e_2,\; w_1 \vec e_1 + w_2 \vec e_2 \rangle \\
&= v_1 w_1 \langle \vec e_1, \vec e_1 \rangle
+ v_1 w_2 \langle \vec e_1, \vec e_2 \rangle
+ v_2 w_1 \langle \vec e_2, \vec e_1 \rangle
+ v_2 w_2 \langle \vec e_2, \vec e_2 \rangle .
\end{aligned}
$$

Using symmetry, $\langle \vec e_2, \vec e_1 \rangle = \langle \vec e_1, \vec e_2 \rangle$, this becomes
$$
\langle \vec v, \vec w \rangle
=
v_1 w_1 \langle \vec e_1, \vec e_1 \rangle
+
(v_1 w_2 + v_2 w_1)\langle \vec e_1, \vec e_2 \rangle
+
v_2 w_2 \langle \vec e_2, \vec e_2 \rangle .
$$

Substituting the given values,
$$
\langle \vec v, \vec w \rangle
=
(v_1 w_1 + v_1 w_2 + v_2 w_1)a + v_2 w_2 b .
$$

Now let $\vec v = \vec w$. Then
$$
\langle \vec v, \vec v \rangle
=
a v_1^2 + 2a v_1 v_2 + b v_2^2 .
$$

Since the inner product is positive definite, we must have
$$
\langle \vec v, \vec v \rangle \ge 0
\quad \text{for all } \vec v .
$$

Rewrite the expression as
$$
\begin{aligned}
\langle \vec v, \vec v \rangle
&= a v_1^2 + 2a v_1 v_2 + a v_2^2 + (b - a)v_2^2 \\
&= a (v_1 + v_2)^2 + (b - a)v_2^2 \ge 0 .
\end{aligned}
$$

For this inequality to hold for all $v_1, v_2$, we must have
$$
a \ge 0
\quad \text{and} \quad
b \ge a,
$$
with $a$ and $b$ not both equal to $0$.

Therefore, once the inner products of the basis vectors are known, the inner product of any two vectors can be computed from their coordinates in that basis. The only requirement is that these values satisfy the defining properties of an inner product.

## Euclidean Space

The main intuition behind the dot product was its deep connection with the geometry of $n$-dimensional space. This intuition can be generalized with the help of an inner product.

Let $V$ be a vector space and let $\langle \cdot , \cdot \rangle$ be an inner product on $V$.
The pair
$$
E = (V, \langle \cdot , \cdot \rangle)
$$
that is, a vector space equipped with an inner product, is called a **Euclidean space**.

It is no coincidence that the name of the ancient geometer appears here. The reason is that, in Euclidean spaces, every geometric statement we are familiar with continues to hold true. Promising, isn’t it?

For all of this to work, however, we first need to learn how to *measure* things in Euclidean spaces. Fortunately, with the dot product as a guiding example, this will not be difficult.

### Length of a Vector

The length (or norm) of a vector $\vec v$ in a Euclidean space is defined as the square root of its *inner square*:
$$
\|\vec v\| = \sqrt{\langle \vec v, \vec v \rangle}.
$$

(This is precisely why we required the inner product to be positive definite.)

### Angle Between Vectors

The cosine of the angle $\angle(\vec v, \vec w)$ between two vectors $\vec v$ and $\vec w$ is defined by
$$
\cos\bigl(\angle(\vec v, \vec w)\bigr)
=
\frac{\langle \vec v, \vec w \rangle}{\|\vec v\| \cdot \|\vec w\|}.
$$

The angle itself is then given by the inverse cosine:
$$
\angle(\vec v, \vec w)
=
\arccos\!\left(
\frac{\langle \vec v, \vec w \rangle}{\|\vec v\| \cdot \|\vec w\|}
\right).
$$

### Orthogonality

Two vectors $\vec v$ and $\vec w$ in a Euclidean space are called **orthogonal** if their inner product is zero:
$$
\langle \vec v, \vec w \rangle = 0,
$$
which is commonly denoted by
$$
\vec v \perp \vec w.
$$

## Geometry Without Visualization

Let $V$ be a two-dimensional vector space with a basis $\{\vec e_1, \vec e_2\}$. The following relations

$$
\langle \vec e_1, \vec e_1 \rangle = 2,
\qquad
\langle \vec e_1, \vec e_2 \rangle = -1,
\qquad
\langle \vec e_2, \vec e_2 \rangle = 1
$$

define an inner product $\langle \cdot , \cdot \rangle$ (it’s not hard to check that this is indeed an inner product, using the above-mentioned argument).

Let’s find the angle between the vectors

$$
\vec v = \vec e_1 + \vec e_2,
\qquad
\vec w = \vec e_1 + 2\vec e_2
$$

in the Euclidean space

$$
E = (V, \langle \cdot , \cdot \rangle).
$$

Before solving this problem, we want to emphasize the main idea: the solution is **purely algebraic**. There is no specific geometric picture behind $V$, and $\{\vec e_1, \vec e_2\}$ is a completely arbitrary basis: its vectors are not necessarily perpendicular, and $\vec e_1$ has length $\sqrt{2}$ in $E$. This example is more abstract than the ones considered in the dot product discussion.

One could ask why we need such a hypothetical construction. The answer is **universality**: this algebraic approach allows us to introduce geometric meaning in any setting where vector spaces appear—for example, in vector analysis, statistics, or model building. Euclidean spaces are tools, and the concrete interpretation comes from the particular problem.

### Step 1: Compute the lengths of $\vec v$ and $\vec w$

First,

$$
\begin{aligned}
\|\vec v\|^2
&= \langle \vec v, \vec v \rangle \\
&= \langle \vec e_1 + \vec e_2,\; \vec e_1 + \vec e_2 \rangle \\
&= \langle \vec e_1, \vec e_1 \rangle
+ 2\langle \vec e_1, \vec e_2 \rangle
+ \langle \vec e_2, \vec e_2 \rangle \\
&= 2 + 2(-1) + 1 \\
&= 1.
\end{aligned}
$$

So,

$$
\|\vec v\| = \sqrt{1} = 1.
$$

Next,

$$
\begin{aligned}
\|\vec w\|^2
&= \langle \vec w, \vec w \rangle \\
&= \langle \vec e_1 + 2\vec e_2,\; \vec e_1 + 2\vec e_2 \rangle \\
&= \langle \vec e_1, \vec e_1 \rangle
+ 4\langle \vec e_1, \vec e_2 \rangle
+ 4\langle \vec e_2, \vec e_2 \rangle \\
&= 2 + 4(-1) + 4 \\
&= 2.
\end{aligned}
$$

So,

$$
\|\vec w\| = \sqrt{2}.
$$

### Step 2: Compute $\langle \vec v, \vec w \rangle$

$$
\begin{aligned}
\langle \vec v, \vec w \rangle
&= \langle \vec e_1 + \vec e_2,\; \vec e_1 + 2\vec e_2 \rangle \\
&= \langle \vec e_1, \vec e_1 \rangle
+ 2\langle \vec e_1, \vec e_2 \rangle
+ \langle \vec e_2, \vec e_1 \rangle
+ 2\langle \vec e_2, \vec e_2 \rangle \\
&= 2 + 2(-1) + (-1) + 2(1) \\
&= 1.
\end{aligned}
$$

### Step 3: Compute the angle

Using

$$
\cos\bigl(\angle(\vec v, \vec w)\bigr)
=
\frac{\langle \vec v, \vec w \rangle}{\|\vec v\|\;\|\vec w\|},
$$

we get

$$
\cos\bigl(\angle(\vec v, \vec w)\bigr)
=
\frac{1}{1\cdot \sqrt{2}}
=
\frac{1}{\sqrt{2}}.
$$

Therefore,

$$
\angle(\vec v, \vec w)
=
\arccos\left(\frac{1}{\sqrt{2}}\right)
=
45^\circ.
$$

## Which Functions Define an Inner Product?

Given an $n$-dimensional vector space $V$ with a basis
$$
\{\vec e_1, \vec e_2, \dots, \vec e_n\},
$$
any vector
$$
\vec v = x_1 \vec e_1 + x_2 \vec e_2 + \dots + x_n \vec e_n
$$
can be represented by its coordinate tuple
$$
(x_1, x_2, \dots, x_n)^T \in \mathbb{R}^n.
$$

This coordinate-based notation is often more natural for concrete problems. In such situations, it is usually more convenient to define the inner product directly as a function of coordinates.

This may sound abstract, so let us look at a two-dimensional example. An inner product on $\mathbb{R}^2$ can be defined as

$$
\langle \vec x, \vec y \rangle
=
\left\langle
\begin{pmatrix} x_1 \\ x_2 \end{pmatrix},
\begin{pmatrix} y_1 \\ y_2 \end{pmatrix}
\right\rangle
=
x_1 y_1 - x_1 y_2 - x_2 y_1 + 2 x_2 y_2 .
$$

It is straightforward to verify that this function is **symmetric** and **bilinear** with respect to the vectors $\vec x$ and $\vec y$. What remains is to check that it is **positive definite**.

### Positive Definiteness

Consider the inner product of a vector with itself:

$$
\begin{aligned}
\langle \vec x, \vec x \rangle
&=
\left\langle
\begin{pmatrix} x_1 \\ x_2 \end{pmatrix},
\begin{pmatrix} x_1 \\ x_2 \end{pmatrix}
\right\rangle \\
&=
x_1^2 - 2 x_1 x_2 + 2 x_2^2 .
\end{aligned}
$$

Now rewrite this expression:

$$
\begin{aligned}
x_1^2 - 2 x_1 x_2 + 2 x_2^2
&=
(x_1^2 - 2 x_1 x_2 + x_2^2) + x_2^2 \\
&=
(x_1 - x_2)^2 + x_2^2 \\
&\ge 0 .
\end{aligned}
$$

This quantity is nonnegative for all $(x_1, x_2)$ and equals zero if and only if
$$
x_1 = x_2 = 0.
$$

Therefore, the inner product is positive definite, and all axioms of an inner product are satisfied.

### Computing an Inner Product

To compute the inner product of two vectors, simply substitute their coordinates into the defining formula. For example,

$$
\begin{aligned}
\left\langle
\begin{pmatrix} -7 \\ 3 \end{pmatrix},
\begin{pmatrix} 13 \\ 10 \end{pmatrix}
\right\rangle
&=
(-7)(13) - (-7)(10) - 3(13) + 2 \cdot 3 \cdot 10 \\
&=
-91 + 70 - 39 + 60 \\
&=
0.
\end{aligned}
$$

Hence, the vectors
$$
(-7, 3)^T \quad \text{and} \quad (13, 10)^T
$$
are **orthogonal** in $V$ with respect to this inner product.

## Conclusion

- An **inner product** is a bilinear, symmetric, and positive definite operation that takes two vectors and returns a real number.

- The **dot product** is the simplest and most familiar example of an inner product.

- To define an inner product, no specific basis is required. However, for explicit calculations, it is often convenient to work with the inner products of basis vectors.

- A vector space equipped with a chosen inner product is called a **Euclidean space**. In such a space, one can meaningfully compute lengths and angles.

- The formulas for the length of a vector and the angle between two vectors are exactly the same as in the case of the dot product:

  $$
  \|\vec v\| = \sqrt{\langle \vec v, \vec v \rangle},
  $$

  $$
  \angle(\vec v, \vec w)
  =
  \arccos\!\left(
  \frac{\langle \vec v, \vec w \rangle}{\|\vec v\| \cdot \|\vec w\|}
  \right).
  $$

- An inner product can also be defined by specifying a suitable function of the coordinates of vectors, provided that this function is bilinear, symmetric, and positive definite.
