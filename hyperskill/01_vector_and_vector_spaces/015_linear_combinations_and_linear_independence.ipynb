# Linear combinations and linear independence

## Introduction: Linear Dependence

You are already familiar with vectors and the key role they play in linear algebra and many other fields. A **vector** is an object that has both **magnitude** and **direction**. Vectors are useful for describing quantities such as velocity, force, or displacement—for example, the velocity of a vehicle in a video game.

However, vectors are rarely studied in isolation. In most applications, we are interested in how vectors **relate to one another** and how they **interact**. For instance, we may want to model whether the velocity of the wind affects the velocity of a car, or whether one physical quantity can be expressed in terms of others.

This is where the concept of **linear dependence** becomes essential. Linear dependence allows us to determine whether one vector can be expressed as a combination of others, helping us understand relationships between vectors.

The idea of linear dependence is a **fundamental concept** in linear algebra and plays a central role in mathematics, physics, engineering, computer graphics, data analysis, and machine learning.

## What Is Linear Dependence?

Let’s try to understand the concept of **linear dependence** through an analogy.

Suppose you enjoy painting and want to buy some paints. Since you have a limited budget, you want to choose your colors wisely.

Does it make sense to buy **red**, **yellow**, and **orange** paint? Not really—because **orange can be obtained by mixing red and yellow**, buying orange is redundant. On the other hand, if you buy **red**, **yellow**, and **blue**, none of these colors can be created by mixing the others. In this case, each color contributes something new.

- **Red, yellow, and orange** form a **dependent** set of colors.
- **Red, yellow, and blue** form an **independent** set of colors.


### From Colors to Vectors

This idea transfers directly to vectors.

A set of vectors is called **linearly dependent** if **at least one vector can be expressed as a linear combination of the others**. Such a vector is redundant—it does not add any new information, because it can be constructed from the remaining vectors.

Conversely, a set of vectors is **linearly independent** if **no vector in the set can be written as a linear combination of the others**. Each vector contributes a new “direction” or degree of freedom.

In the next section, we will formalize these ideas using **linear combinations** and provide precise mathematical definitions of **linear dependence** and **linear independence**.

## Formal Definition

We are now ready to define the key concepts formally.

Consider vectors
$$
\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n
$$
and scalars
$$
c_1, c_2, \dots, c_n.
$$

The vector
$$
\mathbf{w} = c_1 \mathbf{v}_1 + c_2 \mathbf{v}_2 + \cdots + c_n \mathbf{v}_n
$$
is called a **linear combination** of the vectors
$\mathbf{v}_1, \mathbf{v}_2, \dots, \mathbf{v}_n$.

### Trivial and Non-trivial Linear Combinations

- If **all scalars are zero**, the linear combination is called **trivial**.
- If **at least one scalar is non-zero**, the linear combination is called **non-trivial**.

For example, the following is a **non-trivial** linear combination:
$$
1 \cdot
\begin{pmatrix}
2 \\ 3
\end{pmatrix}
-
3 \cdot
\begin{pmatrix}
1 \\ 1.5
\end{pmatrix}
=
\begin{pmatrix}
-1 \\ -1.5
\end{pmatrix}.
$$


### Linear Dependence

A set of vectors is **linearly dependent** if **at least one vector can be expressed as a non-trivial linear combination of the others**.

Formally, vectors
$$
\mathbf{x}_1, \mathbf{x}_2, \dots, \mathbf{x}_n
$$
are **linearly dependent** if there exist scalars
$$
c_1, c_2, \dots, c_n,
$$
not all zero, such that
$$
c_1 \mathbf{x}_1 + c_2 \mathbf{x}_2 + \cdots + c_n \mathbf{x}_n = \mathbf{0}.
$$

### Example (Linear Dependence)

Consider the vectors
$$
\mathbf{x}_1 =
\begin{pmatrix}
2 \\ 3
\end{pmatrix},
\quad
\mathbf{x}_2 =
\begin{pmatrix}
1 \\ 1.5
\end{pmatrix},
\quad
\mathbf{x}_3 =
\begin{pmatrix}
3 \\ 2
\end{pmatrix}.
$$

They are linearly dependent because there exist scalars
$$
c_1 = 1, \quad c_2 = -2, \quad c_3 = 0
$$
such that
$$
1\mathbf{x}_1 - 2\mathbf{x}_2 + 0\mathbf{x}_3
=
\begin{pmatrix}
0 \\ 0
\end{pmatrix}.
$$

### Linear Independence

A set of vectors is **linearly independent** if the **only** scalars that satisfy
$$
c_1 \mathbf{x}_1 + c_2 \mathbf{x}_2 + \cdots + c_n \mathbf{x}_n = \mathbf{0}
$$
are
$$
c_1 = c_2 = \cdots = c_n = 0.
$$

In other words, **only the trivial linear combination produces the zero vector**.

### Example (Linear Independence)

The vectors
$$
\begin{pmatrix}
2 \\ 3
\end{pmatrix}
\quad \text{and} \quad
\begin{pmatrix}
3 \\ 10
\end{pmatrix}
$$
are linearly independent, because the equation
$$
c_1
\begin{pmatrix}
2 \\ 3
\end{pmatrix}
+
c_2
\begin{pmatrix}
3 \\ 10
\end{pmatrix}
=
\begin{pmatrix}
0 \\ 0
\end{pmatrix}
$$
has **only the trivial solution**
$$
c_1 = c_2 = 0.
$$

## Geometric Intuition

There is a very intuitive geometric interpretation of **linear dependence**.

### Linear Dependence in the Plane (2D)

In geometric terms, **two non-zero vectors in the plane are linearly dependent if and only if they are collinear**, meaning they lie on the same straight line.

Algebraically, two vectors $\vec p$ and $\vec q$ are collinear if there exists a scalar $n$ such that
$$
\vec p = n \cdot \vec q.
$$

### Example

![Linear Combination Example](img/linear_combination_example.png)


Consider the vectors
$$
\vec{AB} =
\begin{pmatrix}
1 \\ 2
\end{pmatrix},
\quad
\vec{AC} =
\begin{pmatrix}
2 \\ 4
\end{pmatrix},
\quad
\vec{AD} =
\begin{pmatrix}
2 \\ 2
\end{pmatrix}.
$$

- $\vec{AB}$ and $\vec{AC}$ lie on the same line and are therefore **linearly dependent**:
  $$
  \vec{AC} = 2\,\vec{AB}, \qquad \vec{AB} = \tfrac{1}{2}\vec{AC}.
  $$
- $\vec{AD}$ is **not collinear** with the other two vectors and therefore is **linearly independent** from them.

No scalar multiple of $\vec{AB}$ or $\vec{AC}$ can produce $\vec{AD}$.

### Why Any Three Vectors in 2D Are Dependent

Let three vectors in $\mathbb{R}^2$ be
$$
\vec x =
\begin{pmatrix}
x_1 \\ x_2
\end{pmatrix},
\quad
\vec y =
\begin{pmatrix}
y_1 \\ y_2
\end{pmatrix},
\quad
\vec z =
\begin{pmatrix}
z_1 \\ z_2
\end{pmatrix}.
$$

There are **three unknown coefficients** $c_1, c_2, c_3$ but only **two equations**:
$$
\begin{cases}
c_1 x_1 + c_2 y_1 + c_3 z_1 = 0 \\
c_1 x_2 + c_2 y_2 + c_3 z_2 = 0
\end{cases}
$$

Such a system always has a **non-trivial solution**, meaning
$$
c_1 \vec x + c_2 \vec y + c_3 \vec z = \vec 0
$$
for some coefficients not all zero.

Therefore, **any three vectors in the plane are linearly dependent**.

#### Example
$$
\begin{pmatrix}
1 \\ 2
\end{pmatrix}
+
0\cdot
\begin{pmatrix}
2 \\ 2
\end{pmatrix}
-
0.5\cdot
\begin{pmatrix}
2 \\ 4
\end{pmatrix}
=
\begin{pmatrix}
0 \\ 0
\end{pmatrix}.
$$

**Geometric interpretation:**
In a plane, there are only **two independent directions of movement**.

### Linear Dependence in 3D

In $\mathbb{R}^3$, the geometric picture extends naturally:

- **Three vectors are linearly dependent if and only if they are coplanar**, meaning they lie in the same plane.
- **Four vectors in $\mathbb{R}^3$ are always linearly dependent**, since there are only three independent directions in space.


### General Geometric Principle

In general:

> **A set of $n$ vectors is linearly dependent if and only if they lie in an $(n-1)$-dimensional subspace.**

Equivalently:

> **The dimension of a vector space equals the maximum number of linearly independent directions of movement it contains.**

This geometric viewpoint is fundamental for understanding **dimension, basis, rank, and degrees of freedom** in linear algebra.

## Conclusion

Let us summarize the key ideas about **linear combinations** and **linear dependence**.

- A set of vectors is **linearly dependent** if at least one vector in the set can be expressed as a linear combination of the others, with **at least one scalar different from zero**.

- A **linear combination** has the form
  $$
  c_1 \vec v_1 + c_2 \vec v_2 + \cdots + c_n \vec v_n.
  $$
  - If all scalars satisfy $c_1 = c_2 = \cdots = c_n = 0$, the combination is called **trivial**.
  - If **at least one scalar is non-zero**, the combination is **non-trivial**.

- A set of vectors is **linearly independent** if the **only** linear combination that equals the zero vector is the trivial one.

- From a **geometric perspective**:
  - In $\mathbb{R}^2$, vectors are linearly dependent if and only if they are **collinear**.
  - In $\mathbb{R}^3$, vectors are linearly dependent if and only if they are **coplanar**.
  - In general, $n$ vectors are linearly dependent if they lie in an $(n-1)$-dimensional space.

- If the components of one vector are **proportional** to the components of another vector, then the vectors are **linearly dependent**.

These ideas provide the foundation for understanding **dimension**, **basis**, and **degrees of freedom** in linear algebra.
