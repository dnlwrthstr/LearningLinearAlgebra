# Solving linear systems using Gauss elimination

In this topic, we will learn how to solve systems of linear equations using the **Gauss method**.
We will also introduce **echelon reduction** and show how it can be used to systematically solve a system of linear equations.

Finally, we will walk through a **worked example** to illustrate the entire procedure step by step.

## Gaussian Elimination

Earlier, you studied a method that reduces a matrix to an echelon matrix using elementary row operations. The method you are going to see now is called **Gaussian elimination**. After constructing an **augmented matrix**, we reduce it to an echelon matrix and then solve the system.

Gaussian elimination consists of three main steps:

1. **Construct the augmented matrix** by writing down the coefficients of the equations in a grid and adding an extra column on the right for the constant terms. Each row represents one equation.
2. **Reduce the augmented matrix to an echelon form** using elementary row operations. Optionally, you may continue to the reduced row echelon form to simplify calculations.
3. **Back-solving**, where you reconstruct the system of equations from the reduced matrix and find the values of the unknowns.

### Example

Consider the system:
$$
\begin{cases}
3x + 2y - 5z = -1 \\
2x - y + 3z = 13 \\
x + 2y - z = 9
\end{cases}
$$


#### Step 1: Construct the augmented matrix

$$
\left(
\begin{array}{ccc|c}
3 & 2 & -5 & -1 \\
2 & -1 & 3 & 13 \\
1 & 2 & -1 & 9
\end{array}
\right)
$$


#### Step 2: Reduce to echelon form

**Swap the first and third rows**:
$$
R_1 \leftrightarrow R_3
$$

$$
\left(
\begin{array}{ccc|c}
1 & 2 & -1 & 9 \\
2 & -1 & 3 & 13 \\
3 & 2 & -5 & -1
\end{array}
\right)
$$

**Eliminate entries below the first pivot**:
$$
R_2 \leftarrow R_2 - 2R_1, \quad
R_3 \leftarrow R_3 - 3R_1
$$

$$
\left(
\begin{array}{ccc|c}
1 & 2 & -1 & 9 \\
0 & -5 & 5 & -5 \\
0 & -4 & -2 & -28
\end{array}
\right)
$$


**Normalize the second and third rows**:
$$
R_2 \leftarrow -\frac{1}{5} R_2, \quad
R_3 \leftarrow -\frac{1}{2} R_3
$$

$$
\left(
\begin{array}{ccc|c}
1 & 2 & -1 & 9 \\
0 & 1 & -1 & 1 \\
0 & 2 & 1 & 14
\end{array}
\right)
$$

**Eliminate the entry below the second pivot**:
$$
R_3 \leftarrow R_3 - 2R_2
$$

$$
\left(
\begin{array}{ccc|c}
1 & 2 & -1 & 9 \\
0 & 1 & -1 & 1 \\
0 & 0 & 3 & 12
\end{array}
\right)
$$

**Normalize the third row**:
$$
R_3 \leftarrow \frac{1}{3} R_3
$$

$$
\left(
\begin{array}{ccc|c}
1 & 2 & -1 & 9 \\
0 & 1 & -1 & 1 \\
0 & 0 & 1 & 4
\end{array}
\right)
$$

#### Step 3: Reduce further (optional, reduced echelon form)

Eliminate non-zero entries above pivots:
$$
\begin{aligned}
R_2 &\leftarrow R_2 + R_3 \\
R_1 &\leftarrow R_1 + R_3 \\
R_1 &\leftarrow R_1 - 2R_2
\end{aligned}
$$

$$
\left(
\begin{array}{ccc|c}
1 & 0 & 0 & 3 \\
0 & 1 & 0 & 5 \\
0 & 0 & 1 & 4
\end{array}
\right)
$$

#### Step 4: Back-solving

The corresponding system is:
$$
\begin{cases}
x = 3 \\
y = 5 \\
z = 4
\end{cases}
$$

#### Final Solution

$$
\boxed{
\begin{aligned}
x &= 3 \\
y &= 5 \\
z &= 4
\end{aligned}
}
$$

## Linear System of Equations: Possible Outcomes After Reduction

After reducing an augmented matrix to **row echelon form** (or reduced row echelon form), several different situations may occur. Each of them corresponds to a different type of solution set for the system of linear equations.

### 1. **Unique Solution**

If, after reduction, the coefficient part of the augmented matrix is **triangular** (that is, each row introduces a new pivot), then the system has a **single unique solution**.

For example:
$$
\left(
\begin{array}{ccc|c}
1 & 2 & -1 & 9 \\
0 & 1 & -1 & 1 \\
0 & 0 & 1 & 4
\end{array}
\right)
$$

Each variable corresponds to a pivot position, so the solution is uniquely determined.

### 2. **Infinitely Many Solutions**

If the last row has more than one non-zero element, then the system has an infinite number of solutions. Like in the matrix below:

$$
\begin{pmatrix}
3 & 2 & 6 & 4 \mid 4 \\
0 & 1 & 7 & 1 \mid 2 \\
0 & 0 & 4 & 2 \mid 9
\end{pmatrix}
$$

### 3. **Redundant Equations (Still Infinitely Many Solutions)**

If one or more rows consist entirely of zeros **both on the left-hand side and on the right-hand side**, then these rows represent **redundant equations**. They do not impose any new constraints.

Example:
$$
\left(
\begin{array}{ccc|c}
3 & 2 & 6 & 4 \\
0 & 1 & 7 & 2 \\
0 & 0 & 4 & 9 \\
0 & 0 & 0 & 0
\end{array}
\right)
$$

Such zero rows can be ignored, and the number of solutions is determined by the remaining rows. This typically leads to **infinitely many solutions** if there are free variables.

### 4. **No Solution (Inconsistent System)**

If a row appears where **all coefficients on the left-hand side are zero**, but the **right-hand side is non-zero**, then the system is **inconsistent** and has **no solutions**.

Example:
$$
\left(
\begin{array}{ccc|c}
3 & 2 & 6 & 4 \\
0 & 1 & 7 & 2 \\
0 & 0 & 4 & 9 \\
0 & 0 & 0 & 8
\end{array}
\right)
$$

The last row corresponds to the impossible equation:
$$
0 = 8
$$
which cannot be satisfied.

### 5. **Homogeneous Systems**

A system of linear equations is called **homogeneous** if **all constant terms are zero**, i.e.
$$
A\mathbf{x} = \mathbf{0}
$$

Homogeneous systems **always have at least one solution**, called the **trivial solution**:
$$
x_1 = x_2 = \cdots = x_n = 0
$$

Depending on the number of pivots, homogeneous systems may have:
- only the trivial solution, or
- infinitely many non-trivial solutions.

### Summary

- **Triangular matrix with pivots in every column** → unique solution
- **Fewer pivots than variables** → infinitely many solutions
- **Row of zeros equals zero** → redundant equation
- **Row of zeros equals non-zero** → no solution
- **Homogeneous system** → always at least the trivial solution
