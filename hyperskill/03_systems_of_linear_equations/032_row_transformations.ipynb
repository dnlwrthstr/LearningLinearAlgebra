# Row transformations

## Elementary transformations on rows of matrices

Now let's introduce various elementary transformations on rows of matrices, which we have already seen used in the topic about matrix representation of linear systems. Each system equation corresponds to a row of a matrix. Therefore, transformations of rows correspond to transformations of equations.

To start with, we are going to define which transformations we consider elementary. For illustration, we will use the following matrix:

$$
A =
\begin{pmatrix}
2 & 6 & 3 \\
4 & 2 & 7 \\
8 & 3 & 1
\end{pmatrix}
$$

### Elementary operations

#### 1. Interchanging any two rows of the matrix

Interchanging the second and the third rows of matrix $A$ results in:

$$
\begin{pmatrix}
2 & 6 & 3 \\
8 & 3 & 1 \\
4 & 2 & 7
\end{pmatrix}
$$

#### 2. Multiplying or dividing each element of a row by a fixed non-zero number

After multiplying the first row of matrix $A$ by $2$, the resulting matrix looks like this:

$$
\begin{pmatrix}
4 & 12 & 6 \\
4 & 2 & 7 \\
8 & 3 & 1
\end{pmatrix}
$$

#### 3. Adding or subtracting one row from another

Subtracting the second row from the first transforms matrix $A$ into:

$$
\begin{pmatrix}
-2 & 4 & -4 \\
4 & 2 & 7 \\
8 & 3 & 1
\end{pmatrix}
$$

## Equivalent matrices

We can say that matrix $B$ is equivalent to matrix $A$ (and vice versa) if and only if one matrix can be obtained from the other by applying elementary row operations. Such matrices are called **equivalent**.

Equivalent matrices are denoted as:
$$
A \sim B
$$

Note that the order of matrices does not matter for equivalency:
$A \sim B$ is the same as $B \sim A$, since we can always transform $B$ back into $A$ using inverse elementary operations.

The equivalence property is very important. Using elementary operations, we can transform matrices into **upper triangular form**. In this form, systems of linear equations become easy to solve.

Now we will see how a chain of transformations consisting of several elementary operations preserves equivalence.

### Example

Consider the following matrix:

$$
\begin{pmatrix}
2 & 7 & 1 \\
3 & 9 & -3 \\
2 & 6 & -1
\end{pmatrix}
$$

#### Step 1: Multiply the second row by $\frac{1}{3}$

$$
\begin{pmatrix}
2 & 7 & 1 \\
1 & 3 & -1 \\
2 & 6 & -1
\end{pmatrix}
$$

#### Step 2: Subtract the third row from the first row

$$
\begin{pmatrix}
0 & 1 & 2 \\
1 & 3 & -1 \\
2 & 6 & -1
\end{pmatrix}
$$

#### Step 3: Exchange the first and second rows, then subtract twice the first row from the third row

$$
\begin{pmatrix}
1 & 3 & -1 \\
0 & 1 & 2 \\
0 & 0 & 1
\end{pmatrix}
$$

### Result

Therefore, the following matrices are equivalent:

$$
\begin{pmatrix}
2 & 7 & 1 \\
3 & 9 & -3 \\
2 & 6 & -1
\end{pmatrix}
\;\sim\;
\begin{pmatrix}
1 & 3 & -1 \\
0 & 1 & 2 \\
0 & 0 & 1
\end{pmatrix}
$$

This is a very typical example because we reduced the matrix to an equivalent **upper triangular matrix**. In the future, this will help us solve linear systems efficiently.

### How to check matrix equivalence

To determine whether two matrices are equivalent, we can try to transform both of them into the same matrix (for example, the identity matrix). If this is possible, then the matrices are equivalent.

### Important note on column operations

It is possible to carry out similar operations with **columns**, but we will not consider them here. In terms of linear equations, column operations correspond to changes of variables. Because of this, we **cannot** say that two systems of linear equations have the same solutions if their matrices are related by column operations.

## Non-equivalent matrices

Not all matrices are equivalent. For example, consider the following two matrices:

$$
\begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}
\qquad
\begin{pmatrix}
1 & 0 & 1 \\
0 & 1 & 1 \\
0 & 0 & 0
\end{pmatrix}
$$

The first matrix is the **identity matrix**, so we leave it as it is.

Now consider the second matrix. We can apply an elementary row operation: subtract both the first and the second rows from the third row. After doing this, we obtain:

$$
\begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 0
\end{pmatrix}
$$

You may notice that no sequence of elementary row operations can create a nonzero entry in the third column of this matrix. Therefore, it is impossible to transform the second matrix into the identity matrix.

Hence, these two matrices are **not equivalent**.

## Conclusion

In this topic, we discussed various elementary transformations on matrices as well as the crucial concept of equivalence. Let’s briefly recap the main points:

- There are **three elementary transformations on matrices**:
  1. Interchanging two rows
  2. Multiplying a row by a nonzero scalar (or its inverse operation — division)
  3. Adding one row to another (or its inverse operation — subtraction)

- A matrix $A$ is **equivalent** to a matrix $B$ if and only if $A$ can be transformed into $B$ by applying a finite sequence of elementary row transformations.

- If **no combination of elementary transformations** can transform a matrix $A$ into a matrix $B$, then the matrices are **non-equivalent**.
