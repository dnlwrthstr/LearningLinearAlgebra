# Matrix representation of systems of linear equations

This topic covers one of the most important aspects of linear algebra. Here, you will learn how to represent systems of linear equations in **matrix form** and how to convert this matrix form back to the original system of equations.

This correspondence allows you to solve systems efficiently and demonstrates one of the fundamental applications of matrices. Moreover, the existence of this relationship between systems of equations and matrices forms the foundation of **linear algebra itself**.

It acts as a unifying thread that seamlessly connects the **geometric interpretation** of linear objects with their **algebraic representation**.

## Systems of Linear Equations

Back in school, everyone studied systems of linear equations. Let’s take a look at an example for a quick revision:

$$
\begin{cases}
x + y = 1 \\
2x - y = 3
\end{cases}
$$

On the left-hand side, you have the unknowns ($x$ and $y$), which are the variables you usually aim to determine, along with their respective coefficients. The numbers on the right-hand side of the equations are called the **constant terms**.

There are also other ways of writing the same equations, such as:

$$
\begin{cases}
x + y - 1 = 0 \\
2x - y - 3 = 0
\end{cases}
$$

From this point on, for the sake of convenience, we will consider systems where constant terms are on the **right-hand side**, while unknowns and their coefficients are on the **left-hand side**.

> **Note:** If an equation does not have a constant term, it is treated as having a constant term equal to $0$.


### General Form of a System of Linear Equations

Assume that you have $m$ linear equations and $n$ unknowns. Since the number of Latin letters is limited, unknowns are usually denoted using indices such as $x_1, x_2, \dots, x_n$.

The general form of a system of linear equations is:

$$
\begin{cases}
a_{11}x_1 + \cdots + a_{1n}x_n = b_1 \\
\vdots \\
a_{m1}x_1 + \cdots + a_{mn}x_n = b_m
\end{cases}
$$

Here:
- $x_1, \dots, x_n$ are the **unknown variables**,
- $a_{ij}$ are the **coefficients** of the unknowns,
- $b_1, \dots, b_m$ are the **constant terms**.

### Missing Variables in Equations

If one of the equations in the system does not include some of the unknowns, the coefficient of that unknown is assumed to be zero.

For example, the system

$$
\begin{cases}
3x_1 + 2x_4 - x_5 = 1 \\
2x_2 - 3x_3 + x_5 = 0
\end{cases}
$$

is equivalent to

$$
\begin{cases}
3x_1 + 0x_2 + 0x_3 + 2x_4 - x_5 = 1 \\
0x_1 + 2x_2 - 3x_3 + 0x_4 + x_5 = 0
\end{cases}
$$

## Transforming a System of Linear Equations to Matrix Form

Now it’s time to review the representation of a system of linear equations in matrix form. For instance, consider the following system:

$$
\begin{cases}
x + 2y + 3z = 4 \\
2x + 3y - z = 2 \\
7x - 3y + 2z = 1
\end{cases}
$$

Its **matrix form** (the extended or augmented matrix) is:

$$
\left(
\begin{array}{ccc|c}
1 & 2 & 3 & 4 \\
2 & 3 & -1 & 2 \\
7 & -3 & 2 & 1
\end{array}
\right)
$$

Such a matrix is called **extended** (or **augmented**) because the constant terms are included as a separate column.
The left part of the matrix is called the **coefficient matrix**, and it contains the coefficients of the unknowns.

Although it is sometimes easier to work with the extended form, you will see later that **coefficient matrices** are exactly what is usually referred to as matrices in linear algebra.

### Coefficient Matrix

The coefficient matrix of the system above is:

$$
\left(
\begin{array}{ccc}
1 & 2 & 3 \\
2 & 3 & -1 \\
7 & -3 & 2
\end{array}
\right)
$$

It is important to maintain:
- the **order of rows** (the $k$-th equation corresponds to the $k$-th row),
- the **order of columns** (coefficients of $x_i$ must appear in the $i$-th column).

You *may* change the order of rows if you keep the constant terms aligned, but you **must not change the order of columns**.
In practice, it is safer not to reorder anything while forming the matrix.

### Column Vectors of Unknowns and Constants

You can also write the unknowns and constant terms as column vectors:

$$
\begin{pmatrix}
x \\ y \\ z
\end{pmatrix}
\quad \text{and} \quad
\begin{pmatrix}
4 \\ 2 \\ 1
\end{pmatrix}
$$

### General Form

For a general system of linear equations:

$$
\begin{cases}
a_{11}x_1 + \cdots + a_{1n}x_n = b_1 \\
\vdots \\
a_{m1}x_1 + \cdots + a_{mn}x_n = b_m
\end{cases}
$$

the **extended matrix** is:

$$
\left(
\begin{array}{ccc|c}
a_{11} & \cdots & a_{1n} & b_1 \\
\vdots & \ddots & \vdots & \vdots \\
a_{m1} & \cdots & a_{mn} & b_m
\end{array}
\right)
$$

the **coefficient matrix** is:

$$
\left(
\begin{array}{ccc}
a_{11} & \cdots & a_{1n} \\
\vdots & \ddots & \vdots \\
a_{m1} & \cdots & a_{mn}
\end{array}
\right)
$$

and the column vectors of unknowns and constants are:

$$
\begin{pmatrix}
x_1 \\ \vdots \\ x_n
\end{pmatrix}
\quad \text{and} \quad
\begin{pmatrix}
b_1 \\ \vdots \\ b_m
\end{pmatrix}
$$

### Unequal Numbers of Equations and Unknowns

If the number of equations and unknowns is not the same, the vectors of unknowns and constants will have different lengths.

For example, the system

$$
\begin{cases}
x_1 - x_2 - x_3 = -1 \\
2x_2 - 5x_3 = -3
\end{cases}
$$

has a three-entry vector of unknowns and a two-entry vector of constants:

$$
\begin{pmatrix}
x_1 \\ x_2 \\ x_3
\end{pmatrix}
\quad \text{and} \quad
\begin{pmatrix}
-1 \\ -3
\end{pmatrix}
$$

### Matrix Equation Form

If you multiply the coefficient matrix by the vector of unknowns, you obtain the vector of constants:

$$
\left(
\begin{array}{ccc}
a_{11} & \cdots & a_{1n} \\
\vdots & \ddots & \vdots \\
a_{m1} & \cdots & a_{mn}
\end{array}
\right)
\begin{pmatrix}
x_1 \\ \vdots \\ x_n
\end{pmatrix}
=
\begin{pmatrix}
b_1 \\ \vdots \\ b_m
\end{pmatrix}
$$

Denoting the coefficient matrix by $A$, the vector of unknowns by $\vec{x}$, and the vector of constants by $\vec{b}$, the system can be written compactly as:

$$
A \cdot \vec{x} = \vec{b}
$$

If some coefficient $a_{ij}$ is zero, it is usually omitted in the system of equations rather than written explicitly as $0 \cdot x_j$.
This simply corresponds to a zero entry in the matrix.

### Reversibility

This process can also be reversed.
Given a matrix $A$ and a vector $\vec{b}$ (or an extended matrix), you can always reconstruct the corresponding system of linear equations by reading each row as an equation.

## Examples

Let’s look at some examples.

Given a system

$$
\begin{cases}
y = 2 \\
x + y = 3 \\
x + z = 4
\end{cases}
$$

keeping in mind that the absent unknowns in equations are considered as having zero coefficients, you will end up with a corresponding expanded matrix

$$
\left(
\begin{array}{ccc|c}
0 & 1 & 0 & 2 \\
1 & 1 & 0 & 3 \\
1 & 0 & 1 & 4
\end{array}
\right)
$$

You can also write this using a regular matrix and column vectors:

$$
\left(
\begin{array}{ccc}
0 & 1 & 0 \\
1 & 1 & 0 \\
1 & 0 & 1
\end{array}
\right)
\begin{pmatrix}
x \\ y \\ z
\end{pmatrix}
=
\begin{pmatrix}
2 \\ 3 \\ 4
\end{pmatrix}
$$

Now you can do the same in reverse. Consider an extended matrix

$$
\left(
\begin{array}{cccc|c}
-2 & 0 & 1 & 1 & -2 \\
0 & 3 & -1 & 1 & 6 \\
1 & 1 & 2 & -\frac{3}{2} & 0
\end{array}
\right)
$$

you can conclude that the coefficient matrix and constant term column vector are

$$
\left(
\begin{array}{cccc}
-2 & 0 & 1 & 1 \\
0 & 3 & -1 & 1 \\
1 & 1 & 2 & -\frac{3}{2}
\end{array}
\right)
\quad \text{and} \quad
\begin{pmatrix}
-2 \\ 6 \\ 0
\end{pmatrix}
$$

As the rows of the coefficient matrix have four entries it means that this matrix represents a system with four unknowns. Let’s denote them by $y_1$, $y_2$, $y_3$ and $y_4$. The final system is

$$
\begin{cases}
-2y_1 + y_3 + y_4 = -2 \\
3y_2 - y_3 + y_4 = 6 \\
y_1 + y_2 + 2y_3 - \frac{3}{2}y_4 = 0
\end{cases}
$$

In this topic, you've learned how to represent systems of linear equations using different types of matrices and how to restore a linear system from its matrix representation. In the future, using this knowledge, you will be able to learn how to solve large systems using various methods, determine whether they have solutions, and explore many related applications.
