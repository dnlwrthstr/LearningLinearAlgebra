# Introduction to linear operators

Vector spaces by themselves have many useful properties. However, it is only when we connect them through special functions that we discover their deepest qualities and obtain the strongest results.

In this topic, we will begin to study these functions, their most important properties, and some examples. Later, we will investigate more elaborate results and their relationship with matrices. As in previous topics, we will denote by $V$ and $W$ two vector spaces.


## Preserving the structure

A function $f : A \to B$ connects sets, so it acts as a bridge between them. But vector spaces are not just simple sets; they have an algebraic structure.


This suggests that we should focus primarily on functions that preserve such a structure, that is, functions that are compatible with vector addition and scalar multiplication. These functions are called **linear transformations** (usually denoted by a more dramatic $T$) and they form the core of linear algebra.


From now on in this course, practically all concepts and results will involve linear transformations in one way or another.

Central notions such as matrices, eigenvectors, and orthogonal projections are closely related to these functions. We can also use them to calculate the dimensions of spaces, interpret many geometric transformations, change coordinates, exploit bases of spaces, simplify many calculations, approximate more complicated functions, fully understand systems of linear equations, and much more.

We are now ready to introduce the definition of a linear transformation and explore its immediate properties.


## Definition and basic properties

A function $T : V \to W$ is a **linear transformation** if, for all $v, u \in V$ and for every $\lambda \in \mathbb{R}$,

$T(v + u) = T(v) + T(u)$

$T(\lambda v) = \lambda T(v)$

The first property means that once you have computed the values of $T$ at $v$ and $u$, the value of $T$ at their sum is simply the sum of those values. In the same way, the second property tells us that when transforming $v$ under $T$, we automatically know the transformation of any scalar multiple of $v$.

These two properties come together naturally when analyzing how linear transformations interact with linear combinations:

$T(\lambda_1 v_1 + \lambda_2 v_2 + \cdots + \lambda_n v_n)
= T(\lambda_1 v_1) + T(\lambda_2 v_2) + \cdots + T(\lambda_n v_n)
= \lambda_1 T(v_1) + \lambda_2 T(v_2) + \cdots + \lambda_n T(v_n)$.

It is now time to note a fundamental property of linear transformations: they map the $0$ of $V$ to the $0$ of $W$, that is,

$T(0) = 0$.

You can verify this quickly, since

$T(0) = T(0 + 0) = T(0) + T(0)$,

and subtracting $T(0)$ from both sides yields $T(0) = 0$.

This gives us a simple test to determine whether a transformation is **not** linear:

If $T(0) \neq 0$, then $T$ is not linear.

The most important linear transformations are those that map a space to itself. Their properties are so special that they deserve their own name: if $T : V \to V$ is linear, then we call it a **linear operator**.

## Examples of linear transformations

Undoubtedly, the simplest linear transformation is the zero constant function $O$, which is given by
$O(v) = 0$.
It is immediate that it is linear since

$O(v + u) = 0 = 0 + 0 = O(v) + O(u)$

and

$O(\lambda v) = 0 = \lambda 0 = \lambda O(v)$.

Another well-known function is the identity function $I : V \to V$, which is given by
$I(v) = v$.
It is easy to see that it is also linear, so you can quickly check it. It is a linear operator.

Let us consider a slightly more difficult function
$T : \mathbb{R}^2 \to \mathbb{R}^2$
given by

$T(x, y) = (x + y, -x - y)$.

Let us prove that this function is linear.

### Proof that $T(x, y) = (x + y, -x - y)$ is linear

Let $v = (x_1, y_1)$ and $u = (x_2, y_2)$. Then

$T(v + u)
= T(x_1 + x_2, y_1 + y_2)
= ((x_1 + x_2) + (y_1 + y_2), -(x_1 + x_2) - (y_1 + y_2))$

$= (x_1 + y_1, -x_1 - y_1) + (x_2 + y_2, -x_2 - y_2)
= T(v) + T(u)$.

Also,

$T(\lambda v)
= T(\lambda x_1, \lambda y_1)
= (\lambda(x_1 + y_1), \lambda(-x_1 - y_1))
= \lambda(x_1 + y_1, -x_1 - y_1)
= \lambda T(v)$.

Therefore, $T$ is linear.

In future topics, you will learn to analyze linear transformations geometrically, but for now, note that this operator converts the entire $\mathbb{R}^2$ plane into the line generated by the vector $(1, -1)$.


Now let us look at a very important example. Take an $m \times n$ matrix $A$. We can define a function on $\mathbb{R}^n$ given by the product $Av$, which is a vector in $\mathbb{R}^m$. This function is the transformation associated with $A$ and is denoted by

$L_A : \mathbb{R}^n \to \mathbb{R}^m, \quad L_A(v) = Av$.

From the properties of matrix multiplication, it is clear that this is a linear transformation:

$L_A(v + u) = A(v + u) = Av + Au = L_A(v) + L_A(u)$

and

$L_A(\lambda v) = A(\lambda v) = \lambda(Av) = \lambda L_A(v)$.

Now let us look at a slightly more extravagant transformation. If $M_n$ denotes the set of square $n \times n$ matrices, we can define the trace function

$\operatorname{tr} : M_n \to \mathbb{R}$

given by

$\operatorname{tr}(A) = \sum_{i=1}^n a_{ii}$.

We can easily check that it is linear.

### Proof that $\operatorname{tr}$ is linear

$\operatorname{tr}(A + B)
= \sum_{i=1}^n (a_{ii} + b_{ii})
= \sum_{i=1}^n a_{ii} + \sum_{i=1}^n b_{ii}
= \operatorname{tr}(A) + \operatorname{tr}(B)$

and

$\operatorname{tr}(\lambda A)
= \sum_{i=1}^n (\lambda a_{ii})
= \lambda \sum_{i=1}^n a_{ii}
= \lambda \operatorname{tr}(A)$.

The transpose defines a transformation from the set of $m \times n$ matrices to the set of $n \times m$ matrices. It is easy to see that it is linear thanks to its properties, since

$(A + B)^t = A^t + B^t$

and

$(\lambda A)^t = \lambda(A^t)$.

But let us not rush: not all functions are linear. In fact, only a few are. Consider the function $f(x) = x + 1$. It is clear that

$f(x + y) = x + y + 1 \neq x + y + 2 = f(x) + f(y)$,

so it cannot be linear. Another example is the cosine function. Since $\cos(0) = 1$, it is impossible for it to be linear, due to the result from the previous section.


## Relationship with the basis

A vector space is completely determined by a basis
$\{v_1, v_2, \ldots, v_n\}$.
This means that any vector $u$ can be uniquely written as a linear combination of the basis vectors:

$u = \lambda_1 v_1 + \lambda_2 v_2 + \cdots + \lambda_n v_n$.

But linear transformations interact perfectly with linear combinations. Let us see how a linear transformation $T$ acts on a vector written in terms of a basis:

$T(u)
= T(\lambda_1 v_1 + \lambda_2 v_2 + \cdots + \lambda_n v_n)
= \lambda_1 T(v_1) + \lambda_2 T(v_2) + \cdots + \lambda_n T(v_n)$.

This is the key observation: to know the value of $T$ at $u$, it is enough to know the values of $T$ at the basis vectors.

If we take another vector
$z = \mu_1 v_1 + \mu_2 v_2 + \cdots + \mu_n v_n$,
then

$T(z)
= \mu_1 T(v_1) + \mu_2 T(v_2) + \cdots + \mu_n T(v_n)$,

where only the coefficients have changed.

In summary, once we know the values of $T$ on a basis, we effectively know all of its values. This also means that to construct a linear transformation, it is enough to choose its values on a basis.

A linear transformation $T$ is completely determined by the values it takes on a basis. Just as a basis describes the entire space, it also describes any linear transformation defined on that space.

This is extremely useful: if a linear transformation is complicated to evaluate directly, we only need to compute its values on the basis and then derive a general formula for every vector.

### Example

Let us reconstruct a linear transformation
$T : \mathbb{R}^2 \to \mathbb{R}^2$.
Suppose we only know that

$T(e_1) = (0, 2)$
and
$T(e_2) = (3, 1)$.

Then, for a general vector $(x, y)$,

$T(x, y)
= T(x e_1 + y e_2)
= x T(e_1) + y T(e_2)
= x(0, 2) + y(3, 1)
= (3y, 2x + y)$.


```python
import numpy as np
import matplotlib.pyplot as plt

# ---------- Figure layout ----------
fig = plt.figure(figsize=(14, 5), facecolor="black")
gs = fig.add_gridspec(1, 3, width_ratios=[1, 0.4, 1])

# ---------- LEFT: domain ----------
ax1 = fig.add_subplot(gs[0, 0])
ax1.set_aspect('equal')
ax1.set_xlim(-3, 3)
ax1.set_ylim(-3, 3)
ax1.axis('off')
ax1.set_facecolor("black")

ax1.axhline(0, color='white', linewidth=1)
ax1.axvline(0, color='white', linewidth=1)

ax1.arrow(0, 0, 1.8, 0, head_width=0.12, head_length=0.18,
          length_includes_head=True, color='dodgerblue')
ax1.arrow(0, 0, 0, 1.8, head_width=0.12, head_length=0.18,
          length_includes_head=True, color='dodgerblue')

ax1.text(1.9, -0.2, r"$e_1$", color='dodgerblue', fontsize=12)
ax1.text(-0.25, 1.9, r"$e_2$", color='dodgerblue', fontsize=12)

u = np.array([1.4, 1.2])
ax1.arrow(0, 0, u[0], u[1], head_width=0.12, head_length=0.18,
          length_includes_head=True, color='green')
ax1.text(u[0]*1.05, u[1]*1.05, r"$u$", color='green', fontsize=12)

# ---------- MIDDLE: T arrow ----------
axm = fig.add_subplot(gs[0, 1])
axm.axis('off')
axm.set_facecolor("black")

axm.arrow(0.05, 0.5, 0.9, 0,
          transform=axm.transAxes,
          head_width=0.08, head_length=0.08,
          length_includes_head=True, color='dodgerblue')
axm.text(0.48, 0.62, r"$T$", transform=axm.transAxes,
         fontsize=16, color='dodgerblue')

# ---------- RIGHT: codomain ----------
ax2 = fig.add_subplot(gs[0, 2])
ax2.set_aspect('equal')
ax2.set_xlim(-3, 3)
ax2.set_ylim(-3, 3)
ax2.axis('off')
ax2.set_facecolor("black")

ax2.axhline(0, color='white', linewidth=1)
ax2.axvline(0, color='white', linewidth=1)

T = np.array([[0, 3],
              [2, 1]])

grid_vals = np.arange(-3, 4)
for x in grid_vals:
    p1 = T @ np.array([x, -3])
    p2 = T @ np.array([x,  3])
    ax2.plot([p1[0], p2[0]], [p1[1], p2[1]],
             color='cornflowerblue', linewidth=0.8, alpha=0.6)

for y in grid_vals:
    p1 = T @ np.array([-3, y])
    p2 = T @ np.array([ 3, y])
    ax2.plot([p1[0], p2[0]], [p1[1], p2[1]],
             color='cornflowerblue', linewidth=0.8, alpha=0.6)

Te1 = T @ np.array([1, 0])
Te2 = T @ np.array([0, 1])
Tu  = T @ u

ax2.arrow(0, 0, Te1[0], Te1[1],
          head_width=0.12, head_length=0.18,
          length_includes_head=True, color='dodgerblue')
ax2.arrow(0, 0, Te2[0], Te2[1],
          head_width=0.12, head_length=0.18,
          length_includes_head=True, color='dodgerblue')

ax2.text(Te1[0]*1.05, Te1[1]*1.05, r"$T(e_1)$",
         color='dodgerblue', fontsize=12)
ax2.text(Te2[0]*1.05, Te2[1]*1.05, r"$T(e_2)$",
         color='dodgerblue', fontsize=12)

# ---------- ADDED: arrow for T(u) ----------
ax2.arrow(0, 0, Tu[0], Tu[1],
          head_width=0.12, head_length=0.18,
          length_includes_head=True, color='green')
ax2.text(Tu[0]*1.02, Tu[1]*1.02, r"$T(u)$",
         color='green', fontsize=12)

plt.tight_layout()
plt.show()
```

Thanks to the flexibility of linear transformations, we can sometimes easily construct their inverses. In the previous example, it is enough to denote
$w_1 = (0, 2)$ and $w_2 = (3, 1)$,
and construct the transformation
$S : \mathbb{R}^2 \to \mathbb{R}^2$
given by

$S(w_1) = e_1$
and
$S(w_2) = e_2$.

Thus, $S$ “returns” each vector to its place of origin and transforms the curvilinear space back into the original rectangular one.


## Conclusion

Let $\{v_1, v_2, \ldots, v_n\}$ be a basis for $V$.

Linear transformations are functions that preserve the structure of vector spaces.
A function $T : V \to W$ is a **linear transformation** if, for all $v, u \in V$ and for every $\lambda \in \mathbb{R}$,

$T(v + u) = T(v) + T(u)$

$T(\lambda v) = \lambda T(v)$.

When $T : V \to V$ is linear, it is called a **linear operator**.

If $T$ is linear, then $T(0) = 0$.

Every $m \times n$ matrix $A$ generates a linear transformation defined by

$L_A : \mathbb{R}^n \to \mathbb{R}^m, \quad L_A(v) = Av$.

A linear transformation is completely determined by its values on a basis. For every vector $u \in V$,

$T(u) = \lambda_1 T(v_1) + \lambda_2 T(v_2) + \cdots + \lambda_n T(v_n)$.
