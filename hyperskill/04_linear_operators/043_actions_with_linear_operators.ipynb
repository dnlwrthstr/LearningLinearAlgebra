# Actions with linear operators

Just as vectors are essential in understanding the movement and relationships within a space, **linear transformations** provide a framework to explore how one vector space can be transformed into another.

You're already familiar with the basics of linear transformations, but now you'll learn to manipulate them just as fluently as vectors. You will also start using an effective operation that applies specifically to functions and behaves analogously to **matrix multiplication**.

## The linear transformations are vectors, too!

Vectors may be abstract objects, but so far you have been able to work with them fluently and obtain many useful results thanks to their structure. Operations such as **vector addition** and **scalar multiplication** provide all the flexibility you need. Moreover, we often do not focus on individual vectors but rather on how these operations behave. Whether you are adding vectors in the plane or adding matrices, the underlying algebraic behavior is the same, and the results follow similar rules.

It turns out that **linear transformations** share this same structure. In particular, we can define both a **sum** of linear transformations and a **product by a scalar**. Since the sum of two vectors and the scalar multiple of a vector must again be a vector, the result of combining linear transformations must itself be a linear transformation.

### Linear transformations as a vector space

Let $V$ and $W$ be vector spaces. Denote by
$$
L(V, W)
$$
the set of all linear transformations from $V$ to $W$.

If
$$
S, T \in L(V, W),
$$
we define their **sum** $S + T$ as the transformation that assigns to each vector $v \in V$ the vector
$$
(S + T)(v) = S(v) + T(v) \in W.
$$

Similarly, for any scalar $\lambda$, we define the **scalar multiple** $\lambda T$ by
$$
(\lambda T)(v) = \lambda\, T(v), \quad v \in V.
$$

In summary, the operations in $L(V, W)$ are:

- **Addition**
  $$
  T + S : V \to W, \qquad (T + S)(v) = T(v) + S(v)
  $$

- **Scalar multiplication**
  $$
  \lambda T : V \to W, \qquad (\lambda T)(v) = \lambda T(v)
  $$


### Why this forms a vector space

The crucial question is whether these operations make $L(V, W)$ a vector space. The answer is **yes**. All vector space axioms are satisfied, and the only subtle point is the existence of a zero element.

This role is played by the **zero transformation** $O$, defined by
$$
O(v) = 0 \quad \text{for all } v \in V.
$$

Thus, $L(V, W)$ is a vector space whose elements are linear transformations.

In particular, when $V = W$, the set of linear transformations from $V$ to itself is denoted by
$$
L(V) := L(V, V),
$$
and its elements are called **linear operators on $V$**.

## Composing transformations

Functions can be thought of as *processes* applied to objects in a set. Because of this, they can be **chained together** to form more complex processes that connect different sets. Since linear transformations are functions, they can also be composed.

Let $U$, $V$, and $W$ be vector spaces. If
$$
S \in L(U, V)
\quad \text{and} \quad
T \in L(V, W),
$$
then their **composition** is defined as
$$
T \circ S : U \to W,
\qquad
(T \circ S)(u) = T(S(u)).
$$

That is, first the transformation $S$ maps a vector $u \in U$ to $S(u) \in V$, and then the transformation $T$ maps this result to
$$
T(S(u)) \in W.
$$

### Linearity of the composition

The crucial question is whether the composition of two linear transformations is itself linear. Fortunately, the answer is **yes**:

$$
T \circ S \in L(U, W).
$$

This fact has far-reaching consequences and will play a central role later in linear algebra.

### Properties of composition

Composition is a new operation that is **independent** of vector addition and scalar multiplication. Conceptually, it behaves very much like **matrix multiplication**. In fact, this resemblance is not accidental and will become clearer later.

The composition of linear transformations satisfies the following properties:

- **Associativity**
  $$
  (T \circ S) \circ R = T \circ (S \circ R)
  $$

- **Identity**
  $$
  I \circ T = T = T \circ I
  $$

- **Distributivity over addition**
  $$
  (T_1 + T_2) \circ S = (T_1 \circ S) + (T_2 \circ S)
  $$
  $$
  T \circ (S_1 + S_2) = (T \circ S_1) + (T \circ S_2)
  $$

- **Compatibility with scalar multiplication**
  $$
  T \circ (\lambda S) = (\lambda T) \circ S = \lambda (T \circ S)
  $$

### Not commutative

Despite its similarity to multiplication, composition is **not commutative**. In general,
$$
S \circ T = T \circ S
$$
does not even make sense unless the domains and codomains match appropriately.

Even when
$$
T, S \in L(V),
$$
it is rare for
$$
T \circ S = S \circ T
$$
to hold. This is yet another strong parallel with matrix multiplication.

**Key takeaway:**
Composition equips linear transformations with a powerful algebraic structure, closely mirroring matrix multiplication, while retaining essential differences such as non-commutativity.

## Taking powers of linear transformations

Let us go further with the notion of *product*, that is, **composition**.
As with the multiplication of numbers, we can define **powers** of a linear operator
$$
T \in L(V)
$$
by repeated composition.

### Definition of powers

The **square** of an operator $T$ is defined as its composition with itself:
$$
T^2 = T \circ T.
$$

Higher powers are defined inductively:
$$
T^n = T \circ T^{\,n-1}.
$$

In complete analogy with numbers, we also define:
$$
T^0 = I \quad \text{(the identity operator)},
$$
$$
T^1 = T.
$$

Interpreting composition as the *successive application* of the same transformation, these definitions are natural.

### Algebraic properties of powers

The powers of a linear operator satisfy familiar exponent rules:
$$
T^m T^n = T^{m+n},
$$
$$
(T^m)^n = T^{mn}.
$$

These identities follow directly from the associativity of composition.

### A key difference from numbers

Although the notation resembles ordinary powers, linear transformations are **not numbers**, and composition is **not multiplication**. As a result, operators can exhibit behaviors that have no direct numerical analogue.

One important example is **idempotency**.

### Idempotent operators

An operator $T$ is called **idempotent** if
$$
T^2 = T.
$$

This means that after applying the operator once, further applications have **no additional effect**.

A classic example is a **projection**.

### Example: projection onto a line

Consider a vector $u \in \mathbb{R}^2$ and let
$$
P \in L(\mathbb{R}^2)
$$
be the projection onto a fixed line.

After projecting $u$ once, applying the projection again does not change the result:
$$
P(P(u)) = P(u).
$$

Thus,
$$
P^2 = P,
$$
and the projection operator is idempotent.

### Key takeaway

Powers of linear transformations behave much like powers of numbers, but they also reveal fundamentally new phenomena—such as idempotency—that are unique to operators and have deep geometric meaning.

## The geometry of transformations

Since linear transformations themselves form a vector space, we can study them **geometrically**, just like ordinary vectors. Let us explore this idea with a concrete and intuitive example in the plane.

### Projections in the plane

Consider a vector
$$
u = (x, y) \in \mathbb{R}^2.
$$

We define two linear transformations
$$
P, Q \in L(\mathbb{R}^2)
$$
as follows:
$$
P(x, y) = (x, 0),
$$
$$
Q(x, y) = (0, y).
$$

### Geometric interpretation

- **Transformation $P$**
  Projects any vector onto the **$x$-axis** by removing its $y$-component.

- **Transformation $Q$**
  Projects any vector onto the **$y$-axis** by removing its $x$-component.

Thus, every vector $u = (x,y)$ can be decomposed as
$$
u = P(u) + Q(u).
$$

Geometrically, this means that a vector in the plane is the sum of its horizontal and vertical components.

### Matrix representation

With respect to the standard basis of $\mathbb{R}^2$, these transformations correspond to the matrices
$$
P =
\begin{pmatrix}
1 & 0 \\
0 & 0
\end{pmatrix},
\qquad
Q =
\begin{pmatrix}
0 & 0 \\
0 & 1
\end{pmatrix}.
$$

### Important properties

Both projections satisfy:
$$
P^2 = P, \qquad Q^2 = Q,
$$
so they are **idempotent operators**.

They also satisfy:
$$
P \circ Q = Q \circ P = 0,
$$
since projecting onto one axis and then the other always gives the zero vector.

### Key takeaway

Projections provide a clear geometric example of linear transformations as vectors:

- They decompose space into independent components.
- They illustrate idempotent operators.
- They show how linear transformations can be combined through addition and composition.

Understanding these simple cases builds strong intuition for more complex linear operators.

## Combining projections

Let us continue working with the projections
$P, Q \in L(\mathbb{R}^2)$ defined by
$$
P(x,y) = (x,0), \qquad Q(x,y) = (0,y).
$$
Take a vector
$$
u = (x,y) \in \mathbb{R}^2.
$$

### Sum of projections

By definition of addition in $L(\mathbb{R}^2)$,
$$
(P+Q)(u) = P(u) + Q(u).
$$
Hence,
$$
(P+Q)(u) = (x,0) + (0,y) = (x,y) = u = I(u).
$$

Therefore,
$$
P + Q = I,
$$
where $I$ is the identity transformation.

**Geometric meaning:**
The sum of the projections onto the coordinate axes reconstructs the original vector.

### Scalar multiplication

If we scale $P$ by a scalar, for example $5$, then
$$
(5P)(u) = 5P(u) = 5(x,0) = (5x,0).
$$

**Geometric meaning:**
This is a stretch of the projection onto the $x$-axis by a factor of $5$.

### Difference of projections

Consider the transformation $P - Q$. Then
$$
(P-Q)(u) = P(u) - Q(u) = (x,0) - (0,y) = (x,-y).
$$

Thus, $P - Q$ is the linear transformation that **reflects vectors across the horizontal axis**.

Similarly,
$$
(Q-P)(u) = (0,y) - (x,0) = (-x,y),
$$
which corresponds to a **reflection across the vertical axis**.

### Idempotence of projections

Since $P$ is a projection,
$$
P^2(u) = P(P(u)) = P(x,0) = (x,0) = P(u).
$$
Hence,
$$
P^2 = P.
$$

The same argument shows that
$$
Q^2 = Q.
$$

So, every power of $P$ (or $Q$) is equal to itself:
$$
P^n = P, \qquad Q^n = Q \quad \text{for all } n \ge 1.
$$

### Composition of projections

Now let us compose the transformations.

On one hand,
$$
(P \circ Q)(u) = P(Q(u)) = P(0,y) = (0,0).
$$

On the other hand,
$$
(Q \circ P)(u) = Q(P(u)) = Q(x,0) = (0,0).
$$

Thus,
$$
P \circ Q = Q \circ P = O,
$$
where $O$ is the zero transformation defined by $O(u) = 0$ for all $u$.

**Important remark:**
Here the projections commute under composition, but this is a special case. In general, linear transformations do **not** commute.

### Powers of composed projections

Since
$$
P \circ Q = O,
$$
it follows immediately that
$$
(P \circ Q)^2 = O^2 = O,
$$
and, in fact,
$$
(P \circ Q)^n = O \quad \text{for all } n \ge 1.
$$

### Key takeaway

- Projections can be combined using addition, scalar multiplication, and composition.
- $P + Q = I$ decomposes and reconstructs vectors.
- $P - Q$ and $Q - P$ correspond to reflections.
- Projections are idempotent.
- Although $P$ and $Q$ commute under composition here, this behavior is exceptional in linear algebra.

## Conclusion

Let $V$ and $W$ be two vector spaces.

### Spaces of linear transformations

- $L(V,W)$ is the set of all linear transformations from $V$ into $W$.
- $L(V)$ is the set of all linear operators on $V$ (that is, linear transformations from $V$ to itself).

### Vector space structure of $L(V,W)$

The set $L(V,W)$ is a vector space with the following operations:

- **Addition**
  $$
  T + S : V \to W, \qquad (T+S)(v) = T(v) + S(v)
  $$

- **Scalar multiplication**
  $$
  \lambda T : V \to W, \qquad (\lambda T)(v) = \lambda\,T(v)
  $$

### Composition of linear transformations

If
$$
S \in L(U,V) \quad \text{and} \quad T \in L(V,W),
$$
then their composition belongs to $L(U,W)$ and is defined by
$$
T \circ S : U \to W, \qquad (T \circ S)(u) = T(S(u)).
$$

In general,
$$
T \circ S \neq S \circ T.
$$

### Powers of linear operators

If $T \in L(V)$, then its powers are defined by composition:
$$
\begin{aligned}
T^0 &= I, \\
T^1 &= T, \\
T^2 &= T \circ T, \\
&\;\;\vdots \\
T^n &= T \circ T^{\,n-1}.
\end{aligned}
$$

### Idempotent operators

A linear operator $T \in L(V)$ is called **idempotent** if
$$
T^2 = T.
$$

Such operators satisfy the property that applying them multiple times has the same effect as applying them once.
