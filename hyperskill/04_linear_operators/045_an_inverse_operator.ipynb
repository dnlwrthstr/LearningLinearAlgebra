# An inverse operator

Imagine you are developing some 3D graphics software. How can you transform the scene—for example, rotate it? It turns out that most transformations used in computer graphics are linear operators. This naturally leads to an important question:

> Is it possible to return to the previous scene *without* simply saving it somewhere?

In other words, after applying a linear transformation, can you apply another linear operator that **cancels** its effect?

This question leads directly to the concept of an **inverse operator**.

## Motivation: undoing a transformation

Suppose a linear transformation
$$
T : \mathbb{R}^3 \to \mathbb{R}^3
$$
represents a geometric operation such as a rotation, scaling, or shear of a 3D scene. Applying $T$ changes every vector (and hence every point in the scene).

To reverse this change, we want another linear transformation
$$
T^{-1} : \mathbb{R}^3 \to \mathbb{R}^3
$$
such that applying $T^{-1}$ after $T$ brings every vector back to its original position:
$$
T^{-1}(T(v)) = v \quad \text{for all } v \in \mathbb{R}^3.
$$

Likewise, applying $T$ after $T^{-1}$ should also do nothing:
$$
T(T^{-1}(v)) = v.
$$

## Inverse linear operator

A linear operator
$$
T \in L(V)
$$
is called **invertible** if there exists another linear operator
$$
T^{-1} \in L(V)
$$
such that
$$
T^{-1} \circ T = I
\quad \text{and} \quad
T \circ T^{-1} = I,
$$
where $I$ is the identity operator on $V$.

The operator $T^{-1}$ is called the **inverse** of $T$.

## Interpretation in computer graphics

- A **rotation** has an inverse rotation (rotate back by the opposite angle).
- A **scaling** by a nonzero factor has an inverse scaling (scale by the reciprocal).
- A **shear** may or may not be invertible, depending on its parameters.
- A **projection** is *not* invertible: once information is lost, it cannot be recovered by a linear operator.

Thus, the existence of an inverse operator tells you whether a transformation can be undone *purely mathematically*, without storing the original scene.


## Connection to matrices

Since every linear operator corresponds to a matrix, this idea will soon translate into the notion of an **inverse matrix**. An invertible transformation corresponds exactly to an invertible matrix, and the inverse operator corresponds to multiplying by the inverse matrix.

This connection will allow you to analyze reversibility of transformations using simple algebraic tools—and that is exactly where we are heading next.

## What is an inverse operator?

Let $V$ be a vector space and let
$$
A : V \to V
$$
be a linear operator.

### Definition (invertible operator)

The linear operator $A$ is said to be **invertible** if, for every vector
$$
y \in V,
$$
there exists a **unique** vector
$$
x \in V
$$
such that
$$
A(x) = y.
$$

In other words, $A$ is invertible if and only if:
- every vector in $V$ is reached by $A$ (surjective), and
- no two different vectors are mapped to the same vector (injective).

### Inverse operator

If the operator $A$ is invertible, then the mapping that sends each
$$
y \in V
$$
to the unique vector
$$
x \in V \quad \text{with} \quad A(x) = y
$$
is called the **inverse mapping** of $A$.

A fundamental result is:

> **If a linear operator is invertible, then its inverse mapping is also linear.**

### Characterization via composition

A linear operator $A : V \to V$ is invertible **if and only if** there exists a linear operator
$$
B : V \to V
$$
such that
$$
AB = BA = I,
$$
where $I$ is the identity operator on $V$.

In this case:
- $B$ is called the **inverse operator** of $A$, and
- it is denoted by
$$
A^{-1}.
$$

So the defining relations are:
$$
A^{-1}A = AA^{-1} = I.
$$

### Example

The **identity operator** $I : V \to V$, defined by
$$
I(v) = v \quad \text{for all } v \in V,
$$
is invertible. Its inverse is itself:
$$
I^{-1} = I.
$$

### Intuition

An inverse operator represents a transformation that *undoes* the effect of the original operator. Applying $A$ and then $A^{-1}$ (or vice versa) leaves every vector unchanged.

This concept will soon be mirrored exactly in matrix theory, where invertible linear operators correspond to **invertible matrices**, and inverse operators correspond to **inverse matrices**.

## Properties of an inverse operator

Let $V$ be a vector space and let
$$
A : V \to V
$$
be a linear operator. If $A$ is invertible, its inverse is denoted by $A^{-1}$.
Inverse operators satisfy the following fundamental properties.

### 1. Inverse of a product

If the linear operators
$$
A : V \to V \quad \text{and} \quad B : V \to V
$$
are both invertible, then their product $AB$ is also invertible, and
$$
(AB)^{-1} = B^{-1} A^{-1}.
$$

> **Important:** the order is reversed. This mirrors the behavior of matrix inverses.

### 2. Inverse of a scalar multiple

If $A$ is invertible and $\lambda \neq 0$ is a scalar, then the operator $\lambda A$ is invertible and
$$
(\lambda A)^{-1} = \lambda^{-1} A^{-1}.
$$

### 3. Inverse of the inverse

If a linear operator $A$ is invertible, then its inverse $A^{-1}$ is also invertible, and
$$
(A^{-1})^{-1} = A.
$$

### 4. Characterization via kernel and image

A linear operator
$$
A : V \to V
$$
is invertible **if and only if**
$$
\ker A = \{0\} \quad \text{and} \quad \operatorname{Im} A = V.
$$

That is:
- $A$ is **injective** (only the zero vector is mapped to zero), and
- $A$ is **surjective** (every vector in $V$ is reached).


### 5. Relation with matrices (finite-dimensional case)

If $A : V \to V$ is invertible and
$$
\dim V < \infty,
$$
then the matrix of the inverse operator $A^{-1}$ is the inverse of the matrix of $A$, taken with respect to the **same basis**.

In symbols, if $[A]$ denotes the matrix of $A$, then
$$
[A^{-1}] = [A]^{-1}.
$$

### Summary

Inverse operators behave exactly like inverse matrices:
- products reverse order,
- scalar multiples invert scalars,
- inverses are unique and involutive,
- invertibility is equivalent to injectivity and surjectivity,
- in finite dimensions, operator inversion corresponds directly to matrix inversion.

This tight connection is what allows us to move seamlessly between linear transformations and matrices.

## Example: invertibility of a linear operator on $\mathbb{R}^3$

Consider the linear operator $A : \mathbb{R}^3 \to \mathbb{R}^3$ defined by
$$
A x =
\begin{pmatrix}
2x_1 + 2x_2 - 3x_3 \\
-x_2 + 2x_3 \\
-x_1 - x_2 + x_3
\end{pmatrix},
\qquad
x=
\begin{pmatrix}
x_1\\x_2\\x_3
\end{pmatrix}.
$$

### Step 1: Write $A$ as a matrix

We can rewrite $Ax$ in the form $Ax = A\,x$ with
$$
A=
\begin{pmatrix}
2 & 2 & -3\\
0 & -1 & 2\\
-1 & -1 & 1
\end{pmatrix}.
$$

### Step 2: Decide invertibility

The operator $A$ is invertible **iff** the matrix $A$ is invertible, i.e. iff $A^{-1}$ exists.

To compute $A^{-1}$, form the augmented matrix $[A \mid I]$ and row-reduce:

$$
\left[\begin{array}{ccc|ccc}
2 & 2 & -3 & 1 & 0 & 0\\
0 & -1 & 2 & 0 & 1 & 0\\
-1 & -1 & 1 & 0 & 0 & 1
\end{array}\right]
\;\sim\;
\left[\begin{array}{ccc|ccc}
1 & 0 & 0 & 1 & 1 & 1\\
0 & 1 & 0 & -2 & -1 & -4\\
0 & 0 & 1 & -1 & 0 & -2
\end{array}\right].
$$

Since the left block reduces to $I$, the matrix is invertible, and the right block is the inverse.

### Step 3: Read off the inverse matrix and operator

Therefore,
$$
A^{-1}=
\begin{pmatrix}
1 & 1 & 1\\
-2 & -1 & -4\\
-1 & 0 & -2
\end{pmatrix}.
$$

So the inverse operator $A^{-1} : \mathbb{R}^3 \to \mathbb{R}^3$ is
$$
A^{-1}x =
\begin{pmatrix}
x_1 + x_2 + x_3\\
-2x_1 - x_2 - 4x_3\\
-x_1 - 2x_3
\end{pmatrix}.
$$

### Conclusion

Yes—this linear operator is invertible, because its matrix reduces to the identity and we explicitly obtained $A^{-1}$.

## Conclusion

In this topic, we studied the concept of an **inverse linear operator**.

- A linear operator $A : V \to V$ is **invertible** if there exists another linear operator $A^{-1} : V \to V$ such that
  $$
  A A^{-1} = A^{-1} A = I,
  $$
  where $I$ is the identity (unit) operator.

- The inverse operator is unique, and the order of composition with the original operator does **not** matter.

- We examined several key properties of inverse operators:
  - If $A$ and $B$ are invertible, then $AB$ is invertible and
    $$
    (AB)^{-1} = B^{-1} A^{-1}.
    $$
  - If $\lambda \neq 0$, then $\lambda A$ is invertible and
    $$
    (\lambda A)^{-1} = \lambda^{-1} A^{-1}.
    $$
  - If $A$ is invertible, then $(A^{-1})^{-1} = A$.
  - A linear operator is invertible if and only if
    $$
    \ker A = \{0\} \quad \text{and} \quad \operatorname{Im} A = V.
    $$

- For finite-dimensional spaces, a linear operator is invertible **if and only if** its matrix (in a given basis) is invertible, and the matrix of $A^{-1}$ is the inverse of the matrix of $A$.

- Finally, we worked through a complete example showing how to determine invertibility and compute the inverse operator by finding the inverse of the corresponding matrix using elementary row operations.

This concept is fundamental in both theory and applications, as inverse operators allow us to “undo” linear transformations, which is essential in areas such as solving linear systems, computer graphics, and numerical methods.
