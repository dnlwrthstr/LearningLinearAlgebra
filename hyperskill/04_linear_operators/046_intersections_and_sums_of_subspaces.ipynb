# Intersection and sums of subspaces

Subspaces are usually sets that we know quite well, such as **lines** and **planes**, or spaces that are easy to handleâ€”for example, the set of polynomials of degree less than or equal to $2$ within the set of all functions from $\mathbb{R}$ into $\mathbb{R}$.

Most of the time, subspaces arise as **solutions to specific problems**, such as the solution sets of systems of linear equations. For this reason, it is very common to need to **combine subspaces** with one another. Ideally, the result of such a combination should again be a subspace, so that we can continue to exploit its algebraic structure.

In this topic, we will explore the most useful ways to combine subspaces, their **geometric interpretations**, and the main theoretical results that will soon help us both **simplify calculations** and **understand more advanced concepts** in linear algebra.

## The Dimension of a Subspace

Until now, the subspaces we have encountered have usually been *small* parts of a vector space, such as **lines inside a plane**. The concept that allows us to make this intuition precise is **dimension**.

The dimension of a vector space is our formal notion of its *size*. Thanks to it, we can state a fundamental and intuitive fact:

If $U$ is a subspace of $V$, then
$$
\dim(U) \le \dim(V).
$$

## Intersection of Subspaces

Lines that pass through the origin are our prototype examples of vector subspaces. What happens if we **intersect** such subspaces?

Since every subspace contains the origin, the intersection will always contain $0$. But is that the *only* common point?

### Intersecting Lines

Consider two distinct lines through the origin. Unless the lines are identical, their only common point is
$$
\{0\}.
$$

Even this smallest possible intersection is still a subspace:
- it contains $0$,
- it is closed under addition,
- it is closed under scalar multiplication.

So $\{0\}$ itself is a vector subspace, called the **zero subspace**.

### Intersecting Planes

Now consider two planes through the origin. Their intersection also contains the origin, but in this case there are infinitely many points in common.

Geometrically, the intersection of two planes through the origin is typically a **line through the origin**. Once again, the result is a subspace.

The same reasoning applies if you intersect a plane and a line: the result (either a line through the origin or $\{0\}$) is still a subspace.

### General Result

This behavior is not a coincidence.

> **The intersection of any two subspaces $U$ and $W$ is itself a subspace.**

#### Proof
- Both $U$ and $W$ contain $0$, so $0 \in U \cap W$.
- If $u, v \in U \cap W$, then $u, v \in U$ and $u, v \in W$.
  Since both $U$ and $W$ are closed under addition,
  $$
  u + v \in U \quad \text{and} \quad u + v \in W,
  $$
  hence $u + v \in U \cap W$.
- If $\lambda$ is a scalar and $u \in U \cap W$, then
  $$
  \lambda u \in U \quad \text{and} \quad \lambda u \in W,
  $$
  so $\lambda u \in U \cap W$.

Thus, $U \cap W$ is a subspace.

### A Stronger Statement

The intersection operation is extremely robust:

> The intersection of **any collection** of subspaces (finite or infinite) is also a subspace.

In the worst case, the intersection is simply the zero subspace $\{0\}$, but it never fails to be a subspace.

## Sum of Subspaces

Even though the **union** of two subspaces usually fails to be a subspace (it is not stable under addition), it is still natural to ask how to combine two subspaces in a way that produces a **larger subspace** containing both of them.

Since stability under addition is essential, we define the following construction.

### Definition: Sum of Subspaces

Let $U$ and $W$ be subspaces of a vector space $V$. Their **sum** is defined as
$$
U + W = \{\, u + w \mid u \in U,\; w \in W \,\}.
$$

That is, $U + W$ consists of all vectors that can be written as the sum of one vector from $U$ and one vector from $W$.

### Key Property

> The sum of two subspaces is itself a subspace.

This is exactly what we want: a stable way to combine subspaces.

### Proof

Take two arbitrary vectors $v_1, v_2 \in U + W$.
By definition, there exist vectors
$$
u_1, u_2 \in U \quad \text{and} \quad w_1, w_2 \in W
$$
such that
$$
v_1 = u_1 + w_1, \qquad v_2 = u_2 + w_2.
$$

Now consider their sum:
$$
v_1 + v_2 = (u_1 + w_1) + (u_2 + w_2)
          = (u_1 + u_2) + (w_1 + w_2).
$$

Since $U$ and $W$ are subspaces, they are closed under addition:
$$
u_1 + u_2 \in U, \qquad w_1 + w_2 \in W.
$$
Hence,
$$
v_1 + v_2 \in U + W.
$$

A similar argument shows that scalar multiplication is preserved. Therefore, $U + W$ is a subspace.

Moreover, $U + W$ is **the smallest subspace that contains both $U$ and $W$**.

## Dimension Formula

The sum of subspaces interacts beautifully with dimension:
$$
\dim(U + W) = \dim(U) + \dim(W) - \dim(U \cap W).
$$

This formula captures the idea that dimensions add up, but the overlap $U \cap W$ must be subtracted to avoid double-counting.

## Decomposing a Subspace Using a Basis

The sum operation also helps us **break down** subspaces into simpler pieces.

Suppose a subspace $U$ has a basis
$$
\{u_1, u_2, \dots, u_k\}.
$$

Any vector $u \in U$ can be written as
$$
u = \sum_{j=1}^{k} \lambda_j u_j.
$$

Each term $\lambda_j u_j$ lies in the one-dimensional subspace $\operatorname{span}(u_j)$. Therefore,
$$
U = \operatorname{span}(u_1) + \operatorname{span}(u_2) + \cdots + \operatorname{span}(u_k).
$$

This shows that **every subspace is the sum of the one-dimensional subspaces generated by its basis vectors**, a powerful structural insight that will be used repeatedly in linear algebra.

## Direct Sums

Recall that **linear dependence** introduces redundancy: one vector can be written as a combination of others. This is why we prefer **linearly independent** vectors.
A very similar phenomenon occurs when **adding subspaces**: sometimes the information they contain overlaps.

For example, consider two planes in $\mathbb{R}^3$. Even though every vector in space can be written as a sum of a vector from each plane, the representation is not unique because the planes intersect in a line. This overlap is precisely the redundancy we want to avoid.

This motivates the notion of a **direct sum**.

### Definition: Direct Sum

Let $U$ and $W$ be subspaces of a vector space $V$.
Their sum is called **direct** if
$$
U \cap W = \{0\}.
$$

In this case, we write
$$
U \oplus W.
$$

#### Key Property (Uniqueness)

If $V = U \oplus W$, then **every vector $v \in V$ can be written uniquely** as
$$
v = u + w, \quad u \in U,\; w \in W.
$$

This mirrors the idea of linearly independent vectors: no overlap means no redundancy.

### Dimension Formula for Direct Sums

When the sum is direct, the dimension formula simplifies to
$$
\dim(U \oplus W) = \dim(U) + \dim(W).
$$

This is extremely useful: if we decompose a space into well-understood subspaces, we can compute the dimension of the whole space by simple addition.

You can think of a **direct sum** as the linear-algebra analogue of the union of **disjoint sets**.

### Examples

#### Example 1: The Plane $\mathbb{R}^2$

The canonical basis of $\mathbb{R}^2$ is $\{e_1, e_2\}$. Hence,
$$
\mathbb{R}^2 = \operatorname{span}(e_1) \oplus \operatorname{span}(e_2).
$$

These two one-dimensional subspaces intersect only at the origin, so the sum is direct.

#### Example 2: The Space $\mathbb{R}^3$

We can decompose $\mathbb{R}^3$ as
$$
\mathbb{R}^3
= \operatorname{span}(e_1, e_2) \oplus \operatorname{span}(e_3)
= \operatorname{span}(e_1) \oplus \operatorname{span}(e_2) \oplus \operatorname{span}(e_3).
$$

This decomposition reflects the geometric intuition of a plane plus a perpendicular axis.

#### Example 3: Even and Odd Polynomials

Let $P$ be the vector space of polynomials of degree $\le n$.

- A polynomial $f$ is **even** if
  $$
  f(-x) = f(x).
  $$
- A polynomial $f$ is **odd** if
  $$
  f(-x) = -f(x).
  $$

Let:
- $P_E$ be the subspace of even polynomials, with basis
  $$
  \{1, x^2, x^4, \dots\},
  $$
- $P_O$ be the subspace of odd polynomials, with basis
  $$
  \{x, x^3, x^5, \dots\}.
  $$

The union of these bases forms a basis of $P$, and the intersection is only $\{0\}$. Therefore,
$$
P = P_E \oplus P_O.
$$

Every polynomial can be uniquely decomposed into an even part and an odd part.

#### Example:
$$
x^3 + 4x^2 + 5x + 1
= (4x^2 + 1) + (x^3 + 5x),
$$
where the first polynomial is even and the second is odd.

### Summary

- The **direct sum** $U \oplus W$ means $U + W$ with no overlap except $\{0\}$.
- Every vector has a **unique decomposition** into components from each subspace.
- Dimensions add:
  $$
  \dim(U \oplus W) = \dim(U) + \dim(W).
  $$
- Direct sums are a powerful way to decompose vector spaces into simpler, non-redundant pieces.

## Conclusion

Let $V$ be a vector space and let $U$ and $W$ be subspaces of $V$. The following fundamental results summarize what we have learned:

- **Dimension bound**
  $$
  \dim(U) \le \dim(V).
  $$

- **Intersection**

  The intersection
  $$
  U \cap W
  $$
  is always a subspace of $V$.

- **Union**

  The union
  $$
  U \cup W
  $$
  is a subspace **if and only if**
  $$
  U \subset W \quad \text{or} \quad W \subset U.
  $$

- **Sum of subspaces**

  The sum of $U$ and $W$ is defined as
  $$
  U + W = \{\, u + w \mid u \in U,\; w \in W \,\},
  $$
  and this set is always a subspace of $V$.

- **Dimension of a sum**
  $$
  \dim(U + W) = \dim(U) + \dim(W) - \dim(U \cap W).
  $$

- **Direct sum**

  The sum of $U$ and $W$ is called **direct** if
  $$
  U \cap W = \{0\},
  $$
  and in this case it is denoted by
  $$
  U \oplus W.
  $$

- **Dimension of a direct sum**
  $$
  \dim(U \oplus W) = \dim(U) + \dim(W).
  $$

These results provide the essential tools for combining and decomposing subspaces, both algebraically and geometrically.
