# Eigenvalues and eigenvectors

Have you ever played a game of billiards? When you hit another ball using the cue ball, it moves and changes direction. Similarly, in mathematics, a **linear operator** acts like a function on vectors in a vector space. You apply the operator to a vector, and it changes its direction.

*Billiard-table illustration*

However, among all vectors, there is a special kind called **eigenvectors** that behave differently. In this topic, we will explore eigenvectors in detail. You will see how to formally define eigenvectors and how to find them. The topics you have already learned will help you calculate these values.

By understanding eigenvectors, you can gain insight into how linear operators behave and understand their role in a wide range of applications, from physics to data analysis.

## What are eigenvalues and eigenvectors?

An **eigenvector** $v$ is a vector that, when transformed by a linear operator
$A : V \to V$, is only **scaled** by a scalar value. This scalar $\lambda$ is called the **eigenvalue** of the transformation.

In other words, an eigenvector is a **non-zero** vector that, when multiplied by a matrix or linear transformation, results in a scalar multiple of itself. The eigenvalue represents the factor by which the eigenvector is scaled.

In the billiards analogy, this corresponds to hitting a ball perfectly straight so that it continues moving in the same direction as the shot, only faster or slower.

### Example

Consider the matrix $A$ and the vector $v$:

$$
A =
\begin{pmatrix}
3 & 1 \\
1 & 3
\end{pmatrix},
\qquad
v =
\begin{pmatrix}
1 \\
1
\end{pmatrix}
$$

Now compute the product $Av$:

$$
Av =
\begin{pmatrix}
3 & 1 \\
1 & 3
\end{pmatrix}
\begin{pmatrix}
1 \\
1
\end{pmatrix}
=
\begin{pmatrix}
4 \\
4
\end{pmatrix}
$$

Observe that

$$
Av = 4v
$$

This shows that $v$ is an **eigenvector** of $A$, and the corresponding **eigenvalue** is

$$
\lambda = 4
$$

### Important remark

A matrix $A$ must be **square** in order to have eigenvalues and eigenvectors.

## Finding all eigenvalues of a linear operator

Now let's find the eigenvalues of a linear operator $A$. The set of all eigenvalues of $A$ is called its **spectrum** and is denoted by $\mathrm{Spec}(A)$.

To illustrate the method, we will compute both eigenvalues and eigenvectors for the matrix

$$
A=\begin{pmatrix}3 & 1\\ 2 & 2\end{pmatrix}.
$$

### Step 1: Start from the eigenvalue equation

Eigenvectors $v\neq 0$ and eigenvalues $\lambda$ satisfy

$$
Av=\lambda v.
$$

Assume

$$
v=\begin{pmatrix}x\\y\end{pmatrix}.
$$

Then

$$
\begin{pmatrix}3 & 1\\ 2 & 2\end{pmatrix}
\begin{pmatrix}x\\y\end{pmatrix}
=
\lambda
\begin{pmatrix}x\\y\end{pmatrix}.
$$

Written as a system:

$$
\begin{cases}
3x+y=\lambda x,\\
2x+2y=\lambda y.
\end{cases}
$$

### Step 2: Move everything to one side

Bring the right-hand side to the left:

$$
(A-\lambda I)v=0,
$$

which gives

$$
\begin{cases}
(3-\lambda)x+y=0,\\
2x+(2-\lambda)y=0.
\end{cases}
$$

### Step 3: Use the characteristic equation

A non-trivial solution $v\neq 0$ exists only when

$$
\det(A-\lambda I)=0.
$$

Compute:

$$
\det\!\begin{pmatrix}3-\lambda & 1\\ 2 & 2-\lambda\end{pmatrix}
=(3-\lambda)(2-\lambda)-2=0.
$$

Expand and factor:

$$
(3-\lambda)(2-\lambda)-2
=\lambda^2-5\lambda+4
=(\lambda-1)(\lambda-4)=0.
$$

So the eigenvalues are

$$
\lambda_1=1,\qquad \lambda_2=4,
$$

and therefore

$$
\mathrm{Spec}(A)=\{1,4\}.
$$

## Finding eigenvectors

### Eigenvectors for $\lambda=1$

Solve $(A-I)v=0$:

$$
A-I=
\begin{pmatrix}
2 & 1\\
2 & 1
\end{pmatrix}.
$$

So

$$
2x+y=0 \quad \Longrightarrow \quad x=-\frac{y}{2}.
$$

Let $y=2t$. Then $x=-t$, and the eigenvectors are

$$
v=t\begin{pmatrix}-1\\2\end{pmatrix},\qquad t\in\mathbb{R}.
$$

Thus the eigenspace is

$$
E_{\lambda=1}=\mathrm{span}\left\{\begin{pmatrix}-1\\2\end{pmatrix}\right\}.
$$

### Eigenvectors for $\lambda=4$

Solve $(A-4I)v=0$:

$$
A-4I=
\begin{pmatrix}
-1 & 1\\
2 & -2
\end{pmatrix}.
$$

So

$$
-x+y=0 \quad \Longrightarrow \quad y=x.
$$

Let $x=t$. Then

$$
v=t\begin{pmatrix}1\\1\end{pmatrix},\qquad t\in\mathbb{R}.
$$

Thus the eigenspace is

$$
E_{\lambda=4}=\mathrm{span}\left\{\begin{pmatrix}1\\1\end{pmatrix}\right\}.
$$

## Summary

- **Eigenvalues:** $\lambda=1$ and $\lambda=4$
- **Eigenvectors:**
  - For $\lambda=1$: $\mathrm{span}\left\{\begin{pmatrix}-1\\2\end{pmatrix}\right\}$
  - For $\lambda=4$: $\mathrm{span}\left\{\begin{pmatrix}1\\1\end{pmatrix}\right\}$

## Conclusion

In this topic, you learned about **eigenvalues**, **eigenvectors**, and how to calculate them. The most important points to remember are:

- An **eigenvector** is a nonzero vector that, when transformed by a linear operator, is only scaled by a scalar value:
  $$
  Av = \lambda v
  $$

- The scalar $\lambda$ is called the **eigenvalue** of the transformation.

- Eigenvalues are found by solving the **characteristic equation**:
  $$
  \det(A - \lambda I) = 0
  $$

- This leads to the **characteristic polynomial**:
  $$
  \xi_A(\lambda)
  = b_n \lambda^n + b_{n-1} \lambda^{n-1} + \dots + b_1 \lambda + b_0
  $$

- The set of all eigenvalues of a linear operator $A$ is called its **spectrum**.
