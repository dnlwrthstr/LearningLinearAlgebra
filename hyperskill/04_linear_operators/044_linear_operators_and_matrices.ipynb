# Linear operators and matrices

You will finally discover the intimate connection between **matrices** and **linear transformations**, which is fundamental to understanding both. With this connection, you will be able to analyze a system of equations completely using just a few calculations.

Now you will begin to reap the fruits of your labor. You will put into practice almost everything you have learned so far. This topic will be much easier than the previous ones: all the hard work is already done — now it is time to enjoy the results.

## The transformation generated by a matrix

Recent topics emphasized that any $m \times n$ matrix $A$ induces a linear transformation $L_A$ thanks to the properties of matrix multiplication:

$$
L_A : \mathbb{R}^n \to \mathbb{R}^m,
\qquad
L_A(v) = Av,
$$

where the vector $v$ is written in terms of the canonical basis as

$$
v = v_1 e_1 + \cdots + v_n e_n.
$$

Furthermore, the **null space** and the **range** of $L_A$ encode important information about the linear system associated with $A$.
To streamline notation, instead of focusing on individual entries of $A$, we now emphasize its **columns**.

Let the columns of $A$ be denoted by

$$
A_1, \dots, A_n \in \mathbb{R}^m,
$$

and write the matrix as

$$
A = [\, A_1 \mid \cdots \mid A_n \,].
$$

Each column $A_i$ is a vector in $\mathbb{R}^m$.

### Action on the canonical basis

Recall that a linear transformation is completely determined by its values on a basis.
Let us apply this idea to $L_A$ by evaluating it on the canonical basis vectors.

For the first basis vector $e_1$:

$$
L_A(e_1) = A e_1 = A_1.
$$

#### Proof

By definition of matrix multiplication,

$$
A e_1
=
\begin{bmatrix}
(a_{11}, \dots, a_{1n}) \cdot e_1 \\
\vdots \\
(a_{m1}, \dots, a_{mn}) \cdot e_1
\end{bmatrix}
=
\begin{bmatrix}
a_{11} \\
\vdots \\
a_{m1}
\end{bmatrix}
= A_1.
$$

Thus, transforming $e_1$ retrieves the **first column** of $A$.
The same argument applies to all basis vectors.

### Key result

**The values of $L_A$ on the canonical basis vectors are precisely the columns of $A$.**

### Linear combination form

As a direct consequence, the matrix–vector product can be written as a linear combination of the columns of $A$:

$$
L_A(v) = Av = v_1 A_1 + \cdots + v_n A_n \in \mathbb{R}^m.
$$

This interpretation is fundamental:
multiplying a matrix by a vector is equivalent to forming a linear combination of the matrix columns, with coefficients given by the components of the vector.

## The matrix of a transformation

There is something remarkable going on here. You already know that **all the information** of an arbitrary linear transformation

$$
T : \mathbb{R}^n \to \mathbb{R}^m
$$

is contained in its values on a basis. Indeed, for any vector

$$
v = v_1 e_1 + \cdots + v_n e_n,
$$

linearity gives

$$
T(v) = v_1 T(e_1) + \cdots + v_n T(e_n).
$$

So, if you know the set of vectors

$$
\{\, T(e_1), \ldots, T(e_n) \,\},
$$

you can reconstruct the transformation completely.

Now look closely at this formula. It is **strikingly similar** to the expression you previously obtained for the action of a matrix on a vector as a linear combination of its columns. This observation leads to an unavoidable conclusion:

> **The natural place to store the values of a linear transformation on a basis is a matrix.**

### Definition: matrix associated with a transformation

The **matrix of the linear transformation** $T$ (with respect to the canonical bases) is defined as

$$
[T] = \big[\, T(e_1) \mid \cdots \mid T(e_n) \,\big].
$$

That is, the columns of $[T]$ are exactly the images of the canonical basis vectors under $T$.

### Fundamental identity

With this definition, the action of $T$ on any vector $v \in \mathbb{R}^n$ can be written as

$$
\begin{aligned}
T(v)
&= v_1 T(e_1) + \cdots + v_n T(e_n) \\
&= [T]\,v \\
&= L_{[T]}(v).
\end{aligned}
$$

Thus, applying the transformation $T$ to a vector is **exactly the same** as multiplying that vector by the matrix $[T]$.

### Key consequences

- To fully determine a linear transformation $T : \mathbb{R}^n \to \mathbb{R}^m$, it is enough to compute its values on the basis vectors and place them as columns in a matrix.
- Every $m \times n$ matrix defines a linear transformation from $\mathbb{R}^n$ to $\mathbb{R}^m$.
- Conversely, every linear transformation between these spaces is completely determined by an $m \times n$ matrix.

### Big picture

You can now freely move between **algebra** and **geometry**:

- A matrix is a concrete representation of a linear transformation.
- A linear transformation is an abstract interpretation of a matrix acting on vectors.

This equivalence is one of the central pillars of linear algebra.

## Calculating the matrix of a transformation

Before taking a closer look at the relationship between matrices and linear transformations, let’s work through some concrete examples.

### Example 1: A transformation from $\mathbb{R}^2$ to $\mathbb{R}^3$

Consider the linear transformation

$$
T : \mathbb{R}^2 \to \mathbb{R}^3,
\qquad
T(x,y) =
\begin{pmatrix}
x - y \\
3x + 2y \\
y
\end{pmatrix}.
$$

#### Step 1: Evaluate the transformation on the canonical basis

The canonical basis of $\mathbb{R}^2$ is
$$
e_1 = (1,0), \qquad e_2 = (0,1).
$$

Compute:

$$
T(e_1) = T(1,0) =
\begin{pmatrix}
1 \\
3 \\
0
\end{pmatrix},
\qquad
T(e_2) = T(0,1) =
\begin{pmatrix}
-1 \\
2 \\
1
\end{pmatrix}.
$$

#### Step 2: Build the matrix of $T$

The matrix of $T$ is formed by taking these vectors as columns:

$$
[T] = \big[\, T(e_1) \mid T(e_2) \,\big]
=
\begin{pmatrix}
1 & -1 \\
3 &  2 \\
0 &  1
\end{pmatrix}.
$$

#### Interpretation

Suppose you want to solve the system of linear equations

$$
[T]\,x = b,
\qquad
x \in \mathbb{R}^2,\; b \in \mathbb{R}^3.
$$

- The columns of $[T]$ are **linearly independent**, so the range of $T$ is a **plane** in $\mathbb{R}^3$.
- By the dimension theorem, the null space of $T$ is $\{0\}$, hence $T$ is **injective**.

Consequences:

- The system $[T]x = b$ **does not always have a solution**.
  For example, if $b = (0,0,0)^T$, there is no solution.
- Whenever a solution exists, it is **unique**.
  For example, if
  $$
  b = (0,5,1)^T,
  $$
  the unique solution is
  $$
  x = (1,1)^T.
  $$

From the matrix alone, you can already deduce important properties of the transformation.


### Example 2: A linear operator on $\mathbb{R}^3$

Now consider the operator

$$
T : \mathbb{R}^3 \to \mathbb{R}^3,
\qquad
T(x,y,z) =
\begin{pmatrix}
x + y + z \\
2y - 3z \\
z
\end{pmatrix}.
$$

#### Step 1: Transform the canonical basis

The canonical basis of $\mathbb{R}^3$ is
$$
e_1 = (1,0,0), \quad e_2 = (0,1,0), \quad e_3 = (0,0,1).
$$

Compute:

$$
T(e_1) =
\begin{pmatrix}
1 \\
0 \\
0
\end{pmatrix},
\qquad
T(e_2) =
\begin{pmatrix}
1 \\
2 \\
0
\end{pmatrix},
\qquad
T(e_3) =
\begin{pmatrix}
1 \\
-3 \\
1
\end{pmatrix}.
$$


#### Step 2: Build the matrix of $T$

$$
[T]
=
\big[\, T(e_1) \mid T(e_2) \mid T(e_3) \,\big]
=
\begin{pmatrix}
1 & 1 & 1 \\
0 & 2 & -3 \\
0 & 0 & 1
\end{pmatrix}.
$$


#### Interpretation

From the columns of $[T]$, it is clear that they span all of $\mathbb{R}^3$. Hence:

- The range of $T$ is $\mathbb{R}^3$.
- The system
  $$
  [T]x = b
  $$
  has a solution for **every** $b \in \mathbb{R}^3$.
- By the dimension theorem, the null space of $T$ is $\{0\}$, so $T$ is **injective**.
- Since $T$ is an operator on $\mathbb{R}^3$ and is both injective and surjective, it is **bijective**.

Therefore, the system $[T]x = b$ always has a **unique solution**.

### Key takeaway

By computing the matrix of a linear transformation, you can immediately determine:

- the dimension of its range,
- whether solutions exist for $[T]x = b$,
- and whether those solutions are unique.

This is the power of the matrix–transformation correspondence.

## Two sides of the same coin

The previous examples illustrate the strength of the relationship between **linear transformations** and **matrices**.

Consider:

- The vector space of all linear transformations from $\mathbb{R}^n$ to $\mathbb{R}^m$, denoted
  $$
  L(\mathbb{R}^n, \mathbb{R}^m),
  $$
- The vector space of all $m \times n$ matrices, denoted
  $$
  M_{m \times n}.
  $$

You already know that **both are vector spaces**. The key insight is that they are not just similar—they are **essentially the same**, connected by a natural linear mapping.

## The matrix assignment map

Define a function that assigns to each linear transformation its matrix with respect to the canonical bases:

$$
\Phi : L(\mathbb{R}^n, \mathbb{R}^m) \to M_{m \times n},
\qquad
\Phi(T) = [T].
$$

This map preserves the vector space structure:

### Linearity

For all $S, T \in L(\mathbb{R}^n, \mathbb{R}^m)$ and all scalars $\lambda$,

$$
[T + S] = [T] + [S],
\qquad
[\lambda T] = \lambda [T].
$$

### Proof of linearity

Let $v \in \mathbb{R}^n$ with
$$
v = v_1 e_1 + \cdots + v_n e_n.
$$

**Addition:**
\begin{align*}
(T + S)(v)
&= v_1 (T + S)(e_1) + \cdots + v_n (T + S)(e_n) \\
&= v_1 T(e_1) + \cdots + v_n T(e_n)
  + v_1 S(e_1) + \cdots + v_n S(e_n) \\
&= [T]v + [S]v \\
&= ([T] + [S])v.
\end{align*}

Thus,
$$
[T + S] = [T] + [S].
$$

**Scalar multiplication:**
\begin{align*}
(\lambda T)(v)
&= v_1 (\lambda T)(e_1) + \cdots + v_n (\lambda T)(e_n) \\
&= \lambda \big( v_1 T(e_1) + \cdots + v_n T(e_n) \big) \\
&= \lambda ([T]v) \\
&= (\lambda [T])v.
\end{align*}

Hence,
$$
[\lambda T] = \lambda [T].
$$

## An isomorphism of vector spaces

The map
$$
\Phi : L(\mathbb{R}^n, \mathbb{R}^m) \to M_{m \times n}
$$
is:

### Injective
If $[T] = 0$, then $T(e_i) = 0$ for all basis vectors $e_i$, which implies $T(v) = 0$ for all $v$.
Therefore, $T = 0$.

### Surjective
Given any matrix $A \in M_{m \times n}$, define a transformation $T$ by sending each basis vector $e_i$ to the $i$-th column of $A$.
Then $[T] = A$.

### Consequences

- $\Phi$ is a **vector space isomorphism**
- The two spaces have the same dimension:
  $$
  \dim L(\mathbb{R}^n, \mathbb{R}^m) = \dim M_{m \times n} = mn
  $$
- Linear transformations and matrices are **two representations of the same object**

This allows you to freely move between:
- geometric intuition (transformations),
- algebraic computation (matrices).

## Why matrix multiplication works the way it does

Matrix multiplication is not arbitrary. It is **designed to model composition of linear transformations**.

Let
$$
S : \mathbb{R}^m \to \mathbb{R}^n,
\qquad
T : \mathbb{R}^n \to \mathbb{R}^p
$$
be linear transformations.

Then their composition satisfies the fundamental identity:

$$
[T \circ S] = [T]\,[S].
$$

This explains:
- why matrix multiplication is associative,
- why it is generally **not commutative**,
- and why it perfectly mirrors function composition.

## Final insight

Matrices and linear transformations are **two sides of the same coin**:

- Matrices encode transformations.
- Transformations explain matrix operations.
- Composition corresponds to multiplication.
- Vector space structure is preserved on both sides.

Once you see this, much of linear algebra becomes not just easier—but inevitable.

## Conclusion

Let
$$
L(\mathbb{R}^n, \mathbb{R}^m)
$$
be the set of all linear transformations from $\mathbb{R}^n$ into $\mathbb{R}^m$, and let
$$
M_{m \times n}
$$
be the set of all $m \times n$ matrices.

Let
$$
T \in L(\mathbb{R}^n, \mathbb{R}^m),
\qquad
A \in M_{m \times n}.
$$

### Matrix–transformation correspondence

The linear transformation generated by a matrix $A$ is defined as
$$
L_A : \mathbb{R}^n \to \mathbb{R}^m,
\qquad
L_A(v) = Av.
$$

If the columns of $A$ are denoted by
$$
A = [A_1 \mid A_2 \mid \cdots \mid A_n],
$$
then the values of $L_A$ on the canonical basis are exactly the columns of $A$.

For any vector
$$
v = v_1 e_1 + \cdots + v_n e_n,
$$
we have
$$
L_A(v) = Av = v_1 A_1 + \cdots + v_n A_n.
$$


### Matrix of a linear transformation

For a linear transformation
$$
T : \mathbb{R}^n \to \mathbb{R}^m,
$$
its associated matrix is defined by
$$
[T] = [\, T(e_1) \mid \cdots \mid T(e_n) \,].
$$

This matrix completely determines the transformation, since
$$
T(v) = [T]\,v
$$
for all $v \in \mathbb{R}^n$.


### Compatibility with vector space operations

For all linear transformations $T, S \in L(\mathbb{R}^n, \mathbb{R}^m)$ and all scalars $\lambda$,

$$
[T + S] = [T] + [S],
\qquad
[\lambda T] = \lambda [T].
$$

Thus, the correspondence between transformations and matrices preserves addition and scalar multiplication.


### Isomorphism

There is a vector space isomorphism between
$$
L(\mathbb{R}^n, \mathbb{R}^m)
\quad \text{and} \quad
M_{m \times n}.
$$

Linear transformations and matrices can therefore be regarded as the **same mathematical objects**, viewed from two different perspectives:
- geometric (transformations),
- algebraic (matrices).


### Composition and matrix multiplication

If
$$
S : \mathbb{R}^m \to \mathbb{R}^n,
\qquad
T : \mathbb{R}^n \to \mathbb{R}^p
$$
are linear transformations, then composition corresponds exactly to matrix multiplication:

$$
[T \circ S] = [T][S].
$$

This identity explains why matrix multiplication is defined the way it is: it mirrors the composition of linear transformations.
