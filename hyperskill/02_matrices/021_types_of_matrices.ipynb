# Types of matrices

Matrices, like any other mathematical tool, make our lives easier. Of course, they are not as simple as everyday operations such as multiplication or division, which help us calculate things like how much 2 kilograms of apples cost. Still, matrices are present all around us in many more sophisticated processes.

Every time you surf the web, some data is encrypted—and for that, we need matrices. When an economist prepares a report or data is analyzed, matrices once again play a crucial role. Imagine a manager in a grocery store who wants to compare customers based on their purchases. One possible approach is to represent each receipt as a matrix, where each entry corresponds to the quantity of a particular product purchased. These matrices can then be compared and analyzed.

Now that you have seen how broadly applicable matrices are, we hope you are motivated to continue learning about different types of matrices and the important concepts associated with them.

## Basic Matrices

A **square matrix** of size $n \times n$ is a matrix that has the same number $n$ of rows and columns.

For example, the following matrix is a square matrix of size $3 \times 3$:
$$
\begin{pmatrix}
1 & 0 & 3 \\
2 & -1 & 4 \\
11 & 7 & 11
\end{pmatrix}
$$


A **zero matrix** is a matrix in which every element is equal to $0$.

For example, the $3 \times 3$ zero matrix is
$$
\begin{pmatrix}
0 & 0 & 0 \\
0 & 0 & 0 \\
0 & 0 & 0
\end{pmatrix}
$$

## Diagonal Matrices

The **leading entry** of a row is the first non-zero element in that row.
For example, in the matrix below, the leading entries are highlighted conceptually (they are the first non-zero elements in each row):

$$
\begin{pmatrix}
1 & 2 & 6 & 3 \\
0 & 0 & 0 & 7 \\
0 & 0 & 0 & 0 \\
0 & 1 & -1 & 0
\end{pmatrix}
$$

### Diagonal Matrix

A **diagonal matrix** is a **square matrix** in which:

- all non-zero elements are located on the **main diagonal**,
- all other elements are equal to zero.

The main diagonal runs from the **top-left corner to the bottom-right corner**, and its elements have indices
$$
a_{1,1}, a_{2,2}, a_{3,3}, \dots, a_{n,n}.
$$

Example of a diagonal matrix:
$$
\begin{pmatrix}
1 & 0 & 0 \\
0 & 3 & 0 \\
0 & 0 & 7
\end{pmatrix}.
$$

### Scalar Matrix

A **scalar matrix** is a diagonal matrix in which **all diagonal elements are equal to the same scalar**.

Example:
$$
\begin{pmatrix}
7 & 0 & 0 \\
0 & 7 & 0 \\
0 & 0 & 7
\end{pmatrix}.
$$

### Diagonal Elements in Non-Square Matrices

Even for a non-square $n \times m$ matrix, diagonal elements are defined.
They are the elements
$$
a_{1,1}, a_{2,2}, a_{3,3}, \dots, a_{\min(n,m),\min(n,m)}.
$$

Examples:
$$
\begin{pmatrix}
9 & 0 & 0 \\
0 & 13 & 0 \\
0 & 0 & -7
\end{pmatrix},
\quad
\begin{pmatrix}
1 & 0 & 0 \\
0 & 3 & 0 \\
0 & 0 & 5 \\
0 & 0 & 0
\end{pmatrix},
\quad
\begin{pmatrix}
-1 & 0 & 0 & 0 \\
0 & 42 & 0 & 0 \\
0 & 0 & 8 & 0
\end{pmatrix}.
$$

### Secondary Diagonal

The **secondary diagonal** of a square matrix runs from the **bottom-left corner to the top-right corner**.
Its elements have indices
$$
a_{n,1}, a_{n-1,2}, a_{n-2,3}, \dots, a_{1,n}.
$$

Example:
$$
\begin{pmatrix}
0 & 0 & -3 \\
0 & 4 & 0 \\
6 & 0 & 0
\end{pmatrix}.
$$

### Identity Matrix

The **identity matrix** (sometimes called a *unit matrix*) of size $n$ is the $n \times n$ matrix with:

- $1$ on the main diagonal,
- $0$ everywhere else.

It is denoted by $I_n$, or simply $I$ when the size is clear.

Example of a $3 \times 3$ identity matrix:
$$
I_3 =
\begin{pmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{pmatrix}.
$$

The name *identity matrix* comes from the fact that multiplying any compatible matrix by $I$ leaves it unchanged:
$$
AI = IA = A.
$$

In this sense, the identity matrix plays the same role for matrices as the number $1$ does for real numbers.

## Triangular Matrices

### Upper Triangular Matrix

An **upper triangular matrix** is a square matrix in which all elements **below the main diagonal** are equal to zero:
$$
a_{i,j} = 0 \quad \text{if } i > j.
$$
All other elements (on or above the diagonal) can be arbitrary, including zero.

Example:
$$
\begin{pmatrix}
7 & 2 & 1 & 0 \\
0 & 0 & 3 & 5 \\
0 & 0 & 7 & 4 \\
0 & 0 & 0 & 6
\end{pmatrix}.
$$

### Lower Triangular Matrix

A **lower triangular matrix** is a square matrix in which all elements **above the main diagonal** are equal to zero:
$$
a_{i,j} = 0 \quad \text{if } i < j.
$$
All other elements (on or below the diagonal) can be arbitrary.

Example:
$$
\begin{pmatrix}
7 & 0 & 0 & 0 \\
3 & 7 & 0 & 0 \\
1 & 2 & 0 & 0 \\
6 & 0 & 8 & 9
\end{pmatrix}.
$$

### Symmetric Matrix

A **symmetric matrix** is a square matrix that is equal to its transpose:
$$
a_{i,j} = a_{j,i}.
$$

Example:
$$
\begin{pmatrix}
7 & 0 & 1 \\
0 & 7 & 2 \\
1 & 2 & 2
\end{pmatrix}.
$$

### Relationships Between Matrix Types

- Square **zero matrices** and **identity matrices** are special cases of **scalar matrices**.
- **Scalar matrices** are special cases of **diagonal matrices**.
- **Diagonal matrices** are special cases of:
  - lower triangular matrices,
  - upper triangular matrices,
  - symmetric matrices.

These relationships are important for understanding matrix structure and for simplifying computations in linear algebra.

## Conclusion

Let’s summarize the key ideas from this topic. We explored several important types of matrices, their defining properties, and why they are useful in practice.

- A **square matrix** is a matrix that has the same number of rows and columns. Its dimension is
  $$
  n \times n.
  $$

- A **zero matrix** is a matrix in which **all entries are equal to zero**.

- A **leading entry** of a row is the **first nonzero element** in that row (reading from left to right).

- A **diagonal matrix** is a square matrix whose **only possibly nonzero entries lie on the main diagonal**.
  A matrix with nonzero entries on the **secondary diagonal** is called a **secondary diagonal matrix**.

- An **identity matrix** is a square diagonal matrix with all diagonal entries equal to $1$.
  Multiplying any compatible matrix by the identity matrix leaves it unchanged:
  $$
  A I = I A = A.
  $$

- **Triangular matrices** are square matrices in which all entries are zero
  - **below the main diagonal** (upper triangular), or
  - **above the main diagonal** (lower triangular).

Understanding these matrix types is fundamental for later topics such as solving linear systems, matrix factorizations, determinants, and linear transformations.
