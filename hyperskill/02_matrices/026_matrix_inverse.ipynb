# Matrix inverse

As you already know, multiplying matrices is a much more difficult task than multiplying numbers.
However, in this topic, you will learn about the concept of an **inverse matrix**, which can drastically simplify these calculations in certain cases.

You will also learn its most important application: **finding the unique solution of a square system of linear equations**.
In fact, whenever the solution is unique, the inverse matrix is the perfect tool to compute it **explicitly**.

## Developing the Definition

When you multiply any number by $1$, you get the same number. Moreover, almost every number $x$ (except zero) has a partner $y$ such that
$$
xy = 1,
$$
which is known as the **multiplicative inverse** of $x$. Thanks to this fact, many equations can be solved easily.

But how can we do something similar with matrices? To achieve the same flexibility, we need to be able to *“cancel”* matrices in equations. In other words, we need to multiply matrices by their inverses.

### Identity Matrix as the Neutral Element

As a first step, the **identity matrix** $I$ plays the role of the number $1$, because you already know that
$$
AI = IA = A.
$$

The next step is clear: to *neutralize* a square matrix $A$, we need another matrix $B$ such that
$$
AB = I.
$$

However, matrix multiplication is not commutative, so $B$ must also satisfy
$$
BA = I.
$$

### Definition of the Inverse Matrix

Let $A$ be a square matrix of size $n$.
An **inverse** of $A$ is a square matrix $B$ of the same size $n$ such that
$$
AB = I \quad \text{and} \quad BA = I.
$$

When a matrix has an inverse, it is called **invertible** or **non-singular**.

### Matrices Without Inverses

Just as the number $0$ has no inverse, the zero matrix also has no inverse, since
$$
0A = 0
$$
for any matrix $A$.

But the zero matrix is not the only one without an inverse. Many matrices are non-invertible. For example, consider a matrix $C$ that has an entire row of zeros, say the $i$-th row. No matter which matrix $B$ you choose, the $i$-th row of the product $CB$ will always be zero. Therefore, $CB$ can never equal the identity matrix $I$.

### Uniqueness of the Inverse

A natural question is whether a matrix can have more than one inverse. Suppose
$$
AB = BA = I
$$
and there exists another matrix $C$ such that
$$
AC = CA = I.
$$

Then we can write
$$
B = IB = (CA)B = C(AB) = C(I) = C,
$$
which shows that $B = C$.

Therefore, **if an inverse exists, it is unique**.

### Final Remark

Even if there exists a matrix $B$ such that
$$
AB = I,
$$
you might think that you must also verify
$$
BA = I.
$$
Later, you will see that this second check is not necessary: satisfying either condition is enough to guarantee both.

Now, let’s get our hands dirty with some matrices.

## Checking Out the Inverse

The definition alone does not tell you how to *find* the inverse of a matrix $A$.
If you have a candidate matrix $B$, you must check **both** conditions:
$$
AB = I \quad \text{and} \quad BA = I.
$$

Let’s look at an example.

### First Candidate

Consider the matrices
$$
A =
\begin{pmatrix}
1 & 0 \\
6 & 3
\end{pmatrix},
\qquad
B =
\begin{pmatrix}
-3 & 4 \\
0 & -1
\end{pmatrix}
$$

Compute the product:
$$
AB =
\begin{pmatrix}
1 & 0 \\
6 & 3
\end{pmatrix}
\begin{pmatrix}
-3 & 4 \\
0 & -1
\end{pmatrix}
=
\begin{pmatrix}
-3 & 4 \\
-18& 21
\end{pmatrix}
$$

Since
$$
AB \neq I,
$$
you can immediately conclude that
$$
B \neq A^{-1}.
$$

### Second Candidate

Now consider a new candidate:
$$
C =
\begin{pmatrix}
1 & 0 \\
-2 & \tfrac{1}{3}
\end{pmatrix}.
$$

Compute the product:
$$
AC =
\begin{pmatrix}
1 & 0 \\
6 & 3
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
-2 & \tfrac{1}{3}
\end{pmatrix}
=
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
= I.
$$

Although $AC = I$, this is not enough. You must also check the reverse order:
$$
CA =
\begin{pmatrix}
1 & 0 \\
-2 & \tfrac{1}{3}
\end{pmatrix}
\begin{pmatrix}
1 & 0 \\
6 & 3
\end{pmatrix}
=
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
= I.
$$

Since both conditions hold, you can be completely sure that
$$
C = A^{-1}.
$$

### A Quick Elimination Trick

A fast way to rule out whether two matrices are inverses is to compute **just part** of their product:

- If **any diagonal entry** of the product is not equal to $1$, the matrices are **not** inverses.
- If **any off-diagonal entry** of the product is not equal to $0$, the matrices are **not** inverses.

Now that you know how to check inverses, it’s time to look at **why** inverse matrices are so useful.

## Systems of Linear Equations

Systems of equations are very common. Sometimes they are very large, and it can be difficult to deal with so many equations at once. Fortunately, if a matrix is invertible, we can solve such systems very efficiently. In this case, the problem is reduced to calculating the inverse of a matrix, which, as you will see later, is not always a simple task.


### Solving a System Using the Inverse Matrix

Consider a system of linear equations
$$
A x = b,
$$
where:
- $A$ is a square matrix of size $n$,
- $x$ and $b$ are vectors in $\mathbb{R}^n$.

If the matrix $A$ is invertible, we can premultiply both sides of the equation by $A^{-1}$:
$$
A^{-1} A x = A^{-1} b.
$$

Since $A^{-1}A = I$, this simplifies to:
$$
x = A^{-1} b.
$$

Therefore, if $A$ is invertible, the system has a **unique solution**, given explicitly by:
$$
x = A^{-1} b.
$$

A remarkable fact is that the particular value of $b$ does not affect the *existence* or *uniqueness* of the solution—only the invertibility of $A$ matters. This method works for **any square system**, regardless of its size.

### Uniqueness of the Solution (Proof)

We already know that
$$
x = A^{-1} b
$$
is a solution.

Suppose there exists another solution $y$ such that:
$$
A y = b.
$$

Then:
$$
0 = b - b = A x - A y = A(x - y).
$$

So we have:
$$
A(x - y) = 0.
$$

Now multiply both sides by $A^{-1}$:
$$
x - y
= I(x - y)
= (A^{-1}A)(x - y)
= A^{-1}(A(x - y))
= A^{-1}0
= 0.
$$

This implies:
$$
x = y.
$$

Hence, the solution to the system is **unique**.

## Getting More and More Bulky

Time to see an example! A $1 \times 1$ linear system is just an equation involving numbers, such as
$$
5x = 4.
$$
You can easily solve it and obtain
$$
x = \frac{4}{5}.
$$
However, the really interesting systems are those with higher dimensions.

### A $2 \times 2$ Example

Let us use the matrix from the previous section:
$$
A =
\begin{pmatrix}
1 & 0 \\
6 & 3
\end{pmatrix}.
$$

Take, for example,
$$
b =
\begin{pmatrix}
-4 \\
3
\end{pmatrix}.
$$

Since you already know that
$$
A^{-1} =
\begin{pmatrix}
1 & 0 \\
-2 & \tfrac{1}{3}
\end{pmatrix},
$$
the unique solution of the system is
$$
x = A^{-1} b
=
\begin{pmatrix}
1 & 0 \\
-2 & \tfrac{1}{3}
\end{pmatrix}
\begin{pmatrix}
-4 \\
3
\end{pmatrix}
=
\begin{pmatrix}
-10 \\
1
\end{pmatrix}.
$$

But what if
$$
b =
\begin{pmatrix}
5 \\
-2
\end{pmatrix}?
$$

The method works exactly the same:
$$
x = A^{-1} b
=
\begin{pmatrix}
1 & 0 \\
-2 & \tfrac{1}{3}
\end{pmatrix}
\begin{pmatrix}
5 \\
-2
\end{pmatrix}
=
\begin{pmatrix}
9 \\
-\tfrac{2}{3}
\end{pmatrix}.
$$

### A $3 \times 3$ Example

Now you can see a more bulky example. The inverse of the $3\times3$ matrix

$$
A=
\begin{pmatrix}
2 & 1 & -3\\
0 & 1 & -1\\
1 & -2 & 0
\end{pmatrix}
$$

is

$$
A^{-1}
=\frac{1}{2}
\begin{pmatrix}
2 & 1 & 1\\
-6 & -3 & -5\\
-2 & -2 & -2
\end{pmatrix}.
$$

If you take

$$
b=
\begin{pmatrix}
2\\
8\\
-4
\end{pmatrix},
$$

then the solution of the system $Ax=b$ is just

$$
x=A^{-1}b
=\frac{1}{2}
\begin{pmatrix}
2 & 1 & 1\\
-6 & -3 & -5\\
-2 & -2 & -2
\end{pmatrix}
\begin{pmatrix}
2\\
8\\
-4
\end{pmatrix}
=
\begin{pmatrix}
-18\\
-7\\
-15
\end{pmatrix}.
$$

But how do you calculate the inverse of a matrix $A$? This can be really difficult and requires some new tools. You will learn more about it in the next topic!

## Conclusion

Let $A$ be a square matrix of size $n$.

- The matrix $A$ is **invertible** if there exists a square matrix $B$ of size $n$ such that
  $$
  AB = BA = I.
  $$

- When $A$ is invertible, its inverse is **unique** and is denoted by
  $$
  A^{-1}.
  $$

- The system of linear equations
  $$
  Ax = b
  $$
  has a **unique solution** if and only if $A$ is invertible. In this case, the solution is given by
  $$
  x = A^{-1} b.
  $$
