# Calculating matrix determinant

Now that you are familiar with the determinant of matrices of size $2 \times 2$, we can extend this concept to **square matrices of any size**. In this section, you will:

- Understand the definition of the determinant for $n \times n$ matrices
- Derive an explicit formula for the determinant of a $3 \times 3$ matrix
- Identify classes of matrices whose determinants are especially easy to compute

Be prepared to practice extensively—determinants become intuitive through repetition.

## 1. Determinant of a $2 \times 2$ Matrix (Review)

For a matrix

$$
A =
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix},
$$

the determinant is defined as

$$
\det(A) = ad - bc.
$$

This scalar measures how the linear transformation associated with $A$ scales area and whether it preserves or reverses orientation.


## 2. Determinant of a $3 \times 3$ Matrix

Consider a matrix

$$
A =
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix}.
$$

The determinant of $A$ is given by

$$
\det(A)
= a(ei - fh)
- b(di - fg)
+ c(dh - eg).
$$

This formula is known as **cofactor expansion along the first row**.

## 3. Cofactors and Minors

### Minor

The **minor** $M_{ij}$ of an entry $a_{ij}$ is the determinant of the matrix obtained by deleting row $i$ and column $j$.

### Cofactor

The **cofactor** $C_{ij}$ is defined as

$$
C_{ij} = (-1)^{i+j} M_{ij}.
$$

Using cofactors, the determinant of an $n \times n$ matrix can be written as

$$
\det(A) = \sum_{j=1}^{n} a_{ij} C_{ij},
$$

for any fixed row $i$ (or similarly, any column).

## 4. Determinant of an $n \times n$ Matrix

For a general square matrix $A \in \mathbb{R}^{n \times n}$, the determinant is defined recursively using **cofactor expansion**:

$$
\det(A) = \sum_{j=1}^{n} a_{ij} (-1)^{i+j} \det(A_{ij}),
$$

where $A_{ij}$ is the $(n-1) \times (n-1)$ submatrix obtained by removing row $i$ and column $j$.

## 5. Matrices with Easy Determinants

Some matrices have determinants that can be computed almost immediately.

### Identity Matrix

$$
I_n =
\begin{pmatrix}
1 & 0 & \cdots & 0 \\
0 & 1 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & 1
\end{pmatrix}
\quad \Rightarrow \quad
\det(I_n) = 1.
$$

### Diagonal Matrix

$$
D =
\begin{pmatrix}
d_1 & 0 & \cdots & 0 \\
0 & d_2 & \cdots & 0 \\
\vdots & \vdots & \ddots & \vdots \\
0 & 0 & \cdots & d_n
\end{pmatrix}
\quad \Rightarrow \quad
\det(D) = \prod_{k=1}^{n} d_k.
$$

### Triangular Matrices (Upper or Lower)

If $A$ is triangular, then

$$
\det(A) = \text{product of the diagonal entries}.
$$

### Matrices with a Zero Row or Column

If a matrix has an entire row or column of zeros, then

$$
\det(A) = 0.
$$


## 6. Key Properties of Determinants

Let $A, B \in \mathbb{R}^{n \times n}$ and $\lambda \in \mathbb{R}$.

- $\det(AB) = \det(A)\det(B)$
- $\det(A^T) = \det(A)$
- $\det(\lambda A) = \lambda^n \det(A)$
- $A \text{ is invertible } \iff \det(A) \neq 0$

## 7. Geometric Interpretation

The determinant of a matrix represents the **scaling factor of volume** under the associated linear transformation:

- In $\mathbb{R}^2$: area scaling
- In $\mathbb{R}^3$: volume scaling

The sign of the determinant indicates whether orientation is preserved ($\det > 0$) or reversed ($\det < 0$).


## 8. Summary

- Determinants generalize from $2 \times 2$ to $n \times n$ matrices via cofactors
- The $3 \times 3$ determinant has a concrete closed-form expression
- Certain matrix classes allow instant determinant computation
- Determinants connect algebra, geometry, and invertibility

Practice computing determinants by expansion and by using matrix structure—fluency here pays dividends throughout linear algebra.


## The Protagonists: Minors and Cofactors

In order to define the determinant of a larger matrix, we begin by extracting **smaller sub-matrices** from it.

Let $A$ be a square matrix of size $n$, and assume that the determinant has already been defined for matrices of size $n-1$.

### Minors

The **$(i,j)$-minor** of $A$, denoted by $m_{ij}$, is defined as the determinant of the $(n-1)\times(n-1)$ matrix obtained by deleting the $i$-th row and the $j$-th column of $A$.

### Example: A $3 \times 3$ Matrix

Consider the matrix

$$
A =
\begin{pmatrix}
3 & 0 & -1 \\
4 & 1 & 0 \\
0 & -5 & 1
\end{pmatrix}.
$$

#### The $(1,1)$-minor

The minor $m_{11}$ is obtained by deleting the **first row** and **first column** of $A$.

This leaves the matrix

$$
\begin{pmatrix}
1 & 0 \\
-5 & 1
\end{pmatrix},
$$

and therefore

$$
m_{11}
=
\det
\begin{pmatrix}
1 & 0 \\
-5 & 1
\end{pmatrix}.
$$


#### The $(2,3)$-minor

The minor $m_{23}$ is obtained by deleting the **second row** and the **third column** of $A$.

This leaves the matrix

$$
\begin{pmatrix}
3 & 0 \\
0 & -5
\end{pmatrix},
$$

and therefore

$$
m_{23}
=
\det
\begin{pmatrix}
3 & 0 \\
0 & -5
\end{pmatrix}.
$$

### Cofactors and Alternation

Minors alone are not sufficient to define the determinant. The determinant exhibits an important property known as **alternation**.
To capture this behavior, each minor is multiplied by a sign factor.

The **$(i,j)$-cofactor** of $A$ is defined as

$$
c_{ij} = (-1)^{i+j} m_{ij}.
$$

#### Sign Behavior

- If $i + j$ is **even**, then $(-1)^{i+j} = 1$
- If $i + j$ is **odd**, then $(-1)^{i+j} = -1$

#### Applying This to the Example

For the previous minors:

- Since $1 + 1 = 2$ is even, we have
  $$
  c_{11} = m_{11}.
  $$

- Since $2 + 3 = 5$ is odd, we have
  $$
  c_{23} = -m_{23}.
  $$

### Checkerboard Pattern of Signs

The signs $(-1)^{i+j}$ follow a **checkerboard (chessboard) pattern**:

$$
\begin{pmatrix}
+ & - & + & \cdots \\
- & + & - & \cdots \\
+ & - & + & \cdots \\
\vdots & \vdots & \vdots & \ddots
\end{pmatrix}
$$

- Dark squares correspond to a factor of $+1$
- White squares correspond to a factor of $-1$

This alternating structure is fundamental to the definition of cofactors and, ultimately, to the determinant itself.


## The Determinant of Any Square Matrix

Now that you are familiar with **cofactors**, it is time to take the next step and combine everything.

### Cofactor Expansion Along a Row

Let $A$ be a square matrix of size $n$.
The **cofactor expansion along the $i$-th row of $A$** is defined as the linear combination of the cofactors of the entries in that row, using the entries of the row as coefficients.

Since $a_{i1}, a_{i2}, \ldots, a_{in}$ are precisely the entries of the $i$-th row of $A$, the expansion is

$$
a_{i1} c_{i1} + a_{i2} c_{i2} + \cdots + a_{in} c_{in}.
$$


### Example

Consider again the matrix

$$
A =
\begin{pmatrix}
3 & 0 & -1 \\
4 & 1 & 0 \\
0 & -5 & 1
\end{pmatrix}.
$$

#### Expansion Along the First Row

The cofactor expansion along the **first row** of $A$ is

$$
a_{11} c_{11} + a_{12} c_{12} + a_{13} c_{13}.
$$

Substituting the entries of the first row, we obtain

$$
3c_{11} + 0c_{12} - c_{13}.
$$

Using the definition of cofactors, this becomes

$$
3m_{11} - m_{13}.
$$

Now compute the minors explicitly:

$$
m_{11} =
\det
\begin{pmatrix}
1 & 0 \\
-5 & 1
\end{pmatrix},
\qquad
m_{13} =
\det
\begin{pmatrix}
4 & 1 \\
0 & -5
\end{pmatrix}.
$$

Hence,

$$
\det(A)
=
3
\begin{vmatrix}
1 & 0 \\
-5 & 1
\end{vmatrix}
-
\begin{vmatrix}
4 & 1 \\
0 & -5
\end{vmatrix}
= 23.
$$


### Expansion Along Other Rows and Columns

If you compute the cofactor expansion along the **second row** or the **third row**, you will find that the result is again $23$.

Similarly, you can define the **cofactor expansion along the $j$-th column** of $A$.
Since $a_{1j}, a_{2j}, \ldots, a_{nj}$ are the entries of the $j$-th column, the expansion is

$$
a_{1j} c_{1j} + a_{2j} c_{2j} + \cdots + a_{nj} c_{nj}.
$$

If you compute the expansion along each column of the example matrix, you will observe that:

- All column expansions give the same value
- This value is the same as the value obtained from any row expansion

This is not a coincidence—it holds for **every square matrix**.


### Definition of the Determinant

If $A$ is a square matrix of size $n$, its **determinant** is defined as the value of the cofactor expansion along **any row or any column**.

This definition is recursive:
- Once the determinant is defined for matrices of size $n-1$,
- It can be defined for matrices of size $n$ using cofactors.


### Determinant of a $3 \times 3$ Matrix

Using cofactor expansion along the first row, the determinant of

$$
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix}
$$

is

$$
\begin{vmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{vmatrix}
=
a
\begin{vmatrix}
e & f \\
h & i
\end{vmatrix}
-
b
\begin{vmatrix}
d & f \\
g & i
\end{vmatrix}
+
c
\begin{vmatrix}
d & e \\
g & h
\end{vmatrix}.
$$

Evaluating the $2 \times 2$ determinants gives

$$
\det(A)
=
a(ei - fh)
-
b(di - fg)
+
c(dh - eg).
$$

Expanding further, we obtain

$$
\det(A)
=
aei + bfg + cdh
-
ceg - bdi - afh.
$$


### Diagonal Interpretation (Mnemonic)

Although the formula looks difficult to memorize, there is a helpful pattern:

- The **positive terms** are the products of the three forward diagonals
- The **negative terms** are the products of the three backward diagonals

This diagonal rule provides a convenient mnemonic for computing determinants of $3 \times 3$ matrices.


### Summary

- The determinant is defined via cofactor expansion
- Any row or any column can be used
- All such expansions give the same value
- This recursive definition allows computation of determinants for matrices of any size

With this framework, you can now compute the determinant of any square matrix.


# Putting Theory to Work

Let us look at some concrete examples. A natural starting point is the class of **upper triangular matrices**.
These matrices have zeros below their main diagonal.

For example:

$$
\begin{pmatrix}
1 & 2 \\
0 & 1
\end{pmatrix},
\qquad
\begin{pmatrix}
1 & 2 & 3 \\
0 & 1 & 4 \\
0 & 0 & 1
\end{pmatrix},
\qquad
\begin{pmatrix}
1 & 2 & 3 & 4 \\
0 & 1 & 5 & 6 \\
0 & 0 & 1 & 7 \\
0 & 0 & 0 & 1
\end{pmatrix}.
$$

Their determinant is **extremely easy to compute**: it is simply the product of the entries on the main diagonal.

In fact, the same result holds for **lower triangular matrices** (those with zeros *above* the diagonal), regardless of their size.

## Statement

If $A$ is a triangular matrix of size $n \times n$, then

$$
\det(A) = a_{11} a_{22} \cdots a_{nn}.
$$

## Proof

### Base Case: $n = 2$

For a $2 \times 2$ upper triangular matrix,

$$
\begin{vmatrix}
a_{11} & a_{12} \\
0 & a_{22}
\end{vmatrix}
=
a_{11} a_{22} - 0 \cdot a_{12}
=
a_{11} a_{22}.
$$

Thus, the statement is true for $n = 2$.

### Key Idea

The crucial observation is that we can always perform a **cofactor expansion along the first column**.

In an upper triangular matrix, the first column contains only one nonzero entry, namely $a_{11}$.
Therefore, the determinant reduces to that entry multiplied by its cofactor.

### Induction Step

Assume the result holds for matrices of size $n-1$.
Consider an upper triangular matrix of size $n$:

$$
\begin{vmatrix}
a_{11} & a_{12} & a_{13} & \cdots & a_{1n} \\
0      & a_{22} & a_{23} & \cdots & a_{2n} \\
0      & 0      & a_{33} & \cdots & a_{3n} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0      & 0      & 0      & \cdots & a_{nn}
\end{vmatrix}.
$$

Expanding along the first column gives

$$
\det(A)
=
a_{11}
\begin{vmatrix}
a_{22} & a_{23} & \cdots & a_{2n} \\
0      & a_{33} & \cdots & a_{3n} \\
\vdots & \vdots & \ddots & \vdots \\
0      & 0      & \cdots & a_{nn}
\end{vmatrix}.
$$

The remaining matrix is again upper triangular, now of size $(n-1) \times (n-1)$.
By the induction hypothesis, its determinant is

$$
a_{22} a_{33} \cdots a_{nn}.
$$

Hence,

$$
\det(A) = a_{11} a_{22} a_{33} \cdots a_{nn}.
$$


## Conclusion

By mathematical induction, the determinant of any triangular matrix—upper or lower—is equal to the product of its diagonal entries.

This result makes triangular matrices among the easiest matrices for determinant computations and plays a key role in matrix factorizations such as LU and QR.


## A Larger Example: Strategic Cofactor Expansion

Here is a useful shortcut to keep in mind:

> If a matrix is **upper triangular** and one of its diagonal entries is $0$, then its determinant is $0$.

Now let us work through a larger example to see cofactor expansion in action.

### The Matrix

Consider the $5 \times 5$ matrix

$$
A =
\begin{pmatrix}
2 & 0 & -3 & 2 & 3 \\
-3 & 0 & 4 & -5 & 1 \\
1 & 0 & 5 & 1 & 0 \\
1 & 0 & -3 & 0 & 0 \\
4 & -3 & 4 & 2 & -2
\end{pmatrix}.
$$

Recall that the determinant can be computed by expanding along **any row or column**.
The optimal strategy is to choose a row or column containing many zeros.


### Step 1: Expand Along the Second Column

The second column contains four zeros, so it is an excellent choice:

$$
\det(A)
=
a_{12}c_{12}
+
a_{22}c_{22}
+
a_{32}c_{32}
+
a_{42}c_{42}
+
a_{52}c_{52}.
$$

Since $a_{12}=a_{22}=a_{32}=a_{42}=0$ and $a_{52}=-3$, this reduces to

$$
\det(A) = -3\,c_{52}.
$$

Using the definition of cofactors,

$$
c_{52} = (-1)^{5+2} m_{52} = -m_{52},
$$

so

$$
\det(A) = 3m_{52}.
$$

The minor $m_{52}$ is the determinant of the $4 \times 4$ matrix obtained by deleting row $5$ and column $2$:

$$
m_{52}
=
\begin{vmatrix}
2 & -3 & 2 & 3 \\
-3 & 4 & -5 & 1 \\
1 & 5 & 1 & 0 \\
1 & -3 & 0 & 0
\end{vmatrix}.
$$

### Step 2: Define the $4 \times 4$ Matrix

Let

$$
B =
\begin{pmatrix}
2 & -3 & 2 & 3 \\
-3 & 4 & -5 & 1 \\
1 & 5 & 1 & 0 \\
1 & -3 & 0 & 0
\end{pmatrix}.
$$

Then

$$
\det(A) = 3\det(B).
$$

### Step 3: Expand $\det(B)$ Along the Fourth Row

The fourth row of $B$ has two zeros, so we expand along it:

$$
\det(B)
=
b_{41}c_{41}
+
b_{42}c_{42}
+
b_{43}c_{43}
+
b_{44}c_{44}.
$$

Since $b_{43}=b_{44}=0$, this becomes

$$
\det(B) = b_{41}c_{41} + b_{42}c_{42}.
$$

Substituting values,

$$
\det(B)
=
1\cdot c_{41}
-
3\cdot c_{42}.
$$

Using cofactors,

$$
c_{41} = -m_{41}, \qquad c_{42} = m_{42},
$$

so

$$
\det(B) = -m_{41} - 3m_{42}.
$$


### Step 4: Compute the $3 \times 3$ Minors

#### First Minor

$$
m_{41}
=
\begin{vmatrix}
-3 & 4 & -5 \\
1 & 5 & 1 \\
1 & -3 & 0
\end{vmatrix}.
$$

Expanding along the third row:

$$
m_{41}
=
5
\begin{vmatrix}
4 & -5 \\
5 & 1
\end{vmatrix}
-
\begin{vmatrix}
-3 & -5 \\
1 & 1
\end{vmatrix}.
$$

Evaluating,

$$
m_{41}
=
5(4+25) - (-3-5) = 100.
$$


#### Second Minor

$$
m_{42}
=
\begin{vmatrix}
2 & 2 & 3 \\
-3 & -5 & 1 \\
1 & 1 & 0
\end{vmatrix}.
$$

Expanding along the third row:

$$
m_{42}
=
\begin{vmatrix}
2 & 3 \\
-5 & 1
\end{vmatrix}
-
\begin{vmatrix}
2 & 3 \\
-3 & 1
\end{vmatrix}.
$$

Thus,

$$
m_{42}
=
(2+15) - (2+9) = 6.
$$


### Step 5: Finish the Computation

Now substitute the values:

$$
\det(B) = -100 - 3(6) = -118.
$$

Finally,

$$
\det(A) = 3\det(B) = 3(-118) = -354.
$$


### Final Result

$$
\boxed{\det(A) = -354}
$$

This example illustrates how choosing rows or columns with many zeros can dramatically simplify determinant computations, even for large matrices.


## Conclusion

Let $A$ be a square matrix of size $n$.


### Minors

The $(i,j)$-minor of $A$, denoted by $m_{ij}$, is the determinant of the $(n-1)\times(n-1)$ matrix obtained by deleting the $i$-th row and the $j$-th column of $A$.


### Cofactors

The $(i,j)$-cofactor of $A$ is defined as $c_{ij} = (-1)^{i+j} m_{ij}.$

### Cofactor Expansions

The **cofactor expansion along the $i$-th row** of $A$ is $a_{i1} c_{i1} + a_{i2} c_{i2} + \cdots + a_{in} c_{in}.$

The **cofactor expansion along the $j$-th column** of $A$ is $a_{1j} c_{1j} + a_{2j} c_{2j} + \cdots + a_{nj} c_{nj}.$

### Determinant

The **determinant** of $A$ is defined as the value of its cofactor expansion along **any row or any column**.

### Determinant of a $3 \times 3$ Matrix

For a matrix

$$
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix},
$$

the determinant is

$$
\begin{vmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{vmatrix}
=
aei + bfg + cdh - ceg - bdi - afh.
$$

### Triangular Matrices

The determinant of a triangular matrix is the **product of its diagonal entries**: $\det(A) = a_{11} a_{22} \cdots a_{nn}.$
