# Calculating matrix determinant

Now that you are familiar with the determinant of matrices of size $2 \times 2$, we can extend this concept to **square matrices of any size**. In this section, you will:

- Understand the definition of the determinant for $n \times n$ matrices
- Derive an explicit formula for the determinant of a $3 \times 3$ matrix
- Identify classes of matrices whose determinants are especially easy to compute

Be prepared to practice extensivelyâ€”determinants become intuitive through repetition.

## The building blocks

You already know that the determinant of a $2\times2$ matrix is

$$
\left|\begin{matrix}
a & b\\
c & d
\end{matrix}\right| = ad - bc.
$$

Maybe it is easy to grasp, but not so easy to remember.

A good tip is to think about it in terms of the diagonals of the matrix. The determinant is just the product of its forward diagonal minus the product of its backward diagonal:

![Matrix Determinant_2x2](img/determinant_2x2_ad_minus_bc.png)

But what about the determinant of larger matrices? The focus here will be to define the determinant of $3\times3$ matrices through the determinant of several $2\times2$ matrices. You will use the same strategy to define the determinant of any $n\times n$ matrix in terms of the determinant of several $(n-1)\times(n-1)$ matrices.

Before you go any further on this road, you need to know a bit of terminology.

## The protagonists: minors and cofactors

In order to define the determinant of a larger matrix, you are going to start extracting smaller sub-matrices from it.

Let us take a square matrix $A$ of size $n$ and assume that you have already defined the determinant for matrices of size $n-1$.
The $(i,j)$-minor of $A$, denoted by $m_{ij}$, is the determinant of the $(n-1)\times(n-1)$ matrix that you obtain by deleting the $i$-th row and the $j$-th column of $A$.

Let us illustrate this with a matrix of size $3$:

$$
A=
\begin{pmatrix}
3 & 0 & -1\\
4 & 1 & 0\\
0 & -5 & 1
\end{pmatrix}.
$$

For example, the minor $m_{11}$ of $A$ is the determinant of the matrix that results from eliminating the first row and the first column of $A$:

![Matrix Determinant_2x2_Ex_1](img/determinant_expansion_example.png)

In turn, the minor $m_{23}$ of $A$ is the determinant of the matrix that results from eliminating the second row and the third column of $A$.

![Matrix Determinant_2x2_Ex_02](img/determinant_expansion_example_2.png)

Let us see the last ingredient you need. The determinant has a certain *alternation* behavior. This is a very particular quality and is manifested in the fact that you need to slightly alter the minors by multiplying them by a simple factor. In some cases, this factor is simply $1$ and in the rest it is $-1$. The $(i,j)$-cofactor of $A$ is defined as:

$$
c_{ij} = (-1)^{\,i+j} m_{ij}
$$

This means that the factor of the minor $m_{ij}$ is $1$ when $i+j$ is even, and $-1$ when it is odd. In the previous example, the factor of $m_{11}$ is therefore $1$, and the factor of $m_{23}$ is $-1$. Thus, $c_{11}$ = $m_{11}$ and $c_{23}$ = $-m_{23}$.

You can visualize the behavior of this factor as if it were a chessboard. White entries have a factor of $-1$, while dark entries have a factor of $1$.

![Cofactor_sign_pattern](img/cofactor_sign_pattern.png)


## The determinant of any square matrix

Now that you are familiar with cofactors, it is time to take the next step and combine them all. We define the **cofactor expansion along the $i$-th row of $A$** as the linear combination of the cofactors of the entries of the $i$-th row, using the entries of that row as coefficients (notice that $a_{i1},a_{i2},\ldots,a_{in}$ are all the entries of the $i$-th row of $A$):

$$
a_{i1}c_{i1}+a_{i2}c_{i2}+\cdots+a_{in}c_{in}.
$$

Let us continue with the matrix from the previous section:

$$
A=
\begin{pmatrix}
3 & 0 & -1\\
4 & 1 & 0\\
0 & -5 & 1
\end{pmatrix}.
$$

For example, the cofactor expansion along the first row of $A$ is

$$
a_{11}c_{11}+a_{12}c_{12}+a_{13}c_{13}
=3c_{11}+0c_{12}-c_{13}
=3m_{11}-m_{13}.
$$

This gives

$$
3\begin{vmatrix}
1 & 0\\
-5 & 1
\end{vmatrix}
-
\begin{vmatrix}
4 & 1\\
0 & -5
\end{vmatrix}
=23.
$$

If you calculate the expansion along either of the other two rows, you will notice that the result is the same. You can also define the **cofactor expansion along the $j$-th column** in a similar way (here $a_{1j},a_{2j},\ldots,a_{nj}$ are all the entries of the $j$-th column of $A$):

$$
a_{1j}c_{1j}+a_{2j}c_{2j}+\cdots+a_{nj}c_{nj}.
$$

Try to calculate the expansion along each column of the example matrix. Not only are they all equal, but their value coincides with the common value of the row expansions. This is not a coincidence and is, in fact, true for any matrix. For this reason, we define the **determinant** of a matrix as any of these expansions.

If $A$ is a square matrix of size $n$, its determinant is defined as the value of its expansion along any row or any column.

The important point of this definition is that, once the determinant has been defined for matrices of size $n-1$, it can be extended to matrices of size $n$, and then recursively to larger sizes.

In particular, the determinant of any $3\times3$ matrix can be computed by expanding along any row or column, and the result will always be the same. For example, expanding along the first row gives

$$
\begin{vmatrix}
a & b & c\\
d & e & f\\
g & h & i
\end{vmatrix}
=
a\begin{vmatrix}
e & f\\
h & i
\end{vmatrix}
-
b\begin{vmatrix}
d & f\\
g & i
\end{vmatrix}
+
c\begin{vmatrix}
d & e\\
g & h
\end{vmatrix}.
$$

This can be written explicitly as

$$
a(ei-fh)-b(di-fg)+c(dh-eg)
=aei+bfg+cdh-ceg-bdi-afh.
$$

This is a difficult formula to memorize, but there is a useful trick similar to the one used for $2\times2$ matrices: the positive terms correspond to the products of the three forward diagonals of the matrix, while the negative terms correspond to the products of the three backward diagonals.

![determinant_3x3_sarrus_rule](img/determinant_3x3_sarrus_rule.png)

## Putting theory to work

Let us see some examples. What is better than starting with the simplest matrices, the upper triangular matrices? They only have zeros below their diagonal. For instance:

$$
\begin{pmatrix}
1 & 2\\
0 & 1
\end{pmatrix},
\qquad
\begin{pmatrix}
1 & 2 & 3\\
0 & 1 & 4\\
0 & 0 & 1
\end{pmatrix},
\qquad
\begin{pmatrix}
1 & 2 & 3 & 4\\
0 & 1 & 5 & 6\\
0 & 0 & 1 & 7\\
0 & 0 & 0 & 1
\end{pmatrix}
$$

Their determinant is extremely easy to calculate: it is simply the product of the entries on the main diagonal.

Actually, the same result holds for lower triangular matrices (those that only have zeros above their diagonal), no matter how large they are.

### Proof

For the case $n=2$, the result is obvious:
$$
\begin{vmatrix}
a_{11} & a_{12}\\
0 & a_{22}
\end{vmatrix}
= a_{11}a_{22}-0\cdot a_{12}
= a_{11}a_{22}.
$$

The trick of the proof is that we can always expand the determinant by cofactors with respect to the first column. Since this column has only one entry different from $0$, the determinant is simply that first entry multiplied by its cofactor.

In general, for any $n$, having proved the result for $n-1$, we can easily extend it to $n$. This technique is called *mathematical induction*. Let us see it in action. If the result holds for $n-1$, then for $n$ we have

$$
\begin{vmatrix}
a_{11} & a_{12} & a_{13} & \cdots & a_{1n}\\
0 & a_{22} & a_{23} & \cdots & a_{2n}\\
0 & 0 & a_{33} & \cdots & a_{3n}\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
0 & 0 & 0 & \cdots & a_{nn}
\end{vmatrix}
=
a_{11}
\begin{vmatrix}
a_{22} & a_{23} & \cdots & a_{2n}\\
0 & a_{33} & \cdots & a_{3n}\\
\vdots & \vdots & \ddots & \vdots\\
0 & 0 & \cdots & a_{nn}
\end{vmatrix}.
$$

By the induction hypothesis, the determinant of the $(n-1)\times(n-1)$ matrix is
$$
a_{22}a_{33}\cdots a_{nn}.
$$

Therefore,
$$
\begin{vmatrix}
a_{11} & a_{12} & a_{13} & \cdots & a_{1n}\\
0 & a_{22} & a_{23} & \cdots & a_{2n}\\
0 & 0 & a_{33} & \cdots & a_{3n}\\
\vdots & \vdots & \vdots & \ddots & \vdots\\
0 & 0 & 0 & \cdots & a_{nn}
\end{vmatrix}
=
a_{11}a_{22}a_{33}\cdots a_{nn}.
$$

This completes the proof.

Here is a little shortcut: if the matrix is upper triangular and one of its diagonal entries is $0$, then its determinant is $0$.

Let us end with a slightly larger matrix, say $5\times5$:

$$
A=
\begin{pmatrix}
2 & 0 & -3 & 2 & 3\\
-3 & 0 & 4 & -5 & 1\\
1 & 0 & 5 & 1 & 0\\
1 & 0 & -3 & 0 & 0\\
4 & -3 & 4 & 2 & -2
\end{pmatrix}.
$$

Remember that to calculate the determinant, you can expand along any row or column you want. The best strategy is to choose one that has a lot of $0$'s, as this will reduce the number of cofactors you have to calculate. If you look closely, the second column has a lot of zeros, so it is a good candidate:

$$
\det(A)
= a_{12}c_{12}+a_{22}c_{22}+a_{32}c_{32}+a_{42}c_{42}+a_{52}c_{52}
=0+0+0+0-3c_{52}
$$

$$
=-3(-1)^{5+2}m_{52}
=-3(-1)m_{52}
=3m_{52}
=3
\begin{vmatrix}
2 & -3 & 2 & 3\\
-3 & 4 & -5 & 1\\
1 & 5 & 1 & 0\\
1 & -3 & 0 & 0
\end{vmatrix}.
$$

Great, now there is only one determinant to solve for a matrix of size $4$. If you define

$$
B=
\begin{pmatrix}
2 & -3 & 2 & 3\\
-3 & 4 & -5 & 1\\
1 & 5 & 1 & 0\\
1 & -3 & 0 & 0
\end{pmatrix},
$$

then

$$
\det(A)=3\det(B).
$$

But now, which is the best row or column to expand along? Undoubtedly the fourth row, since it has $2$ zeros.

$$
\det(B)
=b_{41}c_{41}+b_{42}c_{42}+b_{43}c_{43}+b_{44}c_{44}
=b_{41}c_{41}+b_{42}c_{42}+0c_{43}+0c_{44}
$$

$$
=c_{41}-3c_{42}
=-m_{41}-3m_{42}
=-\begin{vmatrix}
-3 & 2 & 3\\
4 & -5 & 1\\
5 & 1 & 0
\end{vmatrix}
-3\begin{vmatrix}
2 & 2 & 3\\
-3 & -5 & 1\\
1 & 1 & 0
\end{vmatrix}.
$$

Now there are $2$ determinants of size $3$. You can either use the explicit formula or expand across the third line (because it has a $0$).

For the first one:

$$
\begin{vmatrix}
-3 & 2 & 3\\
4 & -5 & 1\\
5 & 1 & 0
\end{vmatrix}
=
5c_{31}+c_{32}+0c_{33}
=
5\begin{vmatrix}
2 & 3\\
-5 & 1
\end{vmatrix}
-
\begin{vmatrix}
-3 & 3\\
4 & 1
\end{vmatrix}
$$

$$
=5(2+15)-(-3-12)=100.
$$

And for the second one:

$$
\begin{vmatrix}
2 & 2 & 3\\
-3 & -5 & 1\\
1 & 1 & 0
\end{vmatrix}
=
\begin{vmatrix}
2 & 3\\
-5 & 1
\end{vmatrix}
-
\begin{vmatrix}
2 & 3\\
-3 & 1
\end{vmatrix}
=(2+15)-(2+9)=6.
$$

Thus,

$$
\det(B)=-100-3(6)=-118.
$$

Finally,

$$
\det(A)=3\det(B)=3(-118)=-354.
$$

## Conclusion

Let us take a square matrix $A$ of size $n$.

- The $(i,j)$-minor of $A$, denoted as $m_{ij}$, is the determinant of the $(n-1)\times(n-1)$ matrix that you obtain by deleting the $i$-th row and the $j$-th column of $A$.

- The $(i,j)$-cofactor of $A$ is
$$
c_{ij}=(-1)^{i+j}m_{ij}.
$$

- The cofactor expansion along the $i$-th row of $A$ is
$$
a_{i1}c_{i1}+a_{i2}c_{i2}+\cdots+a_{in}c_{in}
$$

- The cofactor expansion along the $j$-th column of $A$ is
$$
a_{1j}c_{1j}+a_{2j}c_{2j}+\cdots+a_{nj}c_{nj}
$$

- The determinant of $A$ is the value of its expansion along any row or any column.

- The determinant of a matrix of size $3$ is
$$
\begin{vmatrix}
a & b & c\\
d & e & f\\
g & h & i
\end{vmatrix}
=aei+bfg+cdh-ceg-bdi-afh.
$$

- The determinant of a triangular matrix is the product of its diagonal entries.
