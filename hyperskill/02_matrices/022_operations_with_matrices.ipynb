# Operations with matrices

We often need to work with matrices. For example, you might look at a monthly sales table and want to determine how many items were sold over the last five years. As another example, you may multiply a table of office supplies by the number of offices to calculate how much equipment the entire company needs.

To perform tasks like these, we must be able to carry out different operations on matrices.

In this topic, we will learn how to:
- multiply a matrix by a scalar,
- add and subtract matrices,
- and understand what a transposed matrix is.

## Scalar Multiplication

We can multiply a matrix by any real number. Since this number *scales* the matrix, it is called a **scalar**, and the operation is known as **scalar multiplication**.

To multiply a matrix by a scalar, simply multiply **each entry** of the matrix by that scalar.

### Example

$$
-2 \cdot
\begin{pmatrix}
-2 & 0 \\
-3 & 1 \\
-1 & 0
\end{pmatrix}
=
\begin{pmatrix}
-2\cdot(-2) & -2\cdot 0 \\
-2\cdot(-3) & -2\cdot 1 \\
-2\cdot(-1) & -2\cdot 0
\end{pmatrix}
=
\begin{pmatrix}
4 & 0 \\
6 & -2 \\
2 & 0
\end{pmatrix}.
$$

### Why Scalar Multiplication Is Useful

Scalar multiplication is frequently used to **simplify matrices**, especially when solving systems of linear equations. For example, it allows you to scale rows or entire matrices to obtain cleaner coefficients or to match terms during elimination procedures.

## Matrix Sum

To add two matrices, they must have the **same dimensions**. This means they must have the same number of rows and the same number of columns.

When this condition is satisfied, matrix addition is performed **entry by entry**: each element of the first matrix is added to the element in the corresponding position of the second matrix.

### Example

Consider the matrices
$$
A =
\begin{pmatrix}
3 & 0 & -3 \\
2 & -1 & 8
\end{pmatrix},
\qquad
B =
\begin{pmatrix}
1 & 2 & 3 \\
3 & 1 & -7
\end{pmatrix}.
$$

Both matrices are of size $2 \times 3$, so their sum is defined.

Compute the sum:
$$
A + B =
\begin{pmatrix}
3+1 & 0+2 & -3+3 \\
2+3 & -1+1 & 8-7
\end{pmatrix}
=
\begin{pmatrix}
4 & 2 & 0 \\
5 & 0 & 1
\end{pmatrix}.
$$

### Key Points

- Matrix addition is only defined for matrices of the **same size**.
- Addition is performed by adding corresponding entries.
- Matrix addition behaves similarly to vector addition and is used extensively in linear algebra computations.

## Matrix Difference

To subtract two matrices, they must have the **same dimensions**, just as in the case of matrix addition.

Matrix subtraction is performed **entry by entry**, subtracting the elements of the second matrix from the corresponding elements of the first matrix.

### Example

Consider the matrices
$$
A =
\begin{pmatrix}
1 & 2 \\
2 & 3
\end{pmatrix},
\qquad
B =
\begin{pmatrix}
-2 & 2 \\
0 & 5
\end{pmatrix}.
$$

Their difference is
$$
A - B =
\begin{pmatrix}
1 - (-2) & 2 - 2 \\
2 - 0 & 3 - 5
\end{pmatrix}
=
\begin{pmatrix}
3 & 0 \\
2 & -2
\end{pmatrix}.
$$

### Relation to Addition

As you may have noticed, matrix subtraction is closely related to matrix addition. Subtracting a matrix is equivalent to adding its **negative**.

If we define
$$
- A = (-1)\cdot A,
$$
then for any two matrices $A$ and $B$ of the same size,
$$
A - B = A + (-B).
$$

This mirrors the familiar rules of arithmetic and simplifies many matrix computations.

## Transposed Matrix

Starting with a matrix $A$, you can obtain a new matrix $A^T$ by **reflecting the elements of $A$ across its main diagonal**. This operation is called **transposition**.

When a matrix is transposed:
- rows become columns,
- columns become rows.

The superscript $T$ stands for *transpose*.
Formally, the relationship between the elements of $A$ and $A^T$ is
$$
(A^T)_{i,j} = A_{j,i}.
$$

### Example

Consider the matrix
$$
F =
\begin{pmatrix}
1 & 2 \\
1 & 2 \\
1 & 2
\end{pmatrix}.
$$

This is a $3\times 2$ matrix.

### Step 1: Identify the Main Diagonal

The **main diagonal** starts at the top-left corner and proceeds downwards to the right.
Elements on the main diagonal **do not change** their values during transposition.


### Step 2: Swap Rows and Columns

- The first row of $F$ becomes the first column of $F^T$.
- The second row of $F$ becomes the second column of $F^T$.
- The third row of $F$ becomes the third column of $F^T$.

Carrying this out gives:
$$
F^T =
\begin{pmatrix}
1 & 1 & 1 \\
2 & 2 & 2
\end{pmatrix}.
$$

So explicitly,
$$
F^T =
\begin{pmatrix}
1 & 2 \\
1 & 2 \\
1 & 2
\end{pmatrix}^T
=
\begin{pmatrix}
1 & 1 & 1 \\
2 & 2 & 2
\end{pmatrix}.
$$

### Important Property

If you transpose a matrix **twice**, you return to the original matrix:
$$
(A^T)^T = A.
$$

### Key Takeaways

- Transposition swaps rows and columns.
- The main diagonal remains unchanged.
- A $m\times n$ matrix becomes an $n\times m$ matrix after transposition.
- Transposition is a fundamental operation used in matrix multiplication, dot products, and many proofs in linear algebra.
