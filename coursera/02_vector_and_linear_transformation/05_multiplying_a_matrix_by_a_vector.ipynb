# Multiplying a Matrix by a Vector

*Course notes for **Math for Machine Learning**, C1 · W3 · L1 · V08 — "Multiplying a Matrix by a Vector" (DeepLearning.AI).*

This lecture ties the whole lesson together: a **matrix-times-vector** product is just a stack of **dot products**, and it is precisely how a **system of linear equations** looks in compact form. We cover:

- A single linear equation as one **dot product**.
- A system of equations as a **matrix $\times$ vector** product (stacked dot products).
- The **dimension rule** for when the product is defined, and **rectangular** matrices.

```python
import numpy as np

np.set_printoptions(precision=3, suppress=True)
```

## 1. One equation is a dot product

Recall the dot product is the sum of products of corresponding entries. An equation with unknowns such as

$$ 2a + 4b + c = 28 $$

is the dot product of a **row vector of coefficients** $(2, 4, 1)$ with a **column vector of unknowns** $(a, b, c)$:

$$ \begin{pmatrix} 2 & 4 & 1 \end{pmatrix} \begin{pmatrix} a \\ b \\ c \end{pmatrix} = 28. $$

(Vectors are allowed to contain **variables**.) In the fruit story: the coefficients are *how many of each fruit* you bought, the unknowns are *the price of each fruit*, and $28$ is the total.

```python
coeffs = np.array([2, 4, 1])
prices = np.array([3, 5, 2])   # a, b, c
print('coeffs . prices =', coeffs @ prices, '  (= 28)')
```

## 2. A system of equations as a matrix $\times$ vector

Now take a system of **three equations in three unknowns**:

$$ \begin{aligned} a + \phantom{2}b + \phantom{2}c &= 10 \\ a + 2b + \phantom{2}c &= 15 \\ a + \phantom{2}b + 2c &= 12 \end{aligned} $$

Each equation is a dot product of its coefficient row with the unknowns $(a,b,c)$:

$$ (1,1,1)\cdot(a,b,c)=10, \quad (1,2,1)\cdot(a,b,c)=15, \quad (1,1,2)\cdot(a,b,c)=12. $$

The unknown vector $(a,b,c)$ is the **same** in all three, so we can **stack the coefficient rows into a matrix** and write the whole system as a single **matrix $\times$ vector** product:

$$ \begin{pmatrix} 1 & 1 & 1 \\ 1 & 2 & 1 \\ 1 & 1 & 2 \end{pmatrix} \begin{pmatrix} a \\ b \\ c \end{pmatrix} = \begin{pmatrix} 10 \\ 15 \\ 12 \end{pmatrix}. $$

> **A matrix-vector product is nothing more than several dot products stacked together** — one per row of the matrix.

```python
A = np.array([[1, 1, 1],
              [1, 2, 1],
              [1, 1, 2]])
x = np.array([3, 5, 2])   # the solution a=3, b=5, c=2

# A @ x is the three dot products stacked
print('A @ x =', A @ x, '  (= [10, 15, 12])')
print()
print('row-by-row (each a dot product):')
for i, row in enumerate(A):
    print(f'  row {i}: {row} . {x} = {row @ x}')
```

```python
# We can also SOLVE the system for the unknowns given the right-hand side
b = np.array([10, 15, 12])
solution = np.linalg.solve(A, b)
print('solution (a, b, c) =', solution)   # -> [3, 5, 2]
```

## 3. The dimension rule (and rectangular matrices)

For $A x$ to be defined, the **number of columns of $A$ must equal the length of $x$** — otherwise you'd be dot-producting vectors of different lengths, which is undefined.

The matrix need **not be square**: it can be **rectangular** as long as those inner dimensions match. Adding a fourth equation gives a $4 \times 3$ matrix that still multiplies a length-3 vector — and the result has **length 4, matching the number of rows**:

$$ \underbrace{(4 \times 3)}_{\text{matrix}} \; \underbrace{(3 \times 1)}_{\text{vector}} \;\longrightarrow\; \underbrace{(4 \times 1)}_{\text{result}}. $$

In general an $m \times n$ matrix times an $n$-vector produces an $m$-vector. The **inner dimensions must match**; the **outer dimensions** give the result's shape.

```python
# Rectangular: 4 x 3 matrix times a length-3 vector -> length-4 result
A4 = np.array([[1, 1, 1],
               [1, 2, 1],
               [1, 1, 2],
               [2, 1, 1]])   # a fourth equation
x = np.array([3, 5, 2])

print('A4 shape     :', A4.shape)          # (4, 3)
print('x  shape     :', x.shape)           # (3,)
print('A4 @ x       :', A4 @ x, ' shape', (A4 @ x).shape)   # length 4
```

```python
# Mismatched dimensions are an error: columns(3) != length(2)
try:
    A @ np.array([1, 2])
except ValueError as e:
    print('ValueError:', e)
```

## Summary

- A single linear equation is a **dot product** of a coefficient row with the vector of unknowns.
- A **system** of equations stacks those rows into a **matrix**, so the system becomes one **matrix $\times$ vector** product $Ax = b$ — a stack of dot products, one per row.
- **Dimension rule:** an $m \times n$ matrix times an $n$-vector gives an $m$-vector. The number of **columns** must equal the vector's **length**; the number of **rows** sets the result's length.
- Matrices may be **rectangular** — square isn't required, only the inner dimensions must match.
- In NumPy: `A @ x` computes the product; `np.linalg.solve(A, b)` solves $Ax = b$ for the unknowns.
