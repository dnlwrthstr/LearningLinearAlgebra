A motivating overview: the linear model we built in Weeks 1–2 is exactly the **building block of a neural network**. Week 3 develops the linear algebra — **vectors**, **matrices as linear transformations** — that makes this work.

## The thread

$$ \underbrace{W \cdot X + b = \hat{y}}_{\text{linear regression}} \;\longrightarrow\; \underbrace{\sigma(W \cdot X + b)}_{\text{one neuron}} \;\longrightarrow\; \underbrace{\text{layers of neurons}}_{\text{neural network}} $$

```python
import numpy as np
np.set_printoptions(precision=3, suppress=True)
```

## 1. Recap: linear regression is a system of linear equations

With $n$ features and $m$ examples, the model $w_1 x_1 + \dots + w_n x_n + b = y$ written for every example is a **system of linear equations**, compactly

$$ W \cdot X + b = \hat{y}, $$

a weight **vector** times a data **matrix**, plus a bias, giving the **vector** of predictions.

```python
# 3 features, 4 examples
X = np.array([[5.0, 6.0, 8.0, 4.0],
              [10.0, 15.0, 20.0, 8.0],
              [1.0, 0.0, 1.0, 1.0]])
W = np.array([2.0, -0.5, 3.0])
b = 1.0

y_hat = W @ X + b
print("predictions W . X + b =", y_hat)
```

## 2. The artificial neuron

An **artificial neuron** takes the same linear combination $W \cdot X + b$ and passes it through a non-linear **activation function** $\sigma$:

$$ a = \sigma(W \cdot X + b). $$

A common choice is the **sigmoid** $\sigma(z) = \dfrac{1}{1 + e^{-z}}$, which squashes any value into $(0, 1)$. The linear part is pure linear algebra; the activation adds the non-linearity that lets networks model complex relationships.

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

a = sigmoid(W @ X + b)
print("neuron output  sigma(W . X + b) =", a)
```

## 3. The neural network

Stack many neurons into **layers**. Each neuron in layer 1 computes its own $\sigma(W_i^{[1]} \cdot X + b_i^{[1]})$, producing the **activations** $a^{[1]}$. Those become the input to layer 2, and so on until the output $\hat{y}$:

$$ a^{[1]} = \sigma\!\left(W^{[1]} X + b^{[1]}\right), \qquad a^{[2]} = \sigma\!\left(W^{[2]} a^{[1]} + b^{[2]}\right), \qquad \dots $$

Every layer is a **matrix–vector product plus a bias, then an activation** — linear algebra repeated. Here is a forward pass through a small 2-layer network for one input.

```python
rng = np.random.default_rng(0)
x = np.array([0.5, -1.2, 2.0])          # one input with 3 features

# Layer 1: 4 neurons  ->  W1 is (4 x 3), b1 is (4,)
W1 = rng.normal(size=(4, 3)); b1 = rng.normal(size=4)
a1 = sigmoid(W1 @ x + b1)

# Layer 2 (output): 1 neuron  ->  W2 is (1 x 4), b2 is (1,)
W2 = rng.normal(size=(1, 4)); b2 = rng.normal(size=1)
y_out = sigmoid(W2 @ a1 + b2)

print("layer-1 activations a1 =", a1)
print("network output      y =", y_out)
```

## 4. The data structures: vectors, matrices, tensors

Everything above is stored in a small hierarchy of array types:

| Object | Rank | Example in a network |
|--------|------|----------------------|
| **vector** $\vec{a}$ | 1D | an input, a bias, a layer's activations |
| **matrix** | 2D | a layer's weights $W$, a batch of inputs |
| **tensor** | 3D+ | a batch of images, sequences, higher-order data |

Week 3 focuses on **vectors** and **matrices** — and, crucially, on seeing a matrix as a **linear transformation** that moves vectors around space.

```python
vector = np.array([1.0, 2.0, 3.0])
matrix = np.array([[1.0, 2.0], [3.0, 4.0]])
tensor = np.arange(2 * 2 * 3).reshape(2, 2, 3).astype(float)
for name, obj in [("vector", vector), ("matrix", matrix), ("tensor", tensor)]:
    print(f"{name:7s} ndim={obj.ndim}  shape={obj.shape}")
```

## 5. Takeaways

- Linear regression $W \cdot X + b = \hat{y}$ is the **artificial neuron's** linear core; adding an **activation** $\sigma$ makes it a neuron.
- A **neural network** stacks neurons into layers — each layer is a **matrix–vector product + bias + activation**, so the whole thing runs on linear algebra.
- Data lives in **vectors, matrices, and tensors**.
