# The Determinant

*Course notes for **Math for Machine Learning**, C1 · W1 · L1 — "The Determinant" (DeepLearning.AI).*

We've now seen three equivalent faces of singularity — no unique solution, dependent rows, rank deficiency. This lecture gives the single **number** that detects all of them: the **determinant**.

> $\det(A) = 0 \Leftrightarrow$ **singular** (dependent rows)  ·  $\det(A) \neq 0 \Leftrightarrow$ **non-singular** (independent rows).

```python
import numpy as np
np.set_printoptions(precision=3, suppress=True)
```

## 1. Where the 2×2 formula comes from

Take a matrix $\begin{bmatrix} a & b \\ c & d \end{bmatrix}$. It is **singular** when row 2 is a multiple $k$ of row 1:

$$ a k = c, \qquad b k = d \;\;\Longrightarrow\;\; \frac{c}{a} = \frac{d}{b} = k. $$

Cross-multiplying, $a d = b c$, i.e.

$$ \boxed{a d - b c = 0} \quad\text{for a singular matrix.}$$

That quantity is the **determinant**:

$$ \det\begin{bmatrix} a & b \\ c & d \end{bmatrix} = a d - b c. $$

It is zero exactly when the rows are dependent — which is why $\det = 0$ means singular.

```python
def det2(M):
    (a, b), (c, d) = M
    return a * d - b * c

# The two running examples.
non_singular = [[1, 1], [1, 2]]
singular     = [[1, 1], [2, 2]]
print(f"[[1,1],[1,2]]: det = 1*2 - 1*1 = {det2(non_singular)}  -> non-singular")
print(f"[[1,1],[2,2]]: det = 1*2 - 2*1 = {det2(singular)}  -> singular")
```

## 2. Quiz (2×2)

$$ M_1 = \begin{bmatrix} 5 & 1 \\ -1 & 3 \end{bmatrix} \qquad M_2 = \begin{bmatrix} 2 & -1 \\ -6 & 3 \end{bmatrix} $$

Find each determinant; are they singular or non-singular?

```python
for name, M in {"M1": [[5, 1], [-1, 3]], "M2": [[2, -1], [-6, 3]]}.items():
    d = det2(M)
    print(f"{name}: det = {d:3d}  ->  {'non-singular' if d != 0 else 'singular'}")
# M1: 5*3 - 1*(-1) = 15 + 1 = 16  (non-singular)
# M2: 2*3 - (-1)*(-6) = 6 - 6 = 0  (singular)
```

## 3. The 3×3 determinant — diagonals (Sarrus' rule)

For a 3×3 matrix, sum the products of the three **down-right** diagonals and subtract the products of the three **down-left** diagonals:

$$
\det\begin{bmatrix} a & b & c \\ d & e & f \\ g & h & i \end{bmatrix}
= \underbrace{aei + bfg + cdh}_{\searrow\ \text{add}} \;-\; \underbrace{ceg + bdi + afh}_{\swarrow\ \text{subtract}}.
$$

For $\begin{bmatrix}1&1&1\\1&2&1\\1&1&2\end{bmatrix}$ the lecture computes $(4+1+1) - (2+1+2) = 6 - 5 = 1$ → non-singular.

```python
def det3_sarrus(M):
    (a, b, c), (d, e, f), (g, h, i) = M
    down_right = a * e * i + b * f * g + c * d * h
    down_left  = c * e * g + b * d * i + a * f * h
    return down_right - down_left

A = [[1, 1, 1], [1, 2, 1], [1, 1, 2]]
print("Sarrus det      =", det3_sarrus(A))
print("numpy det       =", round(np.linalg.det(A)))   # cross-check

# An upper-triangular matrix: determinant is the product of the diagonal.
U = [[1, 1, 1], [0, 2, 2], [0, 0, 3]]
print("\nUpper-triangular [[1,1,1],[0,2,2],[0,0,3]] det =", det3_sarrus(U), "(= 1*2*3)")
```

## 4. Exercise: determinants confirm singularity

These are the four matrices from the *linear dependence* lecture. Compute each determinant and confirm the ones with $\det = 0$ are exactly the singular (dependent-row) ones.

$$
\begin{bmatrix}1&0&1\\0&1&0\\3&2&3\end{bmatrix}\quad
\begin{bmatrix}1&1&1\\1&1&2\\0&0&-1\end{bmatrix}\quad
\begin{bmatrix}1&1&1\\0&2&2\\0&0&3\end{bmatrix}\quad
\begin{bmatrix}1&2&5\\0&3&-2\\2&4&10\end{bmatrix}
$$

```python
problems = {
    "M1": [[1, 0, 1], [0, 1, 0], [3, 2, 3]],
    "M2": [[1, 1, 1], [1, 1, 2], [0, 0, -1]],
    "M3": [[1, 1, 1], [0, 2, 2], [0, 0, 3]],
    "M4": [[1, 2, 5], [0, 3, -2], [2, 4, 10]],
}
for name, M in problems.items():
    d = det3_sarrus(M)
    print(f"{name}: det = {d:3d}  ->  {'non-singular' if d != 0 else 'singular'}")
# Expected: M1=0 singular, M2=0 singular, M3=6 non-singular, M4=0 singular.
```

## 5. Takeaways

- The **determinant** is one number summarizing a square matrix; $\det = 0 \Leftrightarrow$ **singular**, $\det \neq 0 \Leftrightarrow$ **non-singular**.
- **2×2:** $\det = ad - bc$, derived directly from the condition that one row is a multiple of the other.
- **3×3:** Sarrus' rule — add the three down-right diagonal products, subtract the three down-left ones.
- A **triangular** matrix's determinant is just the product of its diagonal entries.
- This closes the Week-1 chain: *unique solution ⇔ independent rows ⇔ full rank ⇔ $\det \neq 0$ ⇔ non-singular.*

*In practice, use `numpy.linalg.det` (LU-based) for matrices larger than 3×3 — Sarrus' rule only works for 2×2 and 3×3.*
