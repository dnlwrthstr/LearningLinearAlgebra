# Calculating Eigenvalues and Eigenvectors

*Course notes for **Math for Machine Learning**, C1 · W4 · L2 · V04b — "Calculating Eigenvalues and Eigenvectors" (DeepLearning.AI).*

We know *what* eigenvalues and eigenvectors are — now *how* do we find them? The recipe uses a **determinant**:

1. Eigenvalues are the roots of the **characteristic polynomial** $\det(A - \lambda I) = 0$.
2. For each eigenvalue $\lambda$, the eigenvectors solve $(A - \lambda I)v = 0$.

We build the intuition, work a $2\times 2$ and a $3\times 3$ example, and note that only **square** matrices have eigenvalues at all.

```python
import numpy as np
from scipy.linalg import null_space

np.set_printoptions(precision=3, suppress=True)
```

## 1. Intuition: why a determinant appears

Compare $A = \begin{pmatrix} 2 & 1 \\ 0 & 3 \end{pmatrix}$ with the pure scaling $3I = \begin{pmatrix} 3 & 0 \\ 0 & 3 \end{pmatrix}$ (stretch everything by 3). These transformations are **different**, yet they act **identically on a whole line** (the $\lambda=3$ eigen-direction) — infinitely many points where they coincide.

If two transformations agree at infinitely many points, their **difference** $A - 3I$ sends infinitely many vectors to $0$. A matrix with infinitely many solutions to $Mv = 0$ is **singular** — so $\det(A - 3I) = 0$. The same holds for $\lambda = 2$.

```python
A = np.array([[2, 1],
              [0, 3]])
I = np.eye(2)
for lam in [2, 3]:
    M = A - lam * I
    print(f'A - {lam}I =\n{M}   det = {np.linalg.det(M):.0f}  -> singular')
    print()
```

## 2. The characteristic polynomial

In general, $\lambda$ is an **eigenvalue** exactly when $A - \lambda I$ is **singular**:

$$ \det(A - \lambda I) = 0. $$

Expanding this determinant gives the **characteristic polynomial** in $\lambda$; its **roots are the eigenvalues**. For our matrix:

$$ \det\begin{pmatrix} 2-\lambda & 1 \\ 0 & 3-\lambda \end{pmatrix} = (2-\lambda)(3-\lambda) = \lambda^2 - 5\lambda + 6 = 0 \;\Rightarrow\; \lambda = 2,\, 3. $$

```python
# np.poly(A) returns the characteristic polynomial's coefficients
coeffs = np.poly(A)
print('characteristic polynomial coeffs [1, -5, 6]  ->  lambda^2 - 5*lambda + 6')
print('coeffs =', coeffs)
print('roots (eigenvalues) =', np.roots(coeffs))
```

## 3. Finding the eigenvectors

With each eigenvalue in hand, the eigenvectors satisfy $A v = \lambda v$, i.e.

$$ (A - \lambda I)\,v = 0. $$

This is a (singular) homogeneous system — its non-zero solutions form the eigenvector direction (any multiple works, since **only the direction matters**). Solving:

- $\lambda = 2$: $(A - 2I)v = 0 \Rightarrow v = (1, 0)$.
- $\lambda = 3$: $(A - 3I)v = 0 \Rightarrow v = (1, 1)$.

```python
for lam in [2, 3]:
    v = null_space(A - lam * np.eye(2))[:, 0]   # basis of the null space
    v = v / v[np.argmax(np.abs(v))]              # normalize so a component is 1
    print(f'lambda = {lam}: eigenvector direction {v}   (check A v = {A @ v})')
```

## 4. Quiz: a $2\times 2$ example

Find the eigenvalues and eigenvectors of $\begin{pmatrix} 9 & 4 \\ 4 & 3 \end{pmatrix}$.

$$ \det\begin{pmatrix} 9-\lambda & 4 \\ 4 & 3-\lambda \end{pmatrix} = \lambda^2 - 12\lambda + 11 = (\lambda - 11)(\lambda - 1) = 0. $$

So $\lambda = 11$ and $\lambda = 1$, with eigenvectors $(2, 1)$ and $(-1, 2)$ respectively.

```python
B = np.array([[9, 4],
              [4, 3]])
print('char poly coeffs =', np.poly(B), ' -> lambda^2 - 12 lambda + 11')
print('eigenvalues      =', np.roots(np.poly(B)))
print()
for lam in [11, 1]:
    v = null_space(B - lam * np.eye(2))[:, 0]
    v = v / v[np.argmax(np.abs(v))]
    print(f'lambda = {lam:>2}: eigenvector direction {v}')
```

## 5. A $3\times 3$ example

The process is the same for larger square matrices. For

$$ A = \begin{pmatrix} 2 & 1 & -1 \\ 1 & 0 & -3 \\ -1 & -3 & 0 \end{pmatrix}, $$

the characteristic polynomial is $-\lambda^3 + 2\lambda^2 + 11\lambda - 12 = 0$, whose roots are $\lambda = -3,\, 1,\, 4$. Solving $(A - \lambda I)v = 0$ for each gives the eigenvectors (any scalar multiple is equally valid).

```python
A3 = np.array([[ 2,  1, -1],
               [ 1,  0, -3],
               [-1, -3,  0]])
print('char poly coeffs =', np.poly(A3))
print('eigenvalues      =', np.sort(np.roots(np.poly(A3))))
print()
for lam in [-3, 1, 4]:
    v = null_space(A3 - lam * np.eye(3))[:, 0]
    v = v / v[np.argmax(np.abs(v))]
    print(f'lambda = {lam:>2}: eigenvector direction {np.round(v, 3)}   (A v = {np.round(A3 @ v, 3)})')
```

## 6. Only square matrices have eigenvalues

The whole method hinges on a **determinant**, and the determinant is only defined for **square** matrices. So:

- **Square** matrix $\Rightarrow$ it has eigenvalues and eigenvectors.
- **Non-square** matrix $\Rightarrow$ **no** eigenvalues or eigenvectors (there's no $\det(A - \lambda I)$ to solve).

```python
nonsquare = np.array([[1, 2, 3],
                      [4, 5, 6]])   # 2 x 3
print('shape', nonsquare.shape, '- not square')
try:
    np.linalg.eig(nonsquare)
except np.linalg.LinAlgError as e:
    print('np.linalg.eig error:', e)
```

## Summary

- **Eigenvalues** are the roots of the **characteristic polynomial** $\det(A - \lambda I) = 0$ — because $\lambda$ is an eigenvalue exactly when $A - \lambda I$ is **singular**.
- **Eigenvectors** for each $\lambda$ solve the homogeneous system $(A - \lambda I)v = 0$; the solution is a direction, so any non-zero multiple is a valid eigenvector.
- Examples: $\begin{pmatrix}2&1\\0&3\end{pmatrix}\to\lambda\in\{2,3\}$; $\begin{pmatrix}9&4\\4&3\end{pmatrix}\to\{11,1\}$; the $3\times3\to\{-3,1,4\}$.
- Only **square** matrices have eigenvalues/eigenvectors (the determinant needs a square matrix).
- In NumPy: `np.poly(A)` / `np.roots(...)` for the characteristic polynomial, `scipy.linalg.null_space(A - lam*I)` for eigenvectors, or `np.linalg.eig(A)` for both at once.
