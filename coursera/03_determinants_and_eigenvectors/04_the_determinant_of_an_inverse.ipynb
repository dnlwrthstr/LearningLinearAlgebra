# The Determinant of an Inverse

*Course notes for **Math for Machine Learning**, C1 · W4 · L1 · V05 — "The Determinant of an Inverse" (DeepLearning.AI).*

A short, satisfying rule you can probably already guess:

$$ \boxed{\;\det(A^{-1}) = \dfrac{1}{\det(A)}\;} $$

The determinant of an inverse is the **inverse of the determinant**. We'll see it on examples, derive it from the product rule, and connect it to why singular matrices can't be inverted.

```python
import numpy as np

np.set_printoptions(precision=4, suppress=True)
```

## 1. Spotting the pattern

Recall two familiar matrices and their inverses:

| matrix | $\det$ | its inverse's $\det$ |
|---|---|---|
| $\begin{pmatrix} 3 & 1 \\ 1 & 2 \end{pmatrix}$ | $5$ | $0.2 = \tfrac15$ |
| $\begin{pmatrix} 5 & 2 \\ 1 & 2 \end{pmatrix}$ | $8$ | $0.125 = \tfrac18$ |

Each inverse's determinant is exactly **one over** the original — $\tfrac15$ and $\tfrac18$.

```python
for M in [np.array([[3, 1], [1, 2]]), np.array([[5, 2], [1, 2]])]:
    d = np.linalg.det(M)
    d_inv = np.linalg.det(np.linalg.inv(M))
    print(f'det(M) = {d:.1f}   det(M^-1) = {d_inv:.4f}   1/det(M) = {1/d:.4f}')
```

## 2. The singular case

The singular matrix $\begin{pmatrix} 1 & 1 \\ 2 & 2 \end{pmatrix}$ has $\det = 0$ and **no inverse** — mirroring the fact that the number **$0$ has no multiplicative inverse** ($1/0$ is undefined). The rule $\det(A^{-1}) = 1/\det(A)$ breaks exactly where it should.

```python
S = np.array([[1, 1], [2, 2]])
print('det(S) =', np.linalg.det(S), ' -> 1/det(S) undefined, no inverse exists')
try:
    np.linalg.inv(S)
except np.linalg.LinAlgError as e:
    print('np.linalg.inv error:', e)
```

## 3. Why it's true — from the product rule

It follows in three lines from $\det(AB) = \det(A)\det(B)$. Set $B = A^{-1}$:

$$ \det(A\,A^{-1}) = \det(A)\,\det(A^{-1}). $$

But $A\,A^{-1} = I$, the identity, and $\det(I) = 1$, so

$$ 1 = \det(A)\,\det(A^{-1}) \quad\Longrightarrow\quad \det(A^{-1}) = \frac{1}{\det(A)}. $$

```python
A = np.array([[3, 1], [1, 2]])
Ainv = np.linalg.inv(A)
print('det(A @ A^-1) =', round(np.linalg.det(A @ Ainv), 6), ' = det(I)')
print('det(A)*det(A^-1) =', round(np.linalg.det(A) * np.linalg.det(Ainv), 6))
print('both equal 1 -> det(A^-1) = 1/det(A)')
```

## 4. Why $\det(I) = 1$

The identity leaves the plane **unchanged**, so it scales area by a factor of $1$ — its determinant is $1$. Algebraically, for the $2\times2$ identity:

$$ \det\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = 1\cdot 1 - 0\cdot 0 = 1, $$

and the same holds for the $n \times n$ identity — its determinant is always $1$.

```python
for n in [2, 3, 5, 10]:
    print(f'det(I_{n:>2}) =', round(np.linalg.det(np.eye(n)), 6))
```

## Summary

- **$\det(A^{-1}) = 1/\det(A)$** — the determinant of the inverse is the reciprocal of the determinant.
- It follows directly from the **product rule**: $\det(A)\det(A^{-1}) = \det(A A^{-1}) = \det(I) = 1$.
- **$\det(I) = 1$** because the identity leaves area unchanged.
- A **singular** matrix ($\det = 0$) has **no inverse** — just as the number $0$ has no reciprocal.
